    <h1>فصل پنجم: یادگیری ماشین</h1>
    <h2>بخش اول: مقدمه‌ای بر یادگیری ماشین و آشنایی اولیه با انواع الگوریتم‌های آن</h2>
</span>

<p style="font-family:B Nazanin">
  تا اینجای  کار مفاهیم اولیه شامل پایتون، آمار و احتمالات، پیش‌پردازش داده‌ها و ... را یاد گرفتیم و اکنون باید از آن‌ها در یادگیری مفاهیمی مانند یادگیری ماشین و هوش مصنوعی استفاده کنیم. در این بخش ابتدا مفاهیم اولیه هوش مصنوعی و یادگیری ماشین را بررسی می‌کنیم.<br>
    <b>هوش مصنوعی:</b> هر تلاشی برای شبیه‌سازی فرایند هوش انسانی را هوش مصنوعی می‌گویند. به عبارت دیگر به هر کاری که بتوانیم انجام دهیم تا چیزی بسازیم که رفتارش شبیه به بخشی از هوش انسانی باشد را هوش مصنوعی می‌گویند. هوش مصنوعی کاربردها و گرایش‌های بسیار متنوعی دارد که بخشی از آن‌ها به شرح زیر هستند:<br>
        * یادگیری ماشین (Machine Learning)<br>
        * شناخت حالت (Computer Vision)<br>
        * پردازش زبان طبیعی (Natural Language Processing یا NLP)<br>
        * هوش عام (General AI)<br>
        * یادگیری تقویتی (Reinforcement Learning)<br>
        * هوش محاسباتی (Computational Intelligence)<br>
        * یادگیری عمیق (Deep Learning)<br>
        * بهینه‌سازی هوشمند (Intelligent Optimization)<br>
        * مدل‌سازی تولید متن (Generative Text Modeling)<br>
        * تحلیل داده‌های پیچیده (Complex Data Analysis)<br>
        * تفکیک الگوهای صوتی (Speech Pattern Recognition)<br>
        * اینترنت اشیا هوشمند (Smart Internet of Things)<br>
        * مدیریت رویدادها (Event Management)<br>
        * تشخیص تقلب (Fraud Detection)<br>
        * مهندسی دانش (Knowledge Engineering)<br>
        * هوش تجاری (Business Intelligence)<br>
        * مدل‌سازی تولید تصویری (Generative Image Modeling)<br>
        * مهندسی اطلاعات (Information Engineering)<br>
        * سامانه‌های تصمیم‌گیری خودکار (Automatic Decision Systems)<br>
        * ... <br>
    این سامانه‌ها مثل انسان فکر می‌کنند و مثل انسان عمل می‌کنند. اگر بتوانیم به جایی برسیم که مثل انسان کاملا عاقلانه فکر و عمل کنند، آن‌گاه می‌توانیم بگوییم توانسته‌ایم که یک سیستم هوش مصنوعی موفق بسازیم. 
</p>

<p style="font-family:B Nazanin">
    <b>تاریخچه هوش مصنوعی:</b> اولین بار فردی به نام Alan Turing در سال 1950 قصد داشت که کامپیوتری بسازد که هوشمند باشد. این فرد تستی را طراحی کرد که اگر از یک انسان و یک کامپیوتر یک سری فعالیت‌ها بخواهیم و نتوانیم تشخیص دهیم که این فعالیت‌ها را انسان انجام داده یا کامپیوتر، می‌توانیم بگوییم هوشمندسازی را انجام داده‌ایم.
</p>

<p style="font-family:B Nazanin">
   در سال 1955 آقای John McCarthy واژه هوش مصنوعی را ایجاد کرد و گفت به دستگاه‌هایی که بتوانند رفتار انسان را شبیه‌سازی کنند، هوش مصنوعی می‌گوییم. این فرد تلاش کرد که یک سیستم برای انجام بازی شطرنج بسازد که البته ناموفق بود. <br>
    در سال 1958 آقای Frank Rosenblatt تلاش کرد تا یک نورون مصنوعی به نام پرسپترون بسازد. نورون‌ها اجزای شبکه عصبی انسان هستند که پیام‌ّای عصبی را منتقل می‌کند و این فرد سعی کرد که یک نورون مصنوعی تولید کند که شبیه به نورون‌های داخل شبکه عصبی انسان کار کند. اینجا بود که شبکه عصبی اختراع شد و امروزه فرایند یادگیری عمیق توسط همین شبکه عصبی انجام می‌شود. <br>
    در دهه 1960 اولین ربات صنعتی ثابت (Unimate)، اولین Chatbot در حوزه درمان (ربات Eliza) و اولین ربات متحرک (Shakey) ساخته شد. <br>
    دهه 1970 به عنوان اولین زمستان تاریخ هوش مصنوعی بود که اتفاق خاصی در این دهه رخ نداد. <br>
    در دهه 1980 اولین سیستم خبره تولید شد. سیستم خبره سیستمی است که دانشی توسط یک سری متخصص به آن داده می‌شود و سپس این سیستم می‌تواند بر اساس یک سری قوانین از آن دانش استفاده کند و با سایر کاربران تعامل داشته باشد. <br>
    پس از آن مجددا تا اواسط دهه 1990 زمستان دوم هوش مصنوعی رقم خورد و اتفاق خاصی در این حوزه نیفتاد. <br>
    در سال 1997 یک ابرکامپیوتر به نام Deep Blue ساخته شد که توانست یکی از مطرح‌ترین شطرنج‌بازهای دنیا را شکست دهد. اما نکته مهم اینجا بود که این سیستم سخت‌افزار بسیار قدرتمندی داشت و الگوریتم‌های هوش مصنوعی آن به اندازه سخت‌افزار این سیستم تاثیرگذار نبودند. <br>
    در سال 2009 اولین خودروی خودران گوگل ساخته شد. <br>
    در سال 2013 دستیار صوتی Alexa توسط آمازون ساخته شد. <br>
    در سال 2014 چت بات Goostman توانست برای اولین بار تست تورینگ را پاسخ دهد و حدود 33 درصد داوران نتوانستند تشخیص دهند که این یک ماشین است. <br>
    در سال 2017 مقاله Attention is All You Need تحول بزرگی در یادگیری عمیق داشت که بسیاری معتقدند سال 2017 نقطه شروع یادگیری عمیق است. <br>
    در سال 2018 اولین نسخه GPT معرفی شد.
</p>

<p style="font-family:B Nazanin">
    <b>انواع هوش مصنوعی:</b><br>
    * Artificial Narrow Intelligence (ANI): یادگیری ماشین که در یک حوزه کار می‌کند و محدود است. هر کدام از هوش مصنوعی‌های این حوزه فقط در یک حوزه محدود کار می‌کنند و نمی‌توانند حوزه‌های مختلفی را پوشش دهند و مثل یک انسان کامل رفتار کنند. <br>
    * Artificial General Intelligence (AGI): ماشین هوشمندی است که می‌تواند به اندازه انسان هوشمند عمل کند. البته بسیاری از دانشمندان معتقدند که هنوز سال‌ها طول می‌کشد تا به این نقطه برسیم. <br>
    * Artificial Super Intelligence (ASI): هوش مصنوعی که بتواند فراتر از انسان کامل عمل کند و در همه حوزه‌ها بتواند بهتر از انسان عمل کند که مشخص نیست بتوانیم به این نقطه برسیم یا خیر.
</p>

<p style="font-family:B Nazanin">
    <b>تفاوت یادگیری ماشین (Machine Learning) و برنامه‌نویسی سنتی (Traditional Programming):</b> در برنامه‌نویسی سنتی ما یک سری قوانین مشخص می‌کردیم و بر اساس ورودی‌ها و آن قوانین، خروجی‌های مورد نظر را به دست می‌آوریم. برنامه ما فقط توانایی پاسخ به مسائلی را دارد که از قبل برای آن برنامه‌ریزی کرده‌ایم و برای آن قانون مشخص کرده‌ایم. اما در یادگیری ماشین بر اساس داده‌هایی که داریم، به برنامه خود آموزش می‌دهیم که چطورعمل کند.
</p>

<p style="font-family:B Nazanin">
    <b>تفاوت هوش مصنوعی و یادگیری ماشین:</b> هوش مصنوعی هر آن چیزی است که به صورت هوشمند می‌بینیم و یادگیری ماشین یک بخش از هوش مصنوعی است که یاد گرفته‌است با داده‌ها کار کند و برای ما کاری را انجام دهد. ضمنا شبکه‌های عصبی زیرمجموعه‌ای از یادگیری ماشین است که با شبیه‌سازی سیستم شبکه‌های عصبی انسان الگوریتم‌هایی را تعریف می‌کند و به مسائل پاسخ می‌دهد و یادگیری عمیق نیز زیرمجموعه‌ای از شبکه‌های عصبی است. 
</p>

<p style="font-family:B Nazanin">
   تفاوت‌های یادگیری ماشین (ML) و یادگیری عمیق (DL): <br>
    * DL نیاز به داده‌های زیادی برای آموزش دارد در حالی که ML با تعداد کمتری داده می‌تواند آموزش ببیند. <br>
    * DL برای انجام پردازش‌های خود به GPU نیاز دارد تا بتواند پردازش‌ها را به صورت موازی انجام دهد ولی ML از CPU استفاده می‌کند. <br>
    * DL دقت بالاتری دارد. <br>
    * آموزش DL زمان بیشتری نیاز دارد. <br>
    * DL امکانات بیشتری نسبت به ML در اختیار ما می‌گذارد.
</p>

<p style="font-family:B Nazanin">
    <b>مراحل کلی یادگیری ماشین:</b><br>
    <b>1. تعریف مسئله (Problem Definition):</b> مشخص کردن مسئله مهمترین کاری است که باید برای حل آن انجام دهیم. باید بفهمیم که ورودی و خروجی چیست و چه کاری باید انجام شود و نوع مسئله چیست؟ (خوشه‌بندی، طبقه‌بندی، رگرسیون و ...)<br>
    <b>2. تهیه داده‌ها (Data Acquisition):</b> از هر مدل یادگیری ماشین که استفاده کنیم، نیاز به داده داریم. باید داده‌های خود را جمع‌آوری کنیم تا بتوانیم با استفاده از آن‌ها هم به مدل خود آموزش دهیم  هم آن را تست کنیم. <br>
    <b>3. پیش‌پردازش داده‌ها (Data Preprocessing):</b> داده‌هایی که جمع‌آوری شده است ممکن است نیاز به پیش‌پردازش داشته باشد. مراحلی که در فصل‌های قبل گفته‌شد مانند حذف داده‌های تکراری، حذف نویزها، استخراج ویژگی‌ها و ...<br>
    <b>4. انتخاب مدل (Model Selection):</b> در این مرحله بر اساس نوع مسئله باید الگوریتمی را انتخاب کنیم که بهترین عملکرد را برای حل مسئله خواهد داشت.<br>
    <b>5. تقسیم داده (Data Splitting):</b> داده‌های موجود خود را باید به دو بخش تقسیم کنیم: Train Data برای آموزش مدل و Test Data برای تست و آزمایش مدل آموزش‌داده‌شده. اگر از کل داده‌ها برای آموزش مدل استفاده کنیم، نمی‌توانیم بفهمیم که مدل به درستی کار می‌کند یا نه. ممکن است آموزش مدل به درستی صورت نگیرد و برای داده‌های دیگر به خوبی کار نکند. بنابراین همیشه بخشی از داده‌ها را برای تست مدل نگه می‌داریم که مطمئن شویم آموزش مدل به درستی صورت گرفته است. <br>
    <b>6. آموزش مدل (Model Training):</b> با آن بخش از داده‌ها که برای آموزش در نظر گرفتیم، مدل را آموزش می‌دهیم. <br>
    <b>7. ارزیابی مدل (Model Evaluation):</b> با آن بخش از داده‌ها که برای تست در نظر گرفتیم، مدل آموزش‌داده‌شده خود را ارزیابی می‌کنیم. در این مرحله ممکن است متوجه شویم که Over Fitting اتفاق افتاده است. یعنی اگر همان داده‌های آموزش را به مدل بدهیم، کاملا دقیق و درست جواب دهد ولی اگر داده‌های تست یا داده‌های دیگر را به مدل بدهیم، جواب‌های نادرست تولید کند. <br>
    <b>8. تنظیم و بهینه‌سازی مدل (Model Tuning and Optimization):</b> در این مرحله با توجه به مشکلات و نقاط ضعفی که در مرحله ارزیابی متوجه شده‌ایم، با استفاده از تکنیک‌هایی مدل خود را بهینه‌سازی می‌کنیم. <br>
    <b>9. استفاده از مدل برای پیش‌بینی (Model Utilization):</b> در این مرحله کاربرد مدل در عمل را می‌بینیم که داده‌های واقعی را به مدل می‌دهیم تا مدل بتواند پاسخ را پیش‌بینی کند. <br>
    <b>10. توسعه و نگهداری (Development and Maintenance):</b> در گذر زمان باید بحث آموزش، ارزیابی، تنظیم و بهینه‌سازی را تکرار کنیم تا بتوانیم عملکرد بهتری از مدل بگیریم و آن را توسعه دهیم. 
</p>

<p style="font-family:B Nazanin">
   یادگیری ماشین به روش‌های مختلفی امکان‌پذیر است. ما می‌خواهیم یک ماشین را آموزش دهیم تا بتواند در آینده با کمک آموزش‌های داده‌شده به ما کمک کند. <br>
    <b>انواع الگوریتم‌های یادگیری ماشین:</b><br>
    <b>1. یادگیری تحت نظارت (Supervised Learning):</b> این روش ساده‌ترین روش است و بسیار پراستفاده است. در این روش به کمک داده‌هایی که ورودی و خروجی مشخصی دارند و از گذشته ما آن‌ها را داریم و می‌توانیم برچسب‌گذاری کنیم، مدل را آموزش می‌دهیم. به عبارت دیگر این روش معیار ما استفاده از داده‌های آموزشی دارای برچسب است. مثلا در پردازش تصویر، تعداد زیادی تصویر سگ را با ویژگی‌های مشخص به ماشین معرفی می‌کنیم و پس از آموزش یک تصویر جدید را به ماشین می‌دهیم و از آن می‌خواهیم که تصویر سگ را در آن تشخیص دهد. <br>
    <b>2. یادگیری بدون نظارت (Unsupervised Learning):</b> در این روش هیچ برچسبی وجود ندارد و فقط تعدادی داده به مدل داده می‌شود و خود مدل (ماشین) باید رفتارهای آن را تشخیص دهد. مثلا تعداد زیادی داده به مدل داده می‌شود و مدل رفتارهای عجیب و غیرعادی آن را تشخیص می‌دهد. مثلا شناسایی شماره حساب‌های نادرست از بین داده‌ها. در این روش ورودی و خروجی و برچسب نداریم و فقط تعدادی داده برای آموزش به مدل داده می‌شود و خود مدل رفتارهای غیرعادی را در بین داده‌ها پیدا می‌کند. <br>
    <b>3. یادگیری نیمه نظارتی (Semi-Supervised Learning):</b> در این روش نه کاملا همه داده‌ها برچسب دارند و نه مانند روش بدون نظارت همه داده‌ها بدون برچسب هستند. در این روش معمولا تعداد کمی داده برچسب‌دار به همراه تعداد زیادی داده بدون برچسب برای آموزش به ماشین می‌دهیم. <br>
    <b>4. یادگیری تقویتی (Reinforcement Learning):</b> در این روش که کاملا با روش‌های قبلی متفاوت است، سعی می‌کنند یک عامل را در یک محیط قرار دهند و بر اساس اقداماتی که آن عامل در آن محیط انجام می‌دهد و بابت انجام آن پاداش می‌گیرد یا مجازات می‌شود، یادگیری اتفاق می‌افتد. این روش پراستفاده است ولی معمولا هزینه زیادی دارد. 
</p>

<p style="font-family:B Nazanin">
    <b>یادگیری تحت نظارت:</b><br>
    در روش‌های یادگیری تحت نظارت مطابق با شکل زیر، تعدادی برچسب وجود دارد (مثلا در شکل زیر مثلث، دایره، مستطیل و ...) و تعدادی داده برچسب‌دار (مثلا شکل‌های مختلف که مشخص شده‌است کدام دایره است، کدام مثلث و ...). با استفاده از این داده‌ها و برچسب‌ها، مدل را آموزش می‌دهیم و سپس می‌توانیم با داده‌هایی مدل را تست کنیم. در نهایت از مدل آموزش‌داده‌شده استفاده می‌کنیم تا داده‌های واقعی را تشخیص دهد. 
</p>

<p style="font-family:B Nazanin">
   الگوریتم‌های یادگیری تحت نظارت خود به دو دسته کلی تقسیم می‌شوند:<br>
    1. اگر خروجی داده‌ها به صورت گسسته باشد (یعنی خروجی‌ها به تعداد محدود یا به صورت دسته‌بندی باشند)، از روش دسته‌بندی(Classification) استفاده می‌کنیم. مثلا تشخیص نوع حیوان مانند سگ و گربه در یک تصویر <br>
    2. اما اگر خروجی داده‌ها به صورت پیوسته باشد، از روش رگرسیون (Regression) استفاده می‌کنیم. مانند پیش‌بینی قیمت مسکن 
</p>

<p style="font-family:B Nazanin">
    <b>یادگیری بدون نظارت:</b><br>
    در این روش داده‌های بدون برچسب داریم. در این روش الگوریتم سعی می‌کند تا داده‌ها را به خوشه‌هایی (Cluster) تقسیم‌بندی کند. یعنی داده‌ها را بر اساس یک سری ویژگی مشترک در دسته‌های مشترک قرار می‌دهد. خروجی این روش همان خوشه‌ها یا Clusterها است.  
</p>

<p style="font-family:B Nazanin">
    الگوریتم‌های یادگیری بدون نظارت به چند دسته تقسیم‌بندی می‌شوند:<br>
    1. Clustering: در این روش (که بسیار شبیه به روش Classification در روش‌های تحت نظارت است) الگوریتم تلاش می‌کند تا داده‌های بدون برچسب را در خوشه‌های مختلف دسته‌بندی کند. (تفاوت این روش با classification این است که معیار اصلی در classification همان برچسب‌ها هستند درحالی که در clustering باید بر اساس خود داده‌ها این تفکیک صورت بگیرد.) <br>
    2. Association: کار این روش معمولا شناسایی رابطه و اتصال و قوانین داخلی داده‌ها است. مثلا تشخیص الگوهای خرید مشتریان یک فروشگاه. این روش بیشتر در حوزه داده‌کاوی استفاده می‌شود. <br>
    3. Dimentionality Reduction: در روش یادگیری مبتنی بر کاهش ابعاد، تلاش می‌کنیم تا تعداد ویژگی یا ابعاد داده‌ها را کم کنیم به شرطی که خود داده‌ها از بین نروند. 
</p>

<p style="font-family:B Nazanin">
    <b>یادگیری نیمه‌نظارتی:</b><br>
    در این الگوریتم‌ها، تعداد کمی داده برچسب‌دار به مدل داده می‌شود و در کنار آن تعداد زیادی داده بدون برچسب داریم که مدل سعی می‌کند بر اساس هر دو دسته داده رابط بین داده‌ها را کشف کند.
</p>

<p style="font-family:B Nazanin">
    <b>یادگیری تقویتی:</b><br>
    در فرایند یادگیری تقویتی عامل‌هایی (Agent) داریم که در یک محبط (Environment) قرار می‌گیرند و کارهایی (Action) را انجام می‌دهند و به ازای انجام هر کار پاداش (Reward) می‌گیرند یا مجازات می‌شوند. بنابراین الگوریتم بر اساس پاداش‌ها و مجازات‌های دریافت کرده یاد میگیرد که چه کارهایی را باید انجام دهد و چه کارهایی را نباید انجام دهد.
</p>
