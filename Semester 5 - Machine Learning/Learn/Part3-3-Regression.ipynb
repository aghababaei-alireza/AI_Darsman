<p style="font-family:B Nazanin">
    <b>مسائل طبقه‌بندی (Classification):</b><br>
    نکته قابل توجه: همان طور که قبلا توضیح داده شد، الگوریتم‌های یادگیری تحت نظارت به دو دسته رگرسیون و طبقه‌بندی تقسیم می‌شوند. اما دلیل اینکه بحث طبقه‌بندی در این فصل بیان می‌شود این است که بعضی از مسائل طبقه‌بندی توسط الگوریتم‌های رگرسیون مدل‌سازی می‌شوند. پس در این بخش مسائلی از مبحث طبقه‌بندی مطرح می‌شود که توسط الگوریتم‌های رگرسیون مدل‌سازی می‌شوند و مابقی مباحث در فصل مربوط به خود طبقه‌بندی بیان می‌شوند.<br>
    در الگوریتم‌های رگرسیون هدف این بود که با بررسی داده‌های ورودی، یک متغیر عددی پیوسته را پیش‌بینی کنیم. مثلا پیش‌بینی قیمت مسکن یا پیش‌بینی قیمت طلا یا پیش‌بینی میزان فروش یک محصول. اما گاهی مسائلی وجود دارد که باید بر اساس داده‌های ورودی پیش‌بینی کنیم که یک متغیر مربوط به کدام دسته‌بندی است. یعنی متغیر هدف ما یک متغیر گسسته و دسته‌بندی‌شده است. مثلا اطلاعات یک حیوان را داریم و می‌خواهیم مشخص کنیم که این حیوان سگ است یا گربه یا ... . به عبارت دیگر در مسائل طبقه‌بندی یا Classification متغیر هدف از نوع داده‌های اسمی یا Nominal است. <br>
    در این نوع مسائل، گاهی داده‌های اسمی ما قابل مرتب‌سازی هستند، مانند مدرک تحصیلی (که ترتیب دارد: دیپلم، لیسانس، فوق لیسانس،‌دکتری و ...) یا به صورت کلی بدوت ترتیب هستند و قابل مرتب‌سازی نمی‌باشند، مانند نوع حیوانات (سگ و گربه و گرگ و ...). 
</p>

<p style="font-family:B Nazanin">
    <b>انواع مدل‌های طبقه‌بندی:</b><br>
    * Binary Classification: در این نوع دسته‌بندی، متغیر هدف به طور کلی در دو دسته قرار می‌گیرد. مثلا اطلاعات یک فرد را داریم و می‌خواهیم مشخص کنیم که بیمار است یا سالم. یا اطلاعات یک محصول را داریم و می‌خواهیم مشخص کنیم که این محصول خریداری می‌شود یا خیر. <br>
    * Multiclass Classification: اگر تعداد دسته‌هایی که متغیر هدف می‌تواند در آن قرار گیرد، بیش از دو دسته باشد، در این نوع قرار می‌گیرد. مثلا اطلاعات یک میوه را داشته باشیم و بخواهیم مشخص کنیم که نوع میوه چیست. یا با داشتن اطلاعات یک خبر، مشخص کنیم جزء چه دسته‌ای (سیاسی، ورزشی، فرهنگی و ...)‌است.  
</p>

<p style="font-family:B Nazanin">
    <b>مدل Multi Label Classification:</b><br>
    گاهی اوقات در بررسی داده‌ها، به داده‌هایی می‌رسیم که ممکن است در دو دسته مختلف قرار می‌گیرد. یعنی نمی‌توان آن داده را فقط در یک دسته قرار داد. بنابراین به این موارد MultiLabel Classification می‌گویند که در آن یک داده می‌تواند متعلق به بیش از یک دسته‌بندی باشد. به عنوان مثال در تصویر زیر در عکس سمت چپ، تصویر یک سگ را می‌بینیم که طبق پیش‌بینی این تصویر نیز در دسته‌بندی سگ قرار گرفته است اما در عکس سمت راست، تصویر در هر دو دسته سگ و گربه قرار گرفته است.  
</p>

<p style="font-family:B Nazanin">
     برای حل چالش مدل‌های multi Label Classification، الگوریتم‌ها و استراتژی‌های متفاوتی وجود دارد.
</p>

<p style="font-family:B Nazanin">
     در این بین سه روش Binary Relevance، Label Powerset و Classifier Chains بیشتین کاربرد را دارد.<br>
    در روش Binary Relevance باید به تعداد دسته‌های موجود، مدل داشته باشیم. یعنی مثلا مطابق با شکل زیر ما چهار دسته‌بندی $Y_1$ تا $Y_4$ داریم که داده‌ها می‌تواند در این دسته‌بندی‌ها قرار گیرد. در این روش ما یک بار مدلی را آموزش می‌دهیم که بررسی می‌کند که هر ورودی در دسته $Y_1$ قرار می‌گیرد یا خیر. یک بار مدلی را آموزش می‌دهیم که بررسی می‌کند که هر ورودی در دسته $Y_2$ قرار می‌گیرد و الی آخر. مثلا در اینجا باید چهار مدل مختلف داشته باشیم. در نهایت برای بررسی هر داده، باید هر چهار مدل را بررسی کنیم و در آخر بگوییم که این داده در کدام یک از دسته‌ها (ممکن است بیش از یک دسته باشد) قرار می‌گیرد.  
</p>

<p style="font-family:B Nazanin">
     در روش Classifier Chains برچسب‌ها به صورت زنجیره‌ای مرتب می‌شوند. سپس به ترتیب ابتدا مشخص می‌شود که آن داده ورودی (نمونه) در دسته اول قرار می‌گیرد یا خیر. سپس نتیجه این مدل در مدلی که برای دسته دوم ایجاد شده است، استفاده می‌شود. یعنی داده ورودی اولیه و خروجی مدل اول، به عنوان ورودی‌های مدل دوم استفاده می‌شوند. به همین ترتیب ورودی اولیه و خروجی مدل‌های اول و دوم به عنوان ورودی مدل سوم استفاده می‌شوند الی آخر. 
</p>

<p style="font-family:B Nazanin">
     در روش Label Powerset ابتدا تمام حالات ممکن را با در نظر گرفتن دسته‌ها (برچسب‌ها) در نظر می‌گیریم (تمام جایگشت‌های ممکن دسته‌بندی‌ها). سپس هر حالت را به صورت یک مدل Multiclass حل می‌کنیم. مثلا اگر سه دسته‌بندی سگ و گربه و گرگ داشتیم، در کل 8 حالت مختلف داریم. در اینجا به ظاهر تعداد برچسب‌ها زیاد می‌شود اما چالش Multi Label حل می‌شود. البته این روش اگر تعداد دسته‌ها زیاد باشد کاربرد ندارد و مورد استفاده قرار نمی‌گیرد.
</p>

<p style="font-family:B Nazanin">
    <b>تعادل در توزیع داده‌ها (Balanced and Imbalanced Data):</b><br>
    چالش دیگری که چه در مدل‌های multi class و چه در مدل‌های multi label وجود دارد، عدم تعادل در توزیع داده‌ها است. در برخی از دیتاست‌ها، فراوانی داده‌های متعلق به هر دسته تقریبا متعادل است. یعنی به عنوان مثال اگر اطلاعات تعدادی از افراد را داشته باشیم و بخواهیم زن یا مرد بودن آن‌ها را تشخیص بدهیم، تقریبا می‌توانیم بگوییم که نزدیک به نیمی از داده‌ها می‌تواند متعلق به مردان و نزدیک به نیمی از داده‌ها می‌تواند متعلق به زنان باشد. اما در بررسی ژن‌های نرمال و غیرنرمال، این طور نیست. مثلا ممکن است 90 درصد ژن‌های مورد بررسی نرمال باشند و فقط 10 درصد غیر نرمال. یا مثلا در بررسی تراکنش‌های انجام‌شده در یک شبکه بانکی ممکن است در طول یک روز تعداد بسیار کمی از تراکنش‌ها تقلبی باشد. این عدم توازن و عدم تعادل در توزیع داده‌های دسته‌بندی‌های مختلف در امر آموزش مدل ما را دچار مشکل می‌کند که در بخش‌های بعدی روش‌های بررسی و برطرف کردن این چالش را بیان می‌کنیم.
</p>

<p style="font-family:B Nazanin">
    <b>ماتریس آشفته یا ماتریس درهم‌ریختگی (Confusion Matrix):</b><br>
    ماتریس درهم‌ریختگی یک ماتریس $n\times n$ است (که در آن $n$ تعداد حالت‌های مدل classification است. مثلا اگر یک مدل باینری داشته باشیم، این ماتریس $2\times 2$ است). فرض کنید برای یک مدل باینری دو کلاس Positive و Negative داریم. مطابق با تصویر زیر اگر یک داده می‌بایست در دسته Positive قرار گیرد و مدل هم همین دسته Positive را پیش‌بینی کند، به این حالت True Positive می‌گوییم؛ یعنی به درستی در دسته Positive قرار گرفته است. اگر داده باید در دسته Positive باشد ولی مدل به اشتباه در دسته Negative پیش‌بینی کند، به این حالت False Negative می‌گوییم. در این حالت می‌گویند خطای نوع 2 (Type II Error) رخ داده است. اگر داده باید در دسته Negative قرار گیرد ولی به اشتباه در دسته Positive پیش‌بینی می‌شود، به این حالت False Positive می‌گویند و خطای نوع 1 (Type I Error) رخ داده است و در نهایت اگر داده در واقعیت Negative باشد و مدل نیز همین Negative را پیش‌بینی کند، به این حالت True Negative می‌گویند.
</p>

<p style="font-family:B Nazanin">
    فرض کنید مدلی داریم که دارای دو دسته‌بندی 0 (که در اینجا Negative در نظر می‌گیریم) و 1 (که در اینجا Positive در نظر می‌گیریم) است و 6 داده تست داریم که این مدل به صورت زیر پیش‌بینی کرده است:
</p>

TN = True Negative $\qquad$
TP = True Positive $\qquad$
FN = False Negative $\qquad$
FP = False Positive

</table>

<p style="font-family:B Nazanin">
    اگر تعداد هر حالت را بشماریم، می‌توانیم ماتریس درهم‌ریختگی را تشکیل دهیم.
</p>

TN = 1 $\qquad$ TP = 2 $\qquad$ FN = 2 $\qquad$ FP = 1

<p style="font-family:B Nazanin">
    <b>معیارهای ارزیابی در Classification:</b><br>
    همانند رگرسیون که از معیارهایی همچون R2 یا MSE برای ارزیابی مدل آموزش داده شده استفاده می‌کردیم، در مدل‌های Classification از معیارهای زیر استفاده می‌کنیم:<br>
    <b>* نرخ دقت (نرخ کلاس) یا Accuracy:</b> دقت یک مدل برابر است با نسبت تعداد نمونه‌های درست‌تشخیص‌داده‌شده به کل نمونه‌ها.
</p>

$$
Accuracy = \frac{TP + TN}{TP + TN + FP + FN}
$$

$$
Error\ rate = 1 - Accuracy
$$

<p style="font-family:B Nazanin">
    <b>* نرخ کامل بودن (نرخ پاسخ‌های مثبت) یا Sensitivity یا True Positive Rate یا Recall:</b> در این معیار داده‌هایی مهم است که در واقعیت Positive باشند و با در نظر گرفتن این داده‌ها برابر است با داده‌های Positive درست تشخیص داده‌شده نسبت به کل داده‌های Positive.
</p>

$$
Recall = \frac{TP}{TP + FN}
$$

<p style="font-family:B Nazanin">
    این معیار زمانی مفید است که می‌خواهیم تعداد اشتباهات منفی را کاهش دهیم. <br>
    <b>* نرخ صحت یا Precision:</b> در این معیار داده‌هایی که Positive پیش‌بینی شده‌اند مهم است و برابر است با تعداد داده‌های Positive درست تشخیص داده‌شده نسبت به کل داده‌هایی که positive تشخیص داده شده‌اند.
</p>

$$
Precision = \frac{TP}{TP + FP}
$$

<p style="font-family:B Nazanin">
    <b>* معیار F1:</b> ترکیبی از معیارهای قبل است که به نوعی میانگین هندسی معیارهای Recall و Precision است که از طریق دو فرمول زیر محاسبه می‌شود.
</p>

$$
F1\ score = \frac{TP}{TP + \frac{1}{2}(FP + FN)} = \frac{2 \times Precision \times Recall}{Precision + Recall}
$$

<p style="font-family:B Nazanin">
    مباحث گفته‌شده برای یک مدل باینری بود. برای مدل‌های MultiClass که بیش از دو دسته‌بندی دارند، یکی از کلاس‌ها یا دسته‌بندی‌ها به عنوان کلاس Positive یا کلاس هدف و مابقی کلاس‌ها یا دسته‌بندی‌ها به عنوان کلاس Negative تلقی می‌شوند. در این صورت ماتریس درهم‌ریختگی به صورت زیر در می‌آید: (در شکل زیر کلاس $c_k$ به عنوان کلاس Positive در نظر گرفته شده است)
</p>

<p style="font-family:B Nazanin">
    با این تعاریف، معیارهای ارزیابی را می‌توان با همان فرمول‌های قبلی به دست آورد. در مثال زیر دیتاست معروف diabetes بررسی می‌شود.
</p>

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, recall_score, precision_score, f1_score, classification_report
```

```python
df = pd.read_csv('./datasets/diabetes.csv', index_col=False)
df.shape
```

```python
df.head()
```

<p style="font-family:B Nazanin">
    ستون یا متغیر هدف در این دیتاست ستون Outcome است که مشخص می‌کند یک فرد بیماری دیابت دارد یا خیر. در این مثال می‌خواهیم فقط تاثیر میزان گلوکز (ستون Glucose) را در داشتن دیابت بررسی کنیم. پس فقط دو ستون Glucose و Outcome را در نظر می‌گیریم. همچنین نام ستون Outcome را نیز تغییر می‌دهیم تا خوانایی بیشتری داشته باشد.
</p>

```python
df = df[['Glucose', 'Outcome']].copy().rename(columns={'Outcome':'Diabete'})
df.head()
```

```python
df.shape
```

<p style="font-family:B Nazanin">
    در حل مثال‌های واقعی هر چه تعداد داده‌ها بیشتر باشد، احتمالا دقت مدل بیشتر است اما در اینجا برای اینکه بتوانیم به صورت شهودی داده‌ها را ببینیم، فقط 100 داده نمونه را انتخاب می‌کنیم تا آن‌ها را بررسی کنیم.
</p>

```python
df = df.sample(100)
df.shape
```

```python
X = df[['Glucose']]
y = df['Diabete']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

<p style="font-family:B Nazanin">
    در اینجا برای آموزش داده‌ها از مدل Logistic Regression استفاده می‌کنیم که آموزش کامل آن را در بخش‌های بعدی می‌بینیم.
</p>

```python
lrModel = LogisticRegression()
lrModel.fit(X_train, y_train)
```

```python
y_pred = lrModel.predict(X_test)
```

```python
data = {
    'y_Actual': y_test,
    'y_Pred': y_pred
}
df = pd.DataFrame(data, columns=['y_Actual', 'y_Pred'])
df
```

<p style="font-family:B Nazanin">
    ماتریس Confusion را می‌توان به صورت زیر محاسبه کرد:
</p>

```python
confusion_matrix(df['y_Actual'], df['y_Pred'])
```

```python
accuracy_score = accuracy_score(df['y_Actual'], df['y_Pred'])
accuracy_score
```

```python
recall_score = recall_score(df['y_Actual'], df['y_Pred'])
recall_score
```

```python
precision_score = precision_score(df['y_Actual'], df['y_Pred'])
precision_score
```

```python
f1_score = f1_score(df['y_Actual'], df['y_Pred'])
f1_score
```

```python
classification_report = classification_report(df['y_Actual'], df['y_Pred'])
print(classification_report)
```

<p style="font-family:B Nazanin">

</p>

```python
pd.crosstab(df['y_Actual'], df['y_Pred'], rownames=['Actual'], colnames=['Predicted'])
```
