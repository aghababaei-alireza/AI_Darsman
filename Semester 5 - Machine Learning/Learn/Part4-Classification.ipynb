<p style="font-family:B Nazanin">
     در بخش قبلی ابتدا الگوریتم‌های رگرسیون را بررسی کردیم و سپس الگوریتم‌های رگرسیون که برای حل مسائل طبقه‌بندی استفاده می‌شدند را بیان کردیم. در این فصل به طور کامل الگوریتم‌های طبقه‌بندی را بررسی می‌کنیم. البته در این بین برخی از این الگوریتم‌ها نیز برای رگرسیون به کار می‌روند که آن‌ها را نیز می‌بینیم.<br>
    <b>الگوریتم جنگل تصادفی یا Random Forest:</b><br>
    این الگوریتم بر پایه درخت تصمیم کار می‌کند. یعنی الگوریتم Random Forst از چندین درخت تصمیم تشکیل شده است. اولین بار در سال 2001 معرفی شد و روش کار آن بر پایه یادگیری جمعی است. همان طور که در شکل زیر مشخص است، یکی از روش‌های یادگیری ماشین به این صورت است که در آن چندین مدل پایه یادگیری (Base Model) با یکدیگر ترکیب می‌شوند تا بتوانند یک مدل قوی‌تر و دقیق‌تر را ایجاد کنند. به این روش یادگیری جمعی یا Ensemble Learning می‌گویند. هدف این روش بهبود دقت پیش‌بینی است که در نهایت موجب کاهش واریانس و خطا نیز می‌شود. در این روش مدل‌ها یکی پس از دیگری اجرا می‌شوند و هر مدل سعی می‌کند تا خطاهای مدل قبلی را برطرف کند. به این ترتیب پایداری این روش بالاتر می‌رود و خطای آن کاهش پیدا می‌کند.  
</p>

<p style="font-family:B Nazanin">
    <b>انواع روش‌های یادگیری جمعی:</b><br>
    * <b>Boosting:</b> در این روش مدل‌ها به ترتیب اجرا می‌شوند و هر مدل سعی دارد که خطاهای مدل قبلی را برطرف کند. بدین تطتیب در گذر زمان و با اجرای تمامی مدل‌ها به دقت بالاتر و خطای کمتری می‌رسیم. <br>
    * <b>Stacking:</b> در این روش بر خلاف روش قبلی، مدل‌ها به صورت موازی و هم‌زمان اجرا می‌شوند و نتیجه مدل‌ها به یک مدل بالادستی داده می‌شود تا تصمیم‌گیری نهایی انجام شود. <br>
    * <b>Bagging یا Bootstrap Aggregating:</b> این روش که در فهم الگوریتم Random Forest به کار می‌آید، از دو مرحله Bootstrap Sampling و Aggregating تشکیل شده است. در مرحله اول داده‌های آموزشی به روش Bootstrap Sampling به چند زیرمجموعه تقسیم می‌شود. روش Bootstrap Sampling به این صورت است که نمونه انتخاب‌شده باید با نمونه اصلی هم‌اندازه باشد با این تفاوت که ممکن است داده‌های تکراری انتخاب شوند. سپس هر مجموعه داده‌های آموزشی به یک مدل داده می‌شود تا آن مدل را آموزش دهد و در نهایت هر مدل مقدار نهایی را پیش‌بینی می‌کند. مدل‌های پایه هر کدام می‌تواند از یک نوع باشند. در آخر مقادیر پیش‌بینی‌شده مدل‌ها با یکدیگر تجمیع شده و پیش‌بینی نهایی را ایجاد می‌کند. تجمیع نتایج به این صورت است که اگر مسئله طبقه‌بندی باشد، کلاس‌های پیش‌بینی‌شده مدل‌ها را بررسی می‌کند و کلاسی را که رای اکثریت داشته باشد، انتخاب می‌کند. اما اگر مسئله از نوع رگرسیون باشد، نتایج به‌دست‌آمده که از نوع عددی هستند را میانگین می‌گیرد و میانگین به‌دست‌آمده به عنوان نتیجه نهایی محسوب می‌شود.  
</p>

<p style="font-family:B Nazanin">
      در الگوریتم Random Forest نیز می‌توانیم با استفاده از همین روش Bagging چندین درخت تصمیم را با استفاده از داده‌های آموزشی، آموزش دهیم و در نهایت پیش‌بینی مدل‌ها را با هم تجمیع کنیم. مثلا در شکل زیر درخت اول و دوم به این نتیجه رسیده‌اند که شکل مشخص‌شده یک سیب است و درخت سوم موز را پیش‌بینی کرده است. در نهایت رای اکثریت این است که تصویر مربوط به یک سیب است. 
</p>

<p style="font-family:B Nazanin">
    <b>مفهوم انتخاب ویژگی‌های تصادفی یا Random Feature Selection:</b> با توجه به این که در الگوریتم Random Forest از تعدادی درخت تصمیم تشکیل می‌شود، اگر بخواهیم هر درخت را با تمام ویژگی‌ها بررسی کنیم و از تمام ویژگی‌ها برای ایجاد گره‌های درخت استفاده کنیم، درخت‌ها بسیار بزرگ می‌شوند و هزینه محاسبات و زمان انجام آن بسیار زیاد می‌شود. بنابراین مبحث انتخاب ویژگی‌های تصادفی مطرح می‌شود. به این صورت که به جای این که در هر درخت از تمام ویژگی‌ها استفاده کنیم، تعدادی از ویژگی‌ها (مثلا لگاریتم کل ویژگی‌ها یا جذر کل ویژگی‌ها) را استفاده کنیم و درخت را تشکیل دهیم. در این صورت هزینه محاسبات و زمان اجرا کاهش می‌یابد. 
</p>

<p style="font-family:B Nazanin">
      از مزیت‌های این روش است که همبستگی بین درخت‌ها کاهش می‌یابد و درخت‌ها تقریبا از یکدیگر مستقل می‌شوند. به این دلیل که ویژگی‌ها به صورت تصادفی انتخاب می‌شوند و ویژگی‌هایی که در یک درخت استفاده می‌شوند ممکن است با ویژگی‌های استفاده شده در درخت دیگر متفاوت باشند. مزیت دیگر این است که باعث کاهش واریانس می‌شود. وقتی هر درخت بر اساس مجموعه متفاوتی از ویژگی‌ها آموزش می‌بینند، واریانس کلی کم می‌شود و مدل نهایی در برابر نویز قوی‌تر است و عملکرد بهتری دارد. از طرف دیگر تنوع مدل‌ها و درخت‌ها بیشتر می‌شود چون هر درخت بر اساس ویژگی‌های متفاوتی ساخته می‌شود که خود این کار باعث افزایش پیچیدگی مدل و کاهش overfitting می‌شود. <br>
    در مقابل یکی از معایب این روش این است که چون از چندین درخت استفاده می‌کنیم، هزینه آموزش و انجام محاسبات و همچنین زمان اجرای پردازش‌ها زیاد است. علاوه بر آن تعداد هایپرپارامترها زیادتر می‌شود و انتخاب مقدار مناسب برای آن‌ها نیز سخت‌تر می‌شود و چون تعداد آن‌ها زیاد است در صورت استفاده از روش‌هایی مانند GridSearch یا RandomizedSearch زمان بسیار زیادی صرف می‌شود. <br>
    اما در کل در مدل‌های بزرگ و حساس معمولا صرف این زمان و هزینه برای ساخت این مدل ارزشمند و به‌صرفه است تا با ساخت یک مدل دقیق‌تر، نتایج بهتر و کارآمدتر به‌دست آوریم.
</p>

<p style="font-family:B Nazanin">
    <b>هایپرپارامترهای الگوریتم Random Forest:</b><br>
    * n_estimators: تعداد درخت‌هایی که در جنگل تصادفی ساخته می‌شود. هر چه تعداد درختان بیشتر شود، دقت مدل بالاتر می‌رود و نتایج دقیق‌تر است اما هزینه محاسبات بالاتر می‌رود. <br>
    * criterion: معیار استفاده شده برای اندازه‌گیری میزان خلوص (کیفیت) تقسیم در هر گره (مشابه با درخت تصمیم می‌تواند gini یا entropy باشد.)<br>
    * max_depth: حداکثر عمق درخت را مشخص می‌کند که می‌تواند به جلوگیری از overfitting کمک کند. <br>
    * min_samples_split: حداقل تعداد نمونه‌های مورد نیاز جهت تقسیم یک گره داخلی<br>
    * min_samples_leaf: حداقل تعداد نمونه‌های مورد نیاز برای تشکیل یک گره برگ. <br>
    * max_features: حداکثر تعداد ویژگی‌هایی که برای جست‌وجوی بهترین تقسیم در هر گره استفاده می‌شود. <br>
    * bootstrap: آیا نمونه‌گیری با بازگشت از داده‌های آموزشی برای ساختن هر درخت انجام شود یا خیر. (آیا از داده‌های تکراری در نمونه‌گیری استفاده کنیم یا خیر) که جواب مثبت است. <br>
    * oob_score: آیا از نمونه‌های خارج از کیف (out of bag: نمونه‌هایی که در یک بوت‌استرپ انتخاب نمی‌شوند) برای برآورد دقت مدل استفاده شود یا خیر. به صورت پیش‌فرض False است. <br>
    * max_leaf_nodes: حداکثر تعداد گره‌های برگ در هر درخت است که می‌تواند به کنترل اندازه درخت کمک کند. <br>
    * min_impurity_decrease: حداکثر کاهش ناخالصی لازم برای تقسیم یک گره است که برای تنظیم دقت تقسیمات استفاده می‌شود. <br>
    * n_jobs: تعداد پردازنده‌هایی است که برای اجرای کار استفاده می‌شود. (مقدار -1 به معنای استفاده از همه پردازنده‌ها است.)
</p>

<p style="font-family:B Nazanin">
      تابع BinaryEncoder به این صورت عمل می‌کند که ابتدا تعداد مقادیر یکتا در هر ستون را پیدا می‌کند و به هر مقدار یک عدد از یک به بالا نسبت می‌دهد. سپس تعداد کل مقادیر در یک ستون را به مبنای 2 می‌برد و تعداد رقم‌های آن را پیدا می‌کند و به همان تعداد، ستون ایجاد می‌کند و عدد متناظر با هر مقدار را در آن ستون‌ها در مبنای 2 می‌نویسد. مثلا در اینجا در ستون buying چهار مقدار متمایز داریم. تمامی مقادیری که در این ستون وجود دارند را با اعداد یک تا چهار دسته‌بندی می‌کنیم. اعداد از یک تا چهار را در مبنای دو می‌توانیم با سه رقم نشان دهیم. پس سه ستون ایجاد می‌کنیم. حال اگر مقدار متناظر با یک نمونه برابر با 3 بود که در مبنای دو می‌شود 011، در این سه ستون رقم‌های 0 و 1 و 1 را به ترتیب می‌نویسیم.

```python
X_test_3
```

<p style="font-family:B Nazanin">
      مزیت روش BinaryEncoder نسبت به OneHotEncoder این است که تعداد ستون‌ها به شدت کاهش می‌یابد. مثلا اگر در یک ستون 12 مقدار مختلف داشتیم، در حالت OneHotEncoder باید 12 ستون بسازیم ولی در حالت BinaryEncoder فقط چهار ستون می‌سازیم. (12 در مبنای دو می‌شود 1100 که چهار رقم دارد.)<br>
    برای ادامه کار از روش OrdinalEncoder استفاده می‌کنیم.
</p>

```python
import category_encoders as ce
oe = ce.OrdinalEncoder()
X_train = oe.fit_transform(X_train)
X_test = oe.transform(X_test)
X_train , X_test
```

```python
rfc = RandomForestClassifier(random_state=0)
rfc.fit(X_train, y_train)
y_pred = rfc.predict(X_test)

print("confusion_matrix:\n", confusion_matrix(y_test, y_pred))
print("accuracy_score:", accuracy_score(y_test, y_pred))
print("recall_score:", recall_score(y_test, y_pred,average='micro'))
print("precision_score:", precision_score(y_test, y_pred,average='micro'))
print(f"\nClassification Report:\n{classification_report(y_test, y_pred )}")
```

<p style="font-family:B Nazanin">
      مقدار هایپرپارامتر n_estimators به صورت پیش‌فرض 100 است. در اینجا برای سرعت بیشتر این مقدار را روی 5 تنظیم می‌کنیم و می‌بینیم که مقدار دقت نتایج کاهش می‌یابد. (مقدار accuracy از 0.9457 به مقدار 0.9317 کاهش می‌یابد.)
</p>

```python
rfc_100 = RandomForestClassifier(n_estimators=5, random_state=0)
rfc_100.fit(X_train, y_train)
y_pred_100 = rfc_100.predict(X_test)

print("confusion_matrix:\n", confusion_matrix(y_test, y_pred_100))
print("accuracy_score:", accuracy_score(y_test, y_pred_100))
print("recall_score:", recall_score(y_test, y_pred_100,average='micro'))
print("precision_score:", precision_score(y_test, y_pred_100,average='micro'))
print(f"\nClassification Report:\n{classification_report(y_test, y_pred_100 )}")
```

<p style="font-family:B Nazanin">
      در نهایت با استفاده از روش‌هایی مانند RandomizedSearchCV می‌توانیم بهترین مقادیر هایپرپارامترها را پیدا کنیم.
</p>

```python
rf = RandomForestClassifier()
params = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 15, 20],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2, 4]
}

random_search = RandomizedSearchCV(rf, params, n_iter=100, cv=5, verbose=1, n_jobs=-1, random_state=42)
random_search.fit(X_train, y_train)

print('Best Parameters:', random_search.best_params_)
best_rf = random_search.best_estimator_
y_pred = best_rf.predict(X_test)

print("confusion_matrix:\n", confusion_matrix(y_test, y_pred_100))
print("accuracy_score:", accuracy_score(y_test, y_pred_100))
print("recall_score:", recall_score(y_test, y_pred_100,average='micro'))
print("precision_score:", precision_score(y_test, y_pred_100,average='micro'))
print(f"\nClassification Report:\n{classification_report(y_test, y_pred_100 )}")
```

<p style="font-family:B Nazanin">
    <b>Feature Importance:</b> در بسیاری از الگوریتم‌های یادگیری ماشین مفهومی به نام Feature Importance وجود دارد که اهمیت هر کدام از ویژگی‌ها روی کلاس هدف را مشخص می‌کند. به عبارت دیگر بیان می‌کند که هر کدام از ویژگی‌ها چقدر روی متغیر هدف تاثیر دارند. این مقدار برای هر ویژگی عددی بین صفر تا یک است که هر چه این عدد بزرگ‌تر باشد، مشخص می‌کند که آن ویژگی اثر بیشتری روی متغیر هدف دارد. مجموع Feature Importance تمام ویژگی‌ها همیشه برابر با یک است. همان‌طور که در کد زیر می‌بینیم، خروجی دستور feature_importance_ یک آرایه است که مقدار feature importance را برای تمام ویژگی‌های دیتاست به ترتیب مشخص می‌کند که مجموع همه این اعداد نیز یک است. 
</p>

```python
fi=rfc.feature_importances_
print(f"feature importances : {fi}")
print(f"sum of feature importances : {fi.sum()}")
```

<p style="font-family:B Nazanin">
      اگر بخواهیم ویژگی‌ها را به ترتیب اهمیت و تاثیر روی متغیر هدف ببینیم:
</p>

```python
feature_importances = pd.Series(data=fi, index=X.columns)
feature_importances=feature_importances.sort_values(ascending=False)
print(f"\nfeature_importances :\n{feature_importances}")
```

<p style="font-family:B Nazanin">
    <b>الگوریتم K نزدیک‌ترین همسایه یا K-Nearest Neighbours یا KNN:</b><br>
    الگوریتم KNN یک الگوریتم نظارتی است که هم برای مسائل طبقه‌بندی و هم برای مسائل رگرسیون استفاده می‌شود. این الگوریتم بر اساس فاصله بین نقاط داده عمل می‌کند. یعنی برای پیش‌بینی مقدار متغیر هدف برای یک نمونه، ابتدا k همسایه از بین نزدیک‌ترین نقاط به آن نمونه را پیدا می‌کنیم. سپس اگر مسئله از نوع طبقه‌بندی بود، می‌بینیم که این همسایه‌ها بیشتر مربوط به کدام کلاس یا دسته‌بندی هستند. پس پیش‌بینی می‌کنیم که این نمونه نیز از همان کلاس باشد. اگر مسئله از نوع رگرسیون باشد، میانگین مقدار همسایه‌ها را به عنوان پیش‌بینی مقدار نمونه در نظر می‌گیریم. (پارامتر k تعداد همسایه‌های نزدیک هر نقطه است که برای پیش‌بینی از آن استفاده می‌کنیم.)
</p>

<p style="font-family:B Nazanin">
      برای پیدا کردن نزدیک‌ترین همسایه‌ها نیاز به محاسبه فاصله بین نقاط داریم. در بخش‌های قبل نحوه محاسبه فاصله بین نقاط را با روش‌های مختلف (اقلیدسی، منهتن و ...) را بررسی کردیم که در این الگوریتم بیشتر از روش اقلیدسی استفاده می‌کنیم. روش کار الگوریتم به این صورت است که برای هر نقطه جدید، ابتدا فاصله این نقطه با تمام داده‌های موجود قبلی محاسبه می‌شود. سپس k مورد از نزدیک‌ترین همسایه‌ها انتخاب می‌شوند. سپس اگر مسئله طبقه‌بندی بود، کلاس پیش‌بینی‌شده می‌شود کلاسی که در بین همسایه‌ها در اکثریت باشد و اگر رگرسیون بود، میانگین مقدار همسایه‌ها به عنوان مقدار پیش‌بینی‌شده انتخاب می‌شود. به عنوان مثال در تصویر بالا برای نقطه P، هفت همسایه انتخاب شده است که 3 مورد کلاس B، دو مورد کلاس A و دو مورد کلاس C است. پس پیش‌بینی می‌کنیم که نقطه P هم جزء کلاس B باشد.
</p>

<p style="font-family:B Nazanin">
      از مزایای این الگوریتم می‌توان گفت که <br>
    * بسیار ساده است.<br>
    * هم برای مسائل طبقه‌بندی و هم رگرسیون استفاده می‌شود.<br>
    * همچنین نیازی به فرضیات اولیه در مورد توزیع داده‌ها ندارد (یعنی لزومی ندارد که توزیع داده‌ها در آن نرمال باشد).<br>
    اما از معایب آن می‌توان گفت که<br>
    * برای داده‌های بزرگ کارایی چندانی ندارد. مثلا اگر فرض کنیم که یک میلیون داده داشته باشیم و بخواهیم فاصله یک نقطه جدید را با تمام آن داده‌های قبلی محاسبه کنیم، بسیار زمان‌گیر و پرهزینه است و بهینه نمی‌باشد. چون محاسبه فاصله هم زمان زیادی و هم هزینه زیادی می‌برد. <br>
    * الگوریتم KNN حتما نیاز به مقیاس‌بندی داده‌ها دارد.<br>
    * این الگوریتم بسیار حساس به نویز است و داده‌های نویزی می‌توانند تاثیر منفی در عملکرد KNN و تشخیص همسایه‌ها داشته باشد.<br>
    <br>
    معمولا برای بهبود عملکرد KNN، اگر تعداد ویژگی‌ها زیاد بود، سعی می‌کنیم با استفاده از روش‌های کاهش ابعاد مثلا PCA، تعداد ویژگی‌ها و ابعاد مدل را کاهش دهیم. هر چقدر کاهش ابعاد داشته باشیم و از طرف دیگر نرمال‌سازی یا مقیاس‌بندی نیز انجام دهیم، KNN بهتر می‌تواند عمل کند. 
</p>

<p style="font-family:B Nazanin">
    <b>هایپرپارامترهای KNN:</b><br>
    * n_neighbours: تعداد همسایه‌هایی که باید بررسی شود یا همان K. به صورت پیش‌فرض 5 است. <br>
    * weights: وزن همسایه‌ها را مشخص می‌کند. در حالت پیش‌فرض uniform است یعنی همه مسایه‌ها وزن یکسان و برابر دارند. اما می‌توانیم برای همسایه‌ها وزن در نظر بگیریم. یعنی بگوییم همسایه‌های نزدیک‌تر، وزن بیشتر و همسایه‌های دورتر، وزن کمتری دارند تا بین آن‌ها تعادل برقرار کنیم. در این حالت باید این پارامتر را برابر با distance قرار دهیم. <br>
    * algorithm: الگوریتمی است که برای پیدا کردن نزدیک‌ترین همسایه استفاده می‌شود. مانند auto و ball_tree و kd_tree و brute. <br>
    * leaf_size: اندازه برگ‌ها در الگوریتم‌های BallTree و KDTree. مقدار پیش‌فرض آن 30 است. <br>
    * metric: معیار اندازه‌گیری که برای محاسبه فاصله بین نقاط استفاده می‌شود. (euclidean و manhattan و minkowski)<br>
    * p: توان پارامتر برای معیار اندازه‌گیری minkowski. (مقدار 1 معادل فاصله منهتن و مقدار 2 معادل فاصله اقلیدسی می‌شود).
</p>

<p style="font-family:B Nazanin">
      کلاس MyKNN را تعریف می‌کنیم.

```python
from math import sqrt

class MyKNN():
  def __init__(self,k):
    self.k=k

  def fit(self,X_train,y_train):
    self.x_train=X_train
    self.y_train = y_train.to_numpy()

  def euclidean_distance(self,sample1,sample2):
    distance=0.0
    for i in range(len(sample1)):
      distance+=(sample1[i]-sample2[i])**2
    return sqrt(distance)


  def nearest_neighbors(self,test_sample):
    distances=[]
    for i in range(len(self.x_train)):
      d=self.euclidean_distance(self.x_train[i],test_sample)
      t=(self.y_train[i],d)
      # print(f"{i} => {t}")
      distances.append(t)

    distances.sort(key=lambda x:x[1])
    neighbors=[]
    for i in range(self.k):
      neighbors.append(distances[i][0])
    return neighbors


  def predict(self,test_set):
    predictions=[]
    i=1
    for test_sample in test_set:
      # print(f"{80*'='}\n{i} \n{80*'*'}")
      k_nearest_neighbors=self.nearest_neighbors(test_sample)
      # print(f"{self.k} nearest => {k_nearest_neighbors}")
      i+=1
      prediction=max(k_nearest_neighbors,key=k_nearest_neighbors.count)
      predictions.append(prediction)
    return predictions
```

```python
my_knn_classifier=MyKNN(5)
my_knn_classifier.fit(X_train,y_train)
y_pred=my_knn_classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,recall_score,precision_score,classification_report
print("confusion_matrix:\n", confusion_matrix(y_test, y_pred))
print("accuracy_score:", accuracy_score(y_test, y_pred))
print("recall_score:", recall_score(y_test, y_pred,average='micro'))
print("precision_score:", precision_score(y_test, y_pred,average='micro'))
print(f"\nClassification Report:\n{classification_report(y_test, y_pred )}")
```

<p style="font-family:B Nazanin">
    <b>الگوریتم ماشین بردار پشتیبانی یا Support Vector Machine یا SVM:</b><br>
    الگوریتم SVM یک الگوریتم نظارتی است که به دلیل اینکه از قدرت محاسباتی ریاضی زیادی بهره می‌برد، بسیار معروف شده است و عملکرد بهینه و کارایی بالایی دارد. این الگوریتم در مسائلی همچون طبقه‌بندی متن، طبقه‌بندی تصویر، پردازش تصویر، تشخیص الگوها، شناسایی دست‌خط بسیار کاربرد دارد. البته از این الگوریتم در مسائل رگرسیون نیز استفاده می‌شود (که در این مسائل به آن SVR گفته می‌شود). الگوریتم SVM در ابتدا برای مسائل طبقه‌بندی دو کلاسه مطرح شد ولی در گذر زمان و با پیشرفت این الگوریتم، برای مسائل چند کلاسه و حتی غیر خطی نیز استفاده شده است. 
</p>

<p style="font-family:B Nazanin">
      همان طور که در تصویر بالا مشخص است، روش کار این الگوریتم به این صورت است که سعی می‌کند یک ابرصفحه (hyperplane) بیابد که نمونه‌های دو کلاس را از یکدیگر جدا کند. در حالت کلی ممکن خطوط بسیار زیادی وجود داشته باشد که دو کلاس را از هم جدا کند (مشابه تصویر سمت چپ). اما هدف اصلی الگوریتم SVM این است خط یا صفحه‌ای را پیدا کند که از نمونه‌های هر دو کلاس بیشترین فاصله یا حاشیه را داشته باشد (تصویر سمت راست). البته تصویر بالا حالت بسیار ایده‌آلی است که یک خط یا صفحه می‌تواند تمام نمونه‌های دو کلاس را جدا کند. در عمل ممکن است تعدادی از نمونه‌های هر کلاس در محدوده کلاس دیگر قرار بگیرد. ضمن این که این الگوریتم برای مسائل چندکلاسه نیز به کار می‌رود که پیچیده‌تر از این حالت است و باید صفحه یا ابرصفحه‌ای را پیدا کنیم که کلاس‌ها را جدا کند.
</p>

<p style="font-family:B Nazanin">
    <b>انواع SVM بر اساس ماهیت مرز تصمیم‌گیری:</b><br>
    * Linear SVM: الگوریتم SVM خطی به حالتی می‌گویند که می‌توانیم با استفاده از یک خط (در حالت سه بعدی یک صفحه) کلاس‌ها را از یکدیگر جدا کنیم. <br>
    * Non-Linear SVM: الگوریتم SVM غیرخطی به حالتی می‌گویند که نتوانیم با استفاده از یک خط (یا صفحه) کلاس‌ها را از هم جدا کنیم. در این حالت توابعی به نام کرنل استفاده می‌شوند تا داده‌ها را به ویژگی‌هایی با ابعاد کمی بالاتر تبدیل کنند تا جایی که داده‌ها بتوانند به صورت خطی از هم جدا شوند.  
</p>

<p style="font-family:B Nazanin">
      به عنوان مثال مشابه تصویر زیر، توابع Kernel، داده‌های دو بعدی غیرخطی را طوری به سه بعد تبدیل کرده که بتواند با استفاده از یک صفحه آن‌ها را از هم جدا کند. 
</p>

<p style="font-family:B Nazanin">
      اصطلاحاتی که در این الگوریتم استفاده می‌شود عبارتند از:<br>
    * Hyperplane: صفحه‌ای که داده‌های کلاس‌ها را از هم جدا می‌کند، Hyperplane گفته می‌شود. در فضای دو بعدی به آن خط، در سه بعد به آن صفحه و در ابعاد بیشتر به آن ابرصفحه گفته می‌شود. <br>
    * Margin: به فاصله بین نزدیک‌ترین نقاط هر کلاس به ابرصفحه، Margin یا حاشیه گفته می‌شود.<br>
    * Support Vector: نزدیک‌ترین نقطه هر کلاس به ابرصفحه را Support Vector یا نقطه پشتیبانی می‌گویند. 
</p>

<p style="font-family:B Nazanin">
      * Hard Margin و Soft Margin: در حالت ایده‌آل می‌توانیم با یک ابرصفحه تمام نقاط و داده‌های دو کلاس را از هم جدا کنیم که به این حالت حاشیه سخت یا Hard Margin گفته می‌شود. اما در عمل ممکن است در هر کلاس، داده‌های پرتی وجود داشته باشد که این شرط را بر هم بزند. الگوریتم SVM اجازه داده است که با کمک برخی متغیرها به نام slag variableها، الزامات حاشیه سخت را کاهش دهیم و اجازه دهیم که طبقه‌بندی در برخی جاها با اشتباه همراه باشد. این حالت که به آن حاشیه نرم یا Soft Margin گفته می‌شود، کمک می‌کند که الگوریتم با داده‌های پرت و نویزی بهتر بتواند رفتار کند.
</p>

<p style="font-family:B Nazanin">
      * Kernel: همان طور که قبلا نیز توضیح داده شد، Kernelها توابع یا روش‌هایی هستند که داده‌ها را به یک فضا یا بعد بالاتر منتقل می‌کنند تا بتوان یک ابرصفحه خطی پیدا کنیم تا داده‌های کلاس‌ها را از هم جدا کند. پر استفاده‌ترین (محبوب‌ترین) توابع کرنل در SVM عبارتند از:
</p>

<table>
    <tr>
        <td>Linear</td>
        <td>$k(x_1, x_2) = x_1.x_2$</td>
    </tr>
    <tr>
        <td>Polynomial</td>
        <td>$k(x_1, x_2) = (\gamma x_1.x_2 + c)^d$</td>
    </tr>
    <tr>
        <td>Gaussian or Radial basis</td>
        <td>$k(x_1, x_2) = \exp\left(-\gamma\Vert{x_1 - x_2}\Vert^2 \right)$</td>
    </tr>
    <tr>
        <td>Sigmoid</td>
        <td>$k(x_1, x_2) = \tanh\left(\gamma x_1.x_2 + c\right)$</td>
    </tr>
</table>

<p style="font-family:B Nazanin">
      مهم‌ترین هایپرپارامترهای SVM عبارتند از:<br>
    * C: میزان جریمه‌ای را تعیین می‌کند که مدل برای خطای طبقه‌بندی متحمل می‌شود. مقدار پیش‌فرض آن برابر یک است. در واقع پارامتر c پارامتری است که در حالت حاشیه نرم، تعادل بین حداکثر کردن حاشیه و حداقل کردن خطا را کنترل می‌کند. هر چقدر c را بیشتر کنیم، یعنی می‌خواهیم خطاهای طبقه‌بندی را کمتر کنیم. یعنی اجازه ندهیم که نقاط داده‌ای زیادی در سمت اشتباه باشند. هر چقدر c کمتر باشد، یعنی اجازه اشتباه بیشتری می‌دهیم ولی حاشیه بزرگ‌تر می‌شود. از طرفی اگر مقدار c بیش از حد کوچک انتخاب شود، ممکن است مقدار خطا خیلی زیاد شود و underfitting به وجود آید و اگر مقدار c بیش از حد زیاد شود، حاشیه بسیار کوچک می‌شود و مدل بیش از حد به داده‌های آموزشی وابسته می‌شود و overfitting رخ می‌دهد. <br>
    * پارامتر kernel type: نوع کرنل را تعیین می‌کند. مقدار Linear برای مسائل SVM خطی استفاده می‌شود. Polynomial برای مسائل پیچیده‌تر استفاده می‌شود. RBF بین این دو حالت است که هم برای مسائل خطی و هم مسائل غیرخطی کاربرد دارد که مقدار پیش‌فرض نیز همین RBF است. Sigmoid هم برای مسائل غیرخطی کاربرد دارد ولی کمتر استفاده می‌شود.<br>
    * پارامتر گاما ($\gamma$) در RBF Kernel: میزان گستردگی تابع کرنل را در کرنل RBF تعیین می‌کند. اگر مقدار گاما را زیاد کنیم، حساسیت تابع کرنل به نقاط نزدیک به حاشیه زیادتر می‌شود. اگر بیش از حد زیاد شود، overfitting اتفاق می‌افتد. اگر مقدار گاما را کم کنیم، حساسیت تابع کرنل به نقاط دورتر از حاشیه زیادتر می‌شود ولی در این صورت ممکن است خطا بیشتر شود. <br>
    * پارامتر degree در Polynomial kernel: برای کرنل چندجمله‌ای کاربرد دارد و درجه چند جمله‌ای را تعیین می‌کند. هر چه بالاتر رود، درجه چندجمله‌ای بیشار می‌شود و دقت آن بالاتر می‌رود ولی باید مراقب overfitting باشیم. هر چه درجه چندجمله‌ای کمتر باشد، پیچیدگی مدل و حساسیت آن کمتر می‌شود و خطا افزایش می‌یابد و باید مراقب underfitting باشیم.
</p>

<p style="font-family:B Nazanin">
      از مهم‌ترین مزایای SVM می‌توان به این موارد اشاره کرد: <br>
    * تعمیم‌پذیری بالایی دارد. با استفاده از پارامتر c می‌توان تعادلی بین حاشیه و خطا ایجاد کرد و برای مسائل زیادی کاربرد دارد. <br>
    * برای مسائل غیرخطی کاربرد دارد. یعنی با استفاده از توابع کرنل امکان تغییر فضا داریم و می‌توانیم داده‌های غیرخطی را طوری تغییر دهیم که بتوانیم به صورت خطی با آن‌ها رفتار کنیم. <br>
    * می‌تواند با داده‌های با ابعاد بالا کار کند (در حالی که مثلا در الگوریتم KNN بزرگ‌ترین معضل ابعاد بالای داده‌ها بود). اما در این الگوریتم بزرگ‌شدن ابعاد داده‌ها اهمیتی ندارد چون عملا فقط با support vectorها کار داریم و حجم بالای داده و ویژگی‌ها تاثیری در مسئله ندارد.

<p style="font-family:B Nazanin">
      مثال. در این مثال از دیتاست Pulsar_stars استفاده می‌کنیم که اطلاعات نجومی در مورد تعدادی از ستاره‌ها را مشخص کرده است و در نهایت متغیر هدف این است که تعیین کنیم این ستاره از نوع Pulsar هست یا خیر (ستاره Pulsar ستاره‌ای است که امواج رادیویی قدرتمندی را در بازه‌های زمانی مشخصی ارسال می‌کنند). دلیل استفاده از این دیتاست این است که داده‌های آن اصلا بالانس نیست و تعادل بین کلاس‌ها وجود ندارد. همچنین داده‌های پرت زیادی در این دیتاست وجود دارد که این چالش‌ها باید در الگوریتم SVM مورد توجه قرار گیرد. ویژگی‌های مشخص‌شده در این دیتاست عبارتند از: 
</p>

<p style="font-family:B Nazanin">
      Mean of the integrated profile: میانگین پروفایل یکپارچه (شدت پالس‌ها در یک دوره زمانی) <br>
Standard deviation of the integrated profile: انحراف معیار پروفایل یکپارچه (نوسانات شدت پالس‌ها)<br>
Excess kurtosis of the integrated profile: کشیدگی اضافی پروفایل یکپارچه (میزان تیز بودن توزیع شدت پالس‌ها)<br>
Skewness of the integrated profile: چولگی پروفایل یکپارچه (تقارن یا عدم تقارن توزیع شدت پالس‌ها)<br>
Mean of the DM-SNR curve: میانگین منحنی DM-SNR (تاخیر پراکندگی)<br>
Standard deviation of the DM-SNR curve: انحراف معیار منحنی DM-SNR (نوسانات شدت سیگنال به نویز)<br>
Excess kurtosis of the DM-SNR curve: کشیدگی اضافی منحنی DM-SNR (میزان تیز بودن توزیع شدت سیگنال به نویز)<br>
Skewness of the DM-SNR curve: چولگی منحنی DM-SNR (تقارن یا عدم تقارن توزیع شدت سیگنال به نویز)<br>
target_class: اینکه ستاره از نوع pulsar هست یا خیر.
</p>

      مثال. در این مثال می‌خواهیم با استفاده از دیتاست Position Salaries، از الگوریتم SVM برای مسائل رگرسیون استفاده کنیم.

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.svm import SVR
```

```python
df = pd.read_csv('/content/drive/MyDrive/datasets/Position_Salaries.csv')
df
```

```python
df.shape
```

```python
X = df.iloc[:,1:2].values
y = df.iloc[:,2:3].values
```

```python
sc_X = StandardScaler()
sc_y = StandardScaler()
X = sc_X.fit_transform(X)
y = sc_y.fit_transform(y)
```

```python
svr = SVR(kernel='rbf')
svr.fit(X,y)
```

```python
plt.scatter(X, y, color = 'magenta')
plt.plot(X, svr.predict(X), color = 'green')
plt.title('Support Vector Regression')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV, train_test_split
param_grid = {
    'C': [0.1, 1, 10, 100],
    'gamma': [1, 0.1, 0.01, 0.001],
    'kernel': ['linear', 'rbf', 'poly']
}

grid_search = GridSearchCV(estimator=svr, param_grid =param_grid, cv=5, scoring='neg_mean_squared_error', verbose=2, n_jobs=-1)
grid_search.fit(X, y)
print(f"Best Parameters :\n{grid_search.best_params_}")
```

```python
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X)
mse = mean_squared_error(y, y_pred)
print(f"Mean Squared Error : {mse:.3f}")
r2_score = r2_score(y, y_pred)
print(f"R2 Score : {r2_score:.3f}")
```
