<span dir="rtl" style="font-family:B Nazanin" align="right">
    <h2>بخش چهارم: آشنایی با انواع داده‌ها و ساختار داده‌ها در علم داده</h2> 
</span>

<p style="font-family:B Nazanin">
   انواع داده‌ها از نظر ساختار: <br>
    * داده‌های ساختاریافته (Structured Data) <br>
    * داده‌های غیر ساختاریافته (Unstructured Data) <br>
    * داده‌های نیمه‌ساختاریافته (Semi Structured Data) <br>
    <hr>
    <b>داده‌های ساختاریافته:</b> داده‌هایی هستند که دارای قالب از پیش‌تعریف‌شده هستند. دارای فیلدهای مشخص و معینی هستند و هر رکورد از این داده‌ها دارای تمام آن فیلدها است. مثلا برای مشخصات افراد، نام،‌نام خانوادگی، کد ملی، شماره موبایل و ... ذخیره می‌شود و اطلاعات هر فرد دیگری بخواهد ذخیره شود باید تمام این فیلدها را شامل شود. معمولا این نوع داده‌ها برای انسان‌ها دوست‌داشتنی‌تر هستند. معمولا به صورت جدولی هستند که هر سطر یک رکورد و هر ستون یک فیلد از اطلاعات است. ساختاری بسیار منطم و قابل مقایسه است. ویژگی‌ها: <br>
    - در انجام عملیات درج و حذف و ویرایش بسیار راحت هستند. <br>
    - سرعت جست‌وجو و بازیابی اطلاعات بالا است. <br>
    - امنیت بالایی دارند. <br>
    - قابلیت مقیاس‌پذیری نسبتا بالایی دارند. خیلی راحت می‌توان رکورد جدید ایجاد کرد و افزودن ستون یا فیلد به این اطلاعات تقریبا راحت است. <br>
    - حدود 20٪ از داده‌های حوزه بیگ دیتا از این نوع است. <br>
    پایگاه داده‌های SQL و فایل‌های Excel از این نوع داده‌ها هستند. 
</p>

<p style="font-family:B Nazanin">
    <b>داده‌های غیر ساختاریافته:</b> داده‌هایی که برای ذخیره آن‌ها از هیچ قالب از پیش‌تعریف‌شده‌ای استفاده نمی‌شود. مثل ویدئوها، تصویرها، فایل‌های صوتی. هیچ کدام از این مثال‌ها ساختار منظمی برای ذخیره داده‌ها ندارند. ویژگی‌ها: <br>
    - پردازش نسبتا پیچیده‌تری نسبت به داده‌های ساختاریافته دارند. مثلا در یک فایل word بخواهیم دنبال یک کلمه خاص بگردیم، هیچ قالب مشخص یا مکان مشخصی وجود ندارد که در جست‌وجو به ما کمک کند. <br>
    - انعطاف‌پذیری بالایی دارند. معمولا انعطاف‌پذیری در تضاد با نظم است. هر چه داده‌ها نظم بیشتری داشته‌باشند، کمتر می‌توان آن‌ها را تغییر داد. اما هر چه نظم کمتری داشته‌باشند، انعطاف بیشتری دارند. <br>
    - مقیاس‌پذیری نسبتا زیادی دارند. یعنی محدودیتی در ذخیره‌سازی داده‌ها وجود ندارد. <br> 
    نمونه‌هایی از این نوع داده‌ها مانند فایل‌های word، فایل‌های pdf و فایل‌های صوتی،‌تصویری و ویدئویی هستند. 
</p>

<p style="font-family:B Nazanin">
    <b>داده‌های نیمه‌ساختاریافته:</b> داده‌هایی هستند که از یک طرف می‌توان نظم و قاعده مشخصی در آن‌ها پیدا کرد ولی از جهات دیگر می‌توانند این نظم و قاعده و قالب را از بین ببرند و از آن تبعیت نکنند. مانند فایل‌های Json و XML. مثلا در فایل‌های Json می‌توان گفت از این جهت که آرایه‌ای از آبجکت‌ها هستند، یک نوع داده منظم و دارای قالب است ولی می‌دانیم که هر کدام از این آبجکت‌ها می‌توانند با دیگری متفاوت باشند و فیلدهای کاملا متفاوتی داشته‌باشند. پس از این لحاظ یک نوع داده بدون نظم و ساختار هستند. ویژگی‌های داده‌های نیمه‌ساختاریافته: <br>
    - قابل تبدیل به داده‌های ساختاریافته هستند. یعنی تا حدی می‌توان برای آن‌ها جداولی مانند داده‌های ساختاریافته درست کرد و این داده‌ها را درون آن‌ها ذخیره کرد. <br>
    - انعظاف‌پذیری این نوع داده‌ها بسیار بالا است و به راحتی می‌توان آن‌ها را تغییر داد. <br> 
    - مقیاس‌پذیری این داده‌ها بسیار بالا است و به خاطر این که نظم مشخصی ندارند، به راحتی می‌توان آن‌ها را گسترش و توسعه داد و داده‌ها و فیلدهای جدید به آن‌ها اضافه کرد. <hr> 
</p>

<p style="font-family:B Nazanin">
   انواع مقادیر داده‌ای در داده‌های ساختار یافته:<br>
    * داده‌های کیفی (Qualitative or Categorical Data): <br>
      &nbsp  &nbsp - داده‌های اسمی (Nominal Data)<br>
       &nbsp &nbsp - داده‌های ترتیبی (Ordinal Data) <br>
    * داده‌های کمی (Quantitative Data): <br>
      &nbsp &nbsp  - داده‌های گسسته (Descrete Data) <br>
      &nbsp &nbsp  - داده‌های پیوسته (Continous Data)
</p>

<p style="font-family:B Nazanin">
   داده‌های کیفی داده‌هایی هستند که کیفیت یا نامشان مهم است. مثل جنسیت، زبان، رنگ، گروه خونی، متن و ... . در واقع در این نوع داده‌های مقدار عددی وجود ندارد و کیفیت مهم است. به عبارت دیگر قابل شمارش نیستند. اما داده‌های کمی داده‌هایی هستند که در آن‌ها یک عدد هستند و در آن‌ها بحث مقدار عددی و شمارش مطرح می‌شود. <br>
    داده‌های کیفی خود نیز به دو دسته تقسیم‌بندی می‌شوند. داده‌های کیفی اسمی داده‌هایی هستند که نه قابل شمارش‌اند و نه ترتیب در آن‌ها مهم است. مثل جنسیت، زبان، رنگ. اما داده‌های کیفی ترتیبی داده‌هایی هستند که در آن‌ها ترتیب مهم است. مثلا مدرک تحصیلی یک فرد یک نوع داده کیفی ترتیبی است. یعنی مقدار عددی ندارد ولی مقادیر مختلف آن به ترتیب دارای ارزش‌های متفاوتی هستند. مثلا دیپلم، لیسانس، فوق‌لیسانس و ... . <br>
    داده‌های کمی نیز به دو دسته تقسیم‌بندی می‌شوند. داده‌های کمی گسسته داده‌هایی هستند که مقادیرشان اعداد صحیح هستند. مثل تعداد کارمندان یک اداره، تعداد اتاق‌های یک خانه، تعداد پروژه‌های یک فرد. داده‌های کمی گسسته داده‌هایی هستند که با اعداد اعشاری و غیرصحیح مشخص می‌شوند. مثل وزن یک انسان، دما،‌سرعت، فرکانس.
    <hr>
</p>

<p style="font-family:B Nazanin">
    <b>مجموعه‌های داده‌ای:</b> در علم داده مجموعه داده‌ها به چند صورت ذخیره و استفاده می‌شوند:<br>
    * Tabular Data: داده‌هایی هستند که مبتنی بر جدول هستند. این داده‌ها جزء داده‌های ساختاریافته هستند. دارای سطر یا رکورد و همچنین ستون یا فیلد هستند. در این نوع مجموعه داده، امکان دارد بعضی از فیلدها خالی باشند و داده‌ای در آن ذخیره نشده باشد. <br>
    * Data Frames: داده‌هایی هستند که در جدول‌های دوبعدی شامل سطرها و ستون‌ها ذخیره می‌شوند. بسیار شبیه به یک صفحه اکسل هستند. بسیار کاربردی و قابل استفاده هستند و انعطاف‌پذیری بالایی دارند. بسیار شبیه به داده‌های جدولی هستند اما تفاوت این جاست که در داده‌های جدولی تعداد ستون‌ها مهم است و برای همه سطرها، تعداد ستون یکسانی داده وارد می‌شود. اما در دیتافریم‌ها ممکن است تعداد ستون داده‌ها در سطرهای مختلف، متفاوت باشد. بنابراین ممکن است کار با دیتافریم‌ها نیاز به دقت بیشتری داشته باشد. <br>
    * Data Series: سری‌های داده‌ای مجموعه‌ای از داده‌ها هستند که بر روی نمودار آن‌ها را نشان می‌دهیم. معمولا گذشته، حال و آینده یک داده یا پدیده را نشان می‌دهند و وضعیت آن را در طول زمان مشخص می‌کنند. مثلا میزان سود روزانه در طول یک هفته را در قالب نمودار نشان می‌دهیم. این داده‌ها نیز به صورت دوبعدی هستند که هر کدام به صورت یک کلید-مقدار (key-value) ذخیره می‌شوند. <br>
    * Time Series Data: داده‌های سری زمانی، داده‌هایی هستند که در اثر مشاهده یک پدیده در طول زمان به‌دست می‌آیند. یعنی یک سری داده هستند که همیشه یکی از ابعاد آن زمان است. مثلا میزان سود یا ضرر در طول یک روز در بورس، مانیتورینگ بهره‌وری یک مجموعه در طول زمان یک روز یا یک هفته یا یک ماه.گ
    <hr>
</p>

<p style="font-family:B Nazanin">
   وقتی در مورد DataSet یا مجموعه داده صحبت می‌کنیم، منظور همان داده‌های ساختاریافته یا نهایتا نیمه‌ساختاریافته است. دیتاست‌ها یا به صورت آماده به دست ما می‌رسد یا توسط پژوهشگرها تولید می‌شود. معمولا مهم‌ترین قسمت پروژه‌های علم داده دسترسی به یک دیتاست یا مجموعه داده است. این داده‌ها یا داده‌های واقعی هستند که در پژوهش‌های واقعی به‌دست آمده‌اند یا به صورت تصادفی تولید شده‌اند. ولی در پروژه‌های واقعی و بزرگ باید به داده‌های واقعی دسترسی داشته‌باشیم. <br>
    در اینجا با دو مفهوم روبرو هستیم. DataSet و DataTable. <br>
    DataSet به طور جمعی از تعدادی جدول تشکیل شده‌است اما DataTable یک جدول از داده‌هاست. پس DataSet در واقع مجموعه‌ای از DataTableهاست. <br>
    Data Source یا منابع داده‌ای به هر جایی گفته می‌شود که می‌توان از آن داده استخراج کرد و به‌دست آورد. مثلا سنسورها یا حسگرها که تولید داده می‌کنند، می‌توانند یک منبع داده‌ای باشند. <br>
    یک سری شرکت‌ها و وبسایت‌ها وجود دارند که در زمینه‌های مختلف فعالیت کرده و یک سری دیتاست‌های آماده استاندارد را تهیه کرده‌اند تا سایر افراد در پروژه‌های خود از این دیتاست‌ها بتوانند استفاده کنند. مانند:<br>
    - وبسایت Kaggle به آدرس <a href="https://www.kaggle.com/">https://www.kaggle.com/</a> <br>
    - وبسایت UCI به آدرس <a href="https://archive.ics.uci.edu/">https://archive.ics.uci.edu/</a> <br>
    - موتور جست‌وجوگر google datasets به آدرس <a href="https://datasetsearch.research.google.com/">https://datasetsearch.research.google.com/</a><br>
    - کتابخانه scikit learn در پایتون<br>
    <br>
    سایر مجموعه‌ها عبارتند از :<br>
    - Amazon Datasets <br>
    - Microsoft Datasets <br>
    - Government Datasets <br>
    - Awesome public dataset collections <br>
    - Papers with code <br>
    - Computer Vision Datasets <br>
    - Lionbridge AI <br> 
    <hr> 
</p>

<p style="font-family:B Nazanin">
    <b>مخازن داده:</b> داده‌هایی که از طریق Data Sourceها ایجاد می‌شوند و به دست ما می‌رسند، درون مجموعه‌هایی به نام مخازن داده نگهداری می‌شوند. در اینجا انواع این مخازن داده را بررسی می‌کنیم: <br>
    * <b>دریاچه داده (Data Lake):</b> یک مخزن بسیار بزرگ از داده‌های مختلف که می‌تواند ساختاریافته، نیمه‌ساختاریافته و غیرساختاریافته باشد و از روش‌های مختلف وارد این دریاچه شده‌اند. مثلا از طریق یک API خوانده‌شده، یا از طریق سنسورها خوانده‌شده یا از یک سایت آن‌ها را دریافت کرده‌ایم. داده‌هایی که درون دریاچه ذخیره می‌شوند، معمولا طبیعی هستند یعنی هیچگونه دستکاری روی آن‌ها انجام نشده‌است. <br>
    * <b>انبار داده (Data Warehouse):</b> مجموعه داده‌های ساختاریافته است که از منابع مختلف جمع شده‌است و آماده پاسخگویی به یک موضوع مشخص است. انبار داده منظم است و می‌تواند پاسخگوی نیاز ما باشد. <br>
    * <b>بازار داده (Data Mart):</b> بازار داده یک نمونه کوچک یا زیرمجموعه‌ای از انبار داده است. بنابراین داده‌های آن نیز ساختاریافته است که می‌توانیم برای هر واحد سازمانی از آن استفاده کنیم. در واقع بازار داده، بخشی از یک انبار داده است که یک واحد سازمانی مشخص با آن کار می‌کند. بازارهای داده سه دسته دارند: <br>
    &nbsp &nbsp &nbsp - <b>بازار داده وابسته (Dependent Data Mart):</b> اگر ابتدا یک انبار داده بسازیم و سپس بخشی از آن را به عنوان بازار داده جدا کنیم، به آن بازار داده وابسته می‌گویند. یعنی هر بار که داده‌ای را بخواهیم باید آن را از انبار داده استخراج کنیم. <br>
    &nbsp &nbsp &nbsp - <b>بازار داده مستقل (Independent Data Mart):</b> بازارهای داده مستقل یعنی بدون دخالت انبار داده، داده‌ها را از منابع داده‌ای استخراج کرده و ذخیره می‌کنیم. <br>
    &nbsp &nbsp &nbsp - <b>بازار داده ترکیبی (Hybrid Data Mart):</b> بازار داده‌ای است که هم به صورت مستقیم و هم به صورت غیرمستقیم با انبار داده کار می‌کند. یعنی بخشی از آن وابسته به انبار داده است و بخشی از آن مستقل. 
    <hr>
</p>

<p style="font-family:B Nazanin">
    <b>استخراج داده‌ها، تبدیل داده‌ها و بارگذاری داده‌ها (Extract, Transform, Load) یا ETL:</b>
    ETL یک مبحث بسیار مهم در علم داده است که در واقع یک فرایند یکپارچه‌سازی داده‌هاست. برای انجام هر پروژه یا فرایندی در علم داده، باید ابتدا داده‌ها را استخراج کنیم. سپس باید آن‌ها را Transform کنیم. یعنی آن‌ها را به داده‌های ساده و واحد تبدیل کنیم. به عبارت دیگر داده‌ها را به داده‌هایی که برای ما کاربردی است،‌تبدیل کنیم و در نهایت داده‌ها را در یک سیستم داده یا انبار داده ذخیره کنیم. به این فرایند ETL می‌گویند. <br>
    در مرحله استخراج یا Extract، داده‌ها از منابع مختلف جمع‌آوری شده و در Staging Area که یک محل ذخیره موقت است، ذخیره می‌شوند. در این مرحله هم دریافت داده‌ها و هم اعتبارسنجی انجام می‌شود. سپس وقت این می‌رسد که تبدیل داده‌ها به داده‌هایی که ارزشمند است،‌انجام شود. تبدیل داده نیز در همان فضای حافظه Staging Area اتفاق می‌افتد. در نهایت داده‌های تبدیل‌شده در یک فضای حافظه ماندگار (حافظه هدف یا Target) ذخیره می‌شود. <hr> 
</p>

<p style="font-family:B Nazanin">
    <b>هوش تجاری (Business Intelligence):</b> علمی است شامل یک سری نرم‌افزار است که باعث دسترسی و تحلیل داده‌ها می‌شود. با این ابزارهایی که در هوش تجاری وجود دارد، می‌توان به یک شرکت کمک کرد تا تصمیماتش بهینه شود و کارایی‌اش بالا رود. در واقع BI یا هوش تجاری، از داده‌های گذشته برای تصمیمات امروز یک سازمان استفاده می‌کند. <br>
    <b>تجزیه و تحلیل داده‌ها (Data Analytics):</b> در این علم یک سری الگوی پنهان در داده‌ها، بر اساس مطالعه رفتارها پیدا می‌شوند. بنابراین متخصص تجزیه و تحلیل داده‌ها تاثیرات یک متغیر بر سایر متغیرها را بررسی می‌کند. مثلا یک فروشگاه با تجزیه و تحلیل داده‌ها متوجه می‌شود که با دادن یک اشانتیون روی یک محصول، فروش بیشتری کسب می‌کند تا اینکه بخواهد روی آن محصول تخفیف بدهد. بنابراین با تجزیه و تحلیل داده‌ها، الگوهای پنهان داده‌ها را کشف می‌کنیم. <br>
    <b>گزارش‌دهی (Reporting):</b> علمی است که با استفاده از آن، داده‌های جمع‌آوری‌شده را قالب‌بندی می‌کنیم، به آن الگو می‌دهیم، نمودارهای مرتبط را رسم می‌کنیم و نهایتا گزارش آن را به یک سازمان ارائه می‌دهیم. داده‌های خامی که قالبی نداشتند، نظم می‌گیرند، الگو می‌گیرند و در اختیار یک سازمان قرار می‌گیرند.
