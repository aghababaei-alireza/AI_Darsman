    <h1>فصل ششم: یادگیری عمیق</h1>
    <h2>بخش اول: مفاهیم بنیادین یادگیری عمیق  شبکه‌های عصبی مصنوعی</h2>
</span>

<p style="font-family:B Nazanin">
    در این بخش با مفاهیم و اصطلاحات پرکاربرد در مبحث یادگیری عمیق که در ادامه به آن‌ها نیاز خواهیم داشت، آشنا می‌شویم. <br>
    <b>یادگیری عمیق یا Deep Learning:</b> <br>
    یادگیری عمیق یکی از زیرمجموعه‌های یادگیری ماشین است که بر اساس شبکه‌های عصبی مصنوعی ساخته شده است و این شبکه‌ها شامل چندین لایه از نورون‌ها هستند (در مورد مفاهیمی چون نورون و شبکه عصبی در ادامه مفصل توضیح داده خواهند شد). با توجه به اینکه این مدل‌ها از چندین لایه تضکیل شده‌اند، به آن‌ها مدل‌های یادگیری عمیق گفته می‌شود. یعنی هرچقدر تعداد لایه‌های یک مدل بیشتر باشد، یعنی آن مدل، عمیق‌تر است و پیچیدگی بیشتری نیز دارد. هرجقدر پیچیدگی یک مدل بالاتر می‌رود، امکان تشخیص آن بیشتر می‌شود. کار هر لایه از شبکه عصبی، تشخیص و استخراج برخی از ویژگی‌هاست و هرچقدر که اطلاعات و ویژگی‌های بیشتری از داده‌ها استخراج شود، مدل ما سواد بیشتری به دست می‌آورد، پس می‌تواند پیش‌بینی‌های دقیق‌تری را به ما ارائه دهد. یادگیری عمیق دقت بیشتری نسبت به مدل‌های یادگیری ماشین دارد، اما فقط در مسائلی از آن استفاده می‌کنیم که پیچیدگی نسبتا بالایی داشته باشد. چرا که در مسائل عادی، استفاده از یادگیری عمیق صرفه ندارد. 
</p>

<p style="font-family:B Nazanin">
    <b>تفاوت یادگیری ماشین و یادگیری عمیق:</b> همان طور که قبلا توضیح داده شد، بر خلاف برنامه‌نویسی سنتی که باید در آن تمام قواعد و قوانین را برای ماشین مشخص می‌کردیم تا یک تصمیم‌گیری مشخص انجام شود، در الگوریتم‌های یادگیری ماشین، خود مدل می‌تواند از داده‌ها یاد بگیرد و نتایج را پیش‌بینی کند. اما نکته مهمی که در یادگیری ماشین وجود دارد این است که در ابتدا باید ویژگی‌های مهم را برای الگوریتم مشخص کنیم تا بر اساس آن‌ها فرایند یاگیری و درنهایت پیش‌بینی نتایج را انجام دهد. در حالی که الگوریتم‌های یادگیری عمیق این توانایی را دارند که خودشان فرایند استخراج ویژگی‌ها را از داده‌ها انجام دهد و از روی آن‌ها عملیات یادگیری را کامل کنند. بنابراین زمانی که تعداد داده‌ها و ویژگی‌ها بسیار زیاد است (مانند پردازش تصویر، پردازش متن و ...) اجازه می‌دهیم تا خود مدل ویژگی‌ها را پیدا کند و از روی آن‌ها آموزش ببیند. در کل یادگیری ماشین برای مسائل کوچک‌تر و ساده‌تر و یادگیری عمیق برای مسائل بزرگ‌تر و پیچیده‌تر کاربرد دارد.  
</p>

<p style="font-family:B Nazanin">
    معمولا می‌گویند یادگیری عمیق تشنه داده و سخت‌افزار است. یعنی مدل‌های یادگیری عمیق برای آموزش نیاز به داده‌های زیاد و سخت‌افزارهای قدرتمند هستند. چون وجود داده‌های بیشتر به مدل اجازه می‌دهد تا الگوهای پیچیده‌تری را از روی داده‌ها شناسایی کنند. اما همین حجم زیاد اطلاعات برای پردازش نیاز به انجام محاسبات بیشتر دارد. بنابراین سخت‌افزار قدرتمندتری نیز نیاز دارد. <br>
    بنابراین به طور خلاصه زمانی که حجم داده‌ها کم یا متوسط باشد، زمانی که ویژگی‌های داده‌ها مشخص و قابل تفسیر باشند و استخراج آن‌ها سخت نباشد و زمانی که سخت‌افزار خیلی قوی نداریم، نیاز داریم که به سراغ یادگیری ماشین برویم. اما زمانی که حجم داده‌ها زیاد است، پیچیدگی داده‌ها زیاد است و منابع سخت‌افزاری قدرتمندتری داریم و نیاز داریم که استخراج ویژگی‌ها به صورت خودکار اتفاق بیفتد، باید از یادگیری عمیق استفاده کنیم. 
</p>

<p style="font-family:B Nazanin">
     همان طور که در ابتدا توضیح داده شد، یادگیری عمیق بر اساس شبکه‌های عصبی مصنوعی انجام می‌شود که خود این شبکه‌های عصبی مصنوعی بر پایه شبکه‌های عصبی طبیعی ساخته شده‌اند. بنابراین در ابتدا ساختار این شبکه‌های عصبی را بررسی می‌کنیم. <br>
    <b>سلول عصبی طبیعی یا نورون یا Neuron:</b><br>
    نورون واحد اصلی سیستم عصبی انسان و حیوان است. یعنی مغز ما پر از این سلول‌های عصبی است. کار این نورون‌ها انتقال اطلاعات است. یعنی اطلاعات را دریافت کنند، روی آن‌ها پردازش انجام داده و به جای دیگری منتقل کنند. این اطلاعات ممکن است از یک سری نورون دیگر گرفته شود و به یک سری نورون دیگر یا بافت‌های بدن تحویل داده شود. تصویر زیر ساختار کلی یک نورون را نشان می‌دهد که در ادامه اجزای آن را شرح می‌دهیم.  
</p>

<p style="font-family:B Nazanin">
    <b>* دندریت:</b> یک سری شاخه‌های بزرگ هستند که به سلول عصبی وارد می‌شوند. وظیفه دندریت دریافت پیام‌ها از سلول‌ها یا نورون‌های دیگر است. دندریت‌ها ضخامت‌های مختلفی دارند. برخی ضخیم‌تر و برخی نازک‌تر هستند. علت این امر این است که می‌توانند با سلول‌ها یا بافت‌ها ارتباط قوی‌تر یا ضعیف‌تری داشته باشند. هر چه دندریت ضخیم‌تر باشد به معنای این است که با نورون قبلی خود ارتباط قوی‌تری دارد و سیگنال‌های قوی‌تری از آن دریافت کرده است. <br>
    <b>* هسته:</b> بخش مرکزی سلول است که اطلاعات ژنتیکی سلول در آن نگهداری می‌شود. کنترل تمامی فعالیت‌های یک نورون بر عهده هسته آن نورون است. وقتی سیگنالی وارد یک نورون می‌شود، هسته وظیفه پردازش آن سیگنال را دارد. <br>
    <b>* آکسون:</b> یک رشته بلند و نازک است که سیگنال‌های الکتریکی یا الکتروشیمیایی را از جسم سلولی به سمت نورون‌های دیگر می‌فرستد. <br>
    <b>* مایلین:</b> یک لایه چربی روی آکسون است که هم عایقی برای آکسون عمل می‌کند تا ضربه و آسیب نبیند و هم به سرعت انتقال سیگنال‌های الکتریکی کمک می‌دهد. یعنی کمک می‌دهد تا سیگنال‌ها سریع‌تر و با کارایی بیشتری منتقل شوند. <br>
    <b>* سیناپس:</b> در انتهای هر نورون یک سری اتصالات وجود دارد که به آن سیناپس می‌گویند. نورون از طریق این سیناپس‌ها به نورون‌های دیگر متصل می‌شود و از  طریق همین سیناپس‌ها پیام‌ها و سیگنال‌ها به نورون‌های دیگر منتقل می‌شود. 
</p>

<p style="font-family:B Nazanin">
    <b>پتانسیل عمل نورون یا Neuron Action Potential:</b><br>
    نکته مهمی که در مورد نورون‌ها وجود دارد این است که نورون‌ها همیشه خروجی ندارند. گفتیم که سیگنال‌های عصبی وارد نورون می‌شوند، پردازش شده و از طریق مسیرهای موجود در نورون، به نورون‌های دیگر می‌رسد. اما ممکن است همیشه این اتفاق نیفتد و در طول آکسون سیگنال قطع شود. یعنی آنقدر قدرت سیگنال ضعیف باشد که به سیناپس‌ها نرسد که در این شرایط می‌گوییم نورون خروجی ندارد. به نمودار زیر که مشخص‌کننده تغییرات ولتاژ برحسب زمان است، نودار پتامسیل عمل نورون می‌گویند. پتانسیل عنل، یک تغییر ناگهانی و گذرا در ولتاژ سلولی است که به نورون‌ها اجازه می‌دهد سیگنال‌ها را در طول آکسون خود منتقل کنند. وقتی نورون در حال استراحت است، این ولتاژ برابر با $-70 mV$ است. اگر نورونی بخواهد سیگنالی را انتقال بدهد، باید ولتاژش به یک حد آستانه (Threshold) برسد که طبق شکل زیر $-55 mV$ است. همان طور که در شکل مشاهده می‌شود، سیگنال‌هایی که کمتر از این حد آستانه هستند، منتقل نمی‌شوند و سلول مجددا به حالت استراحت می‌رود. 
</p>

<p style="font-family:B Nazanin">
    <b>شبکه‌های عصبی بیولوژیکی یا Biological Neural Networks:</b><br>
    شبکه‌های عصبی بیولوژیکی همان سیستم عصبی است که در مغز انسان‌ها و حیوانات وجود دارد. یک ساختار پیچیده دارند که از میلیاردها نورون تشکیل شده است که خروجی هر نورون از طریق سیناپس‌ها به نورون بعدی وارد می‌شود. این شبکه با استفاده از تعداد زیاد نورون‌های درونش با صرف انرژی کمی می‌تواند در تشخیص الگوها به خوبی عمل کند. شبکه‌های عصبی مصنوعی نیز از روی همین شبکه‌ها ساخته شده‌اند و در الگوریتم‌های یادگیری عمیق استفاده می‌شوند. 
</p>

<p style="font-family:B Nazanin">
    <b>نورون‌های مصنوعی (Artificial Neurons) یا پرسپترون (Perceptron):</b><br>
    نورون‌هایی هستند که از روی نورون‌های طبیعی الهام گرفته شده و ساخته شده‌اند. این نورون‌ها به صورت شکل زیر شبیه‌سازی می‌شوند. هر نورون یک سری ورودی دارد که در شکل زیر با $x$ نشان داده شده است (در شکل زیر 3 ورودی $x_1$ و $x_2$ و $x_3$ داریم ولی در کل می‌توانیم هر تعدادی ورودی داشته باشیم، مشابه با یادگیری ماشین). این ورودی‌ها یا داده‌های خام اولیه هستند یا خروجی نورون‌های دیگر. به ازای هر ورودی، یک ضریب یا یک وزن داریم. این وزن‌ها هستند که تعیین می‌کنند که هر ورودی چقدر در تصمیم‌گیری نهایی تاثیر داشته باشد. در اینجا این وزن‌ها با $w$ نشان داده شده است ($w_1$ و $w_2$ و $w_3$). در مقایسه با نورون طبیعی می‌توان گفت که وزن این ورودی‌ها الهام‌گرفته از ضخامت دندریت‌هاست. دندریت‌هایی که ضخامت بیشتری داشتند، می‌توانستند سیگنال‌های قوی‌تری را منتقل کنند. در اینجا نیز همین طور است. اگر وزن یک ورودی بیشتر باشد، اثر بیشتری در خروجی دارد. در فرایند یادگیری سعی می‌کنیم تا بهترین وزن‌ها را پیدا کنیم تا بتوانیم بهترین خروجی را پیش‌بینی کنیم. کار هسته نورون پردازش داده‌های ورودی است که در نورون مصنوعی مجموع ورودی‌های وزن‌دار را محاسبه می‌کنیم و یک مقدار بایاس (در اینجا با $b$ نشان داده شده است) را به آن اضافه می‌کنیم. وقتی بایاس به این مجموع اضافه می‌شود، قدرت تصمیم‌گیری بالاتر می‌رود. پس از محاسبه مجموع وزن‌دار ورودی‌ها و بایاس، یک تابع خطی داریم که مطمئنا برای مدل‌سازی کافی نیست. چون در بسیاری از موارد پیچیدگی مسئله آنقدر بالاست که با یک تابع خطی حل نشود. پس نیاز ب ه یک تابع غیرخطی داریم. همچنین در نمودار پتانسیل عمل دیدیم که باید قدرت سیگنال از یک مقدار آستانه بیشتر باشد تا نورون بتواند آن را انتقال دهد. پس اگر بخواهیم نورون مصنوعی را طوری شبیه‌سازی کنیم که این دو مشکل (غیرخطی و پیچیده بودن مدل و حد آستانه قدرت سیگنال) را برطرف کند، باید از یک تابع به نام تابع فعال‌ساز یا Activation Function استفاده کنیم که این تابع تغییراتی روی مقدار محاسبه‌شده می‌دهد و خروجی نهایی را مشخص می‌کند.  
</p>   

<p style="font-family:B Nazanin">
     تابع فعال‌ساز می‌تواند خطی یا غیرخطی باشد (که البته معمولا غیرخطی است) که بر اساس شرایط مختلف مسئله می‌توان توابع مختلفی را برای این کار استفاده کرد. 
</p>

<p style="font-family:B Nazanin">
     مشابه با رگرسیون، اگر بخواهیم همه این عملیات را به صورت برداری بنویسیم، مقدار بایاس را $w_0$ در نظر می‌گیریم و آن را معادل یک ورودی با مقدار همیشه یک در نظر می‌گیریم ($x_0 = 1$). سپس:
</p>

$$
w = [w_0, w_1, w_2, \cdots, w_m] \qquad x = [1, x_1, x_2, \cdots, x_m]
$$

$$
z = w. x^T = x .w^T
$$

<p style="font-family:B Nazanin">
    <b>توابع فعال‌ساز:</b> توابعی که در شکل زیر معرفی شده‌اند، تعدادی از توابع پرکاربرد هستند که می‌توانند به عنوان توابع فعال‌ساز استفاده شوند. 
</p>

<p style="font-family:B Nazanin">
    <b>آموزش پرسپترون:</b> فرایند آموزش دادن یک پرسپترون یا نورون مصنوعی به این صورت است که داده‌های ورودی یعنی بردار $x$ توسط وزن‌های اولیه و مقدار اولیه بایاس با یکدیگر جمع می‌شوند، سپس مقدار عددی به دست آمده به تابع فعال‌ساز داده می‌شود  تا مقدار پیش‌بینی‌شده ($y_p$) به دست آید. مقدار پیش‌بینی‌شده را با مقدار واقعی ($y_t$) مقایسه می‌کنیم. با استفاده از یکی از معیارها مقدار خطا را محاسبه می‌نماییم (مثلا معیار MSE که می‌شود: $e = (y_p - y_t)^2$). در نهایت پس از محاسبه مقدار خطا، اگر این مقدار خطا زیاد بود با استفاده از یک بهینه‌ساز یا Optimizer مقادیر وزن‌ها و بایاس را تغییر می‌دهیم و مجددا همین مراحل را تکرار می‌کنیم. مثلا می‌توانیم از روش گرادیان کاهشی استفاده کنیم تا وزن‌های اصلاح‌شده را به دست آوریم:
</p>

<p style="font-family:B Nazanin">
     در این روش دو هایپرپارامتر خواهیم داشت: یکی $\mu$ که ضریبی است که در روش گرادیان کاهشی استفاده می‌کنیم تا تغییرات وزن‌ها در هر مرحله را کنترل کنیم و دیگری $N$ که تعداد کل تکرارها است. اگر تعداد کل تکرارها از این عدد بیشتر شد، فرایند متوقف می‌شود و خروجی مشخص خواهد شد اما اگر قبل از رسیدن به این تعداد تکرار، خطا از مقدار مورد نظر کمتر شد، در آن لحظه فرایند تکرار را متوقف می‌کنیم و خروجی را مشخص می‌نماییم. 
</p>

<p style="font-family:B Nazanin">
    <b>شبکه عصبی چندلایه یا MultiLayer Perceptron یا MLP:</b><br>
    تا اینجا با نحوه کار نورون‌های مصنوعی یا پرسپترون آشنا شدیم که چگونه داده‌های ورودی را می‌گیرد و پردازش می‌کند و خروجی را مشخص می‌سازد. اما یک نورون تنها فقط می‌تواند ارتباطات ساده بین ورودی‌ها را تشخیص دهد. اما اگر نیاز داشته باشیم تا ارتباطات پیچیده‌تری را بین داده‌ها کشف کنیم، نیاز داریم تا تعداد بیشتری از نورون‌ها را در کنار هم قرار دهیم تا بتوانند رابطه‌های پیچیده‌تری را بین داده‌های ورودی کشف کنند. برای این کار می‌توانیم از شبکه عصبی MLP استفاده کنیم که یکی از ساده‌ترین شبکه‌های عصبی مصنوعی است. از این مدل شبکه عصبی می‌توانیم هم در مسائل رگرسیون و هم در مسائل دسته‌بندی استفاده کنیم.  
</p>

<p style="font-family:B Nazanin">
     این شبکه، داده‌ها را از یک لایه ورودی (Input Layer) دریافت می‌کند. پس اولین لایه یک شبکه عصبی چندلایه، لایه ورودی است. سپس این داده‌های ورودی به همراه وزن‌های تخصیص‌یافته به هر کدام، وارد لایه‌های مخفی (Hidden Layers) می‌شوند. هر کدام از این لایه‌ها شامل تعدادی نورون است، که روی داده‌ها پردازش می‌کنند و داده‌ها را به لایه بعدی تحویل می‌دهند. در نهایت داده به لایه آخر یا لایه خروجی (Output Layer) می‌رسد که در این لایه نیز یک سری نورون وجود دارد که روی داده‌های به دست آمده پردازش انجام می‌دهند و درنهایت خروجی نهایی مشخص می‌شود. این ساختار امکان کشف ارتباطات پیچیده‌تری بین داده‌ها را برای ما فراهم می‌آورد.<br>
    همان طور که در شکل‌های زیر مشاهده می‌شود، نهایت قدرت یک پرسپترون تنها این است که یک مسئله رگرسیون خطی ساده یا یک مسئله دسته‌بندی ساده مشابه شکل‌های زیر را حل کند.
</p>

<p style="font-family:B Nazanin">
     اما به عنوان مثال اگر مسئله دسته‌بندی (Classification) پیچیده‌تر شد و داده‌ها مانند شکل زیر دیگر تنها با یک خط ساده قابل جداسازی نبودند، قطعا با یک پرسپترون تنها نمی‌توان این مسئله را حل کرد و مدل مناسبی را آموزش داد. بلکه باید از چندین نورون در چندین لایه استفاده کرد. 
</p>

<p style="font-family:B Nazanin">
     در لینک زیر می‌تواند تاثیر کم و زیاد کردن تعداد نورون‌ها و تعداد لایه‌ها را در نحوه دسته‌بندی داده‌ها به صورت شهودی ببینید. <br>
    https://playground.tensorflow.org/
</p>

<p style="font-family:B Nazanin">
     می‌خواهیم با یک مثال روش کار را بیشتر توضیح دهیم. فرض کنید در یک مسئله مطابق با شکل زیر، یک لایه ورودی با دو گره ($x_1$ و $x_2$)، یک لایه مخفی با 5 پرسپترون ($h_{11}$ تا $h_{15}$) و یک لایه خروجی با 3 گره ($z_1$ و $z_2$ و $z_3$) داریم. فرض کنید نورون‌ها به صورت Fully Connected هستند. یعنی هر پرسپترون از همه ورودی‌ها داده دریافت می‌کند. دقت کنید که هر کدام از پرسپترون‌های لایه مخفی و هر کدام از پرسپترون‌های لایه خروجی، برای خودشان یک تابع جمع‌کننده (برای محاسبه مجموع وزن‌دار داده‌های ورودی) و یک تابع فعال‌ساز دارند. راهنمای نمادگذاری پرسپترون‌ها و وزن‌ها در شکل زیر مشخص شده است. <br>
    توجه: برای پرسپترون‌های مختلف، می‌توان از توابع فعال‌ساز متفاوت استفاده کرد اما معمولا توابع فعال‌ساز پرسپترون‌های مختلف در همه لایه‌ها یکسان است. (در شبکه‌های پیچیده‌تر می‌توان متفاوت در نظر گرفت). <br>
    توجه: هر پرسپترون یک مقدار متفاوت از بایاس را دارد و این طور نیست که یک بایاس ثابت برای همه پرسپترون‌ها در نظر بگیریم.
</p>

<p style="font-family:B Nazanin">
     در ابتدا اگر فقط لایه اول را بررسی کنیم:
</p>

<p style="font-family:B Nazanin">
     فرض کنیم که تابع فعال‌ساز همه پرسپترون‌ها $f$ باشد. بنابراین روابط ریاضی خروجی این پرسپترون‌ها به صورت زیر خواهد بود:
</p>

$$
\begin{align}
y_1 &= f(1\times w_{0,1}^1 + x_1\times w_{1,1}^1 + x_2\times w_{2,1}^1)\\
y_2 &= f(1\times w_{0,2}^1 + x_1\times w_{1,2}^1 + x_2\times w_{2,2}^1)\\
y_3 &= f(1\times w_{0,3}^1 + x_1\times w_{1,3}^1 + x_2\times w_{2,3}^1)\\
y_4 &= f(1\times w_{0,4}^1 + x_1\times w_{1,4}^1 + x_2\times w_{2,4}^1)\\
y_5 &= f(1\times w_{0,5}^1 + x_1\times w_{1,5}^1 + x_2\times w_{2,5}^1)\\
\end{align}
$$

<p style="font-family:B Nazanin">
     که در اینجا $y$ها همان $h$ها هستند. یعنی: $y_1 = h_{11}$ و $y_2 = h_{12}$ و $y_3 = h_{13}$ و $y_4 = h_{14}$ و $y_5 = h_{15}$. <br>
    اگر همه داده‌ها را به صورت ماتریسی بنویسیم:
</p>

$$
W = \begin{bmatrix}
w_{0,1}^1 & w_{1,1}^1 & w_{2,1}^1 \\
w_{0,2}^1 & w_{1,2}^1 & w_{2,2}^1 \\
w_{0,3}^1 & w_{1,3}^1 & w_{2,3}^1 \\
w_{0,4}^1 & w_{1,4}^1 & w_{2,4}^1 \\
w_{0,5}^1 & w_{1,5}^1 & w_{2,5}^1 
\end{bmatrix}_{5\times3}
\Rightarrow
W^T = \begin{bmatrix}
w_{0,1}^1 & w_{0,2}^1 & w_{0,3}^1 & w_{0,4}^1 & w_{0,5}^1 \\
w_{1,1}^1 & w_{1,2}^1 & w_{1,3}^1 & w_{1,4}^1 & w_{1,5}^1 \\
w_{2,1}^1 & w_{2,2}^1 & w_{2,3}^1 & w_{2,4}^1 & w_{2,5}^1 
\end{bmatrix}_{3\times5}
$$

$$
\begin{align}
X &= \begin{bmatrix}1 & x_1 & x_2 \end{bmatrix}_{1\times 3} \\
Y &= \begin{bmatrix}y_1 & y_2 & y_3 & y_4 & y_5 \end{bmatrix}_{1\times 5}
\end{align}
$$

<p style="font-family:B Nazanin">
     در این صورت داریم:
</p>

$$
Y = X.W^T = \begin{bmatrix}1 & x_1 & x_2 \end{bmatrix}_{1\times 3} \times \begin{bmatrix}
w_{0,1}^1 & w_{0,2}^1 & w_{0,3}^1 & w_{0,4}^1 & w_{0,5}^1 \\
w_{1,1}^1 & w_{1,2}^1 & w_{1,3}^1 & w_{1,4}^1 & w_{1,5}^1 \\
w_{2,1}^1 & w_{2,2}^1 & w_{2,3}^1 & w_{2,4}^1 & w_{2,5}^1 
\end{bmatrix}_{3\times5}
$$

<p style="font-family:B Nazanin">
     به همین ترتیب برای لایه خروجی داریم:
</p>

$$
\begin{align}
z_1 &= f(1\times w_{0,1}^2 + h_{11}\times w_{1,1}^2 + h_{12}\times w_{2,1}^2 + h_{13}\times w_{3,1}^2 + h_{14}\times w_{4,1}^2 + h_{15}\times w_{5,1}^2)\\
z_2 &= f(1\times w_{0,2}^2 + h_{11}\times w_{1,2}^2 + h_{12}\times w_{2,2}^2 + h_{13}\times w_{3,2}^2 + h_{14}\times w_{4,2}^2 + h_{15}\times w_{5,2}^2)\\
z_3 &= f(1\times w_{0,3}^2 + h_{11}\times w_{1,3}^2 + h_{12}\times w_{2,3}^2 + h_{13}\times w_{3,3}^2 + h_{14}\times w_{4,3}^2 + h_{15}\times w_{5,3}^2)\\
\end{align}
$$

<p style="font-family:B Nazanin">
     اگر به صورت ماتریسی بنویسیم:
</p>

$$
W = \begin{bmatrix}
w_{0,1}^2 & w_{1,1}^2 & w_{2,1}^2 & w_{3,1}^2 & w_{4,1}^2 & w_{5,1}^2 \\
w_{0,2}^2 & w_{1,2}^2 & w_{2,2}^2 & w_{3,2}^2 & w_{4,2}^2 & w_{5,2}^2 \\
w_{0,3}^2 & w_{1,3}^2 & w_{2,3}^2 & w_{3,3}^2 & w_{4,3}^2 & w_{5,3}^2 \\
\end{bmatrix}_{3\times6}
\Rightarrow
W^T = \begin{bmatrix}
w_{0,1}^2 & w_{0,2}^2 & w_{0,3}^2 \\
w_{1,1}^2 & w_{1,2}^2 & w_{1,3}^2 \\
w_{2,1}^2 & w_{2,2}^2 & w_{2,3}^2 \\
w_{3,1}^2 & w_{3,2}^2 & w_{3,3}^2 \\
w_{4,1}^2 & w_{4,2}^2 & w_{4,3}^2 \\
w_{5,1}^2 & w_{5,2}^2 & w_{5,3}^2 \\
\end{bmatrix}_{6\times3}
$$

$$
\begin{align}
Y &= \begin{bmatrix}1 & h_{11} & h_{12} & h_{13} & h_{14} & h_{15}\end{bmatrix}_{1\times6} \\
Z &= \begin{bmatrix}z_1 & z_2 & z_3\end{bmatrix}_{1\times3}
\end{align}
$$

<p style="font-family:B Nazanin">
     که در این صورت داریم:
</p>

$$
Z = Y.W^T = \begin{bmatrix}1 & h_{11} & h_{12} & h_{13} & h_{14} & h_{15}\end{bmatrix}_{1\times6} \times \begin{bmatrix}
w_{0,1}^2 & w_{0,2}^2 & w_{0,3}^2 \\
w_{1,1}^2 & w_{1,2}^2 & w_{1,3}^2 \\
w_{2,1}^2 & w_{2,2}^2 & w_{2,3}^2 \\
w_{3,1}^2 & w_{3,2}^2 & w_{3,3}^2 \\
w_{4,1}^2 & w_{4,2}^2 & w_{4,3}^2 \\
w_{5,1}^2 & w_{5,2}^2 & w_{5,3}^2 \\
\end{bmatrix}_{6\times3}
$$

<p style="font-family:B Nazanin">
     بنابراین می‌بینیم که محاسبات نسبتا ساده هستند و از طریق ضرب‌های ماتریسی محاسبه می‌شوند. <br>
    در ادامه حالت‌های مختلف ورودی و خروجی را بررسی می‌کنیم:<br>
    * حالت اول: <b>m ورودی و 1 نورون:</b>
</p>

<p style="font-family:B Nazanin">
     اگر m ورودی و یک نورون داشته باشیم، خروجی به صورت زیر خواهد بود:
</p>

$$
h = x_1 w_1 + x_2w_2 + \cdots + x_mw_m + b
$$

<p style="font-family:B Nazanin">
     که اگر به صورت ماتریسی بنویسیم:
</p>

$$
h = xw + b
$$

<p style="font-family:B Nazanin">
     که در آن $x$ و $w$ به صورت بردار و $h$ و $b$ به صورت عدد حقیقی هستند:
</p>

$$
\begin{matrix}x\in \mathbb{R}^m & w\in \mathbb{R}^m & h\in \mathbb{R} & b\in \mathbb{R}\end{matrix}
$$

<p style="font-family:B Nazanin">
    * حالت دوم: <b>1 ورودی و k نورون:</b>
</p>

<p style="font-family:B Nazanin">
     اگر یک ورودی و k نورون داشته باشیم، خروجی به صورت زیر خواهد بود:
</p>

$$
\begin{align}
h_1 &= xw_1 + b_1 \\
h_2 &= xw_2 + b_2 \\
h_3 &= xw_3 + b_3 \\
&\vdots \\
h_k &= xw_k + b_k
\end{align}
$$

<p style="font-family:B Nazanin">
     که در حالت ماتریسی به صورت زیر خواهد بود:
</p>

$$
h = xW + b
$$

<p style="font-family:B Nazanin">
     که در آن ابعاد متغیرها به صورت زیر خواهد بود:
</p>

$$
\begin{matrix}x\in\mathbb{R}^d & W\in\mathbb{R}^{d\times k} & b\in\mathbb{R}^k & h\in\mathbb{R}^k\end{matrix}
$$

<p style="font-family:B Nazanin">
     این موارد با این فرض است که $x$ یک ویژگی از دیتاست ما است که در این دیتاست به اندازه $d$ نمونه (رکورد) داریم.
</p>

<p style="font-family:B Nazanin">
    بنابراین برای یک <b>شبکه عصبی با m ویژگی و یک لایه با k نورون و n نمونه در دیتاست</b>، فرمول کلی به صورت زیر خواهد بود: 
</p>

$$
h_{n\times k} = X_{n\times m} \cdot W_{m\times k} + b_{n\times k}
$$

     بنابراین شمای کلی یک شبکه MLP به صورت زیر خواهد بود:
</p>
