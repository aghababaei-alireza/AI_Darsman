<p style = "font-family:B Nazanin;">
    <b>همبستگی (Correlation):</b> تا اینجا با استفاده از آمار توصیفی یکی از داده‌ها را در نظر می‌گرفتیم و با ابزارهای مختلف مثل محاسبه میانگین، میانه، مد و ... آن را تحلیل می‌کردیم. به این ابزارها، ابزارهای تک‌متغیره می‌گوییم. یعنی فقط روی یک متغیر کار می‌کنند. اما متغیرهای مختلف ممکن است با هم ارتباط داشته باشند. یعنی تغییر در یکی باعث تغییر در دیگری شود. به عبارت دیگر همبستگی داشته باشند. در این صورت نیاز به یک ابزار تحلیل داریم که دومتغیره باشد. همبستگی یک ابزار دومتغیره است که قدرت و جهت ارتباط بین دو متغیر را اندازه‌گیری می‌کند. <br>
    جهت همبستگی، نحوه تغییرات دو متغیر را نسبت به یکدیگر نشان می‌دهد. اگر با افزایش یک متغیر، متغیر دوم نیز افزایش پیدا کند، می‌گوییم بین این دو متغیر همبستگی مثبت وجود دارد. به عبارت دیگر تغییرات این دو متغیر هم‌جهت و هم‌راستا است. اما اگر با افزایش یک متغیر، متغیر دوم کاهش یابد، می‌گوییم بین این دو متغیر همبستگی منفی وجود دارد. <br>
    میزان و شدت ارتباط بین دو متغیر با قدرت یا ضریب همبستگی تعیین می‌شود. ضریب همبستگی عددی بین -1 تا +1 است. هر چه همبستگی و ارتباط بین دو متغیر شدیدتر باشد، قدرمطلق ضریب همبستگی به یک نزدیک‌تر است و هر چه ارتباط بین دو متغیر کمتر باشد، ضریب همبستگی به صفر نزدیک‌تر است. <br>
    <b>انواع همبستگی:</b><br>
    * <b>همبستگی خطی:</b> اگر ارتباط و همبستگی بین دو متغیر از نوع خطی باشد یعنی نمودار تغییرات یک متغیر بر حسب متغیر دوم یک خط باشد، به ارتباط بین دو متغیر همبستگی خطی یا Linear Correlation می‌گویند. <br>
    * <b>همبستگی صفر:</b> اگر بین دو متغیر نتوان ارتباط بامعنایی پیدا کرد، می‌گوییم این دو متغیر با هم همبستگی صفر دارند (No Correlation). <br>
    * <b>همبستگی غیر خطی:</b> اگر ارتباط بین دو متغیر به صورت غیرخطی باشد، می‌گوییم همبستگی غیرخطی یا Non-linear Correlation وجود دارد.
</p>

<p style = "font-family:B Nazanin;">
    <b>متدهای همبستگی (Correlation Methods):</b> با استفاده از متدهای زیر می‌توان میزان همبستگی دو متغیر را محاسبه کرد:<br>
    * <b>متد Pearson (پیرسون):</b> یکی از اصلی‌ترین و پرکاربردترین متدهای تشخیص میزان همبستگی است که زمانی از این متد استفاده می‌شود که هر دو متغیر مورد نظر دارای توزیع نرمال باشند. <br>
    * <b>متد Kendall (کندال)</b><br>
    * <b>متد Spearman (اسپیرمن)</b><br>
    دو متد kendall و spearman زمانی استفاده می‌شوند که یا توزیع داده‌ها نرمال نباشد یا اساسا ندانیم که توزیع داده‌ها به چه صورت است و همچنین زمانی که داده‌ها شامل داده پرت نیز باشد.
</p>

<p style = "font-family:B Nazanin;">
    <b>سری زمانی (Time Series):</b><br>
    به مجموعه‌ای از داده‌ها که در یک دیتاست به ترتیب زمان ثبت شده‌اند، سری زمانی یا Time series می‌گویند. بعضی از داده‌ها وابسته به زمان هستند. مثلا در یک سیستم فروشگاهی هر روز تعدادی محصول فروخته خواهد شد. هر محصول بعد از محصول قبلی. بسیاری از دیتاست‌ها این ویژگی را دارند که تعدادی از فیلدهای آنها (و شاید همه فیلدها) بر اساس زمان درج شده باشند. ویژگی‌های اصلی این داده‌ها به صورت زیر است:<br>
    1. داده‌ها به ترتیب زمان ثبت شده‌اند (یا از گذشته به آینده یا از آینده به گذشته) <br>
    2. پریود زمانی داده‌ها ثابت است. یعنی معمولا نظم خاصی در ورود اطلاعات در آن‌ها وجود دارد. مثلا فاصله زمانی بین داده‌ها معمولا یکسان است. <br>
    3. همبستگی زمانی: داده‌های سری زمانی نسبت به زمان همبستگی دارند مثلا فروش یک فروشگاه آنلاین به فروش محصولات دیجیتال آن ربط دارد. هر چه زمان کار بیشتر باشد، مثلا میزان فروش نیز بیشتر است. <br>
    4. در طول زمان می‌توان سیگنال‌های محتلفی در این داده‌های سری زمانی پیدا کرد. یعنی مثلا در یک سایت فروش آنلاین متوجه می‌شویم که در اوایل فصل تابستان میزان فروش بیشتر است که بر اساس آن می‌توان الگوهایی را شناسایی کرد. <br>
    <br>
    مهم‌ترین کاربرد داده‌های سری زمانی پیش‌بینی آینده است. با تحلیل داده‌ها و پیدا کردن الگوها، می‌توان پیش‌بینی کرد که در آینده رفتار این داده‌ها چگونه خواهد بود. یا با تحلیل داده‌های سری زمانی می‌توانیم تحلیل موجودی انجام دهیم. مثلا موجودی اولیه یک محصول را می‌دانیم و با بررسی داده‌های فروش محصول در طول زمان بفهمیم که اکنون چه مقدار موجودی از آن محصول داریم. یا با بررسی یک سیستم و عملکرد آن که در طول زمان ثبت شده‌است، عملکرد آن را بهینه کنیم. <br>
    <br>
    برای بررسی داده‌های سری زمانی باید: 1. داده‌ها را بخوانیم. 2. بر اساس تاریخ و زمان آن‌ها را ایندکس‌گذاری کنیم. مانند داده‌های دیگر که به هر رکورد یک آیدی یکتا نسبت می‌دادیم، در این داده‌ها بهتر است که تاریخ و زمان معیار ایندکس‌گذاری باشد. 3. نویزها را پیدا می‌کنیم. 4. پردازش داده‌های پرت را انجام می‌دهیم. 5. اگر لازم است تقسیم‌بندی انجام می‌دهیم. مثلا داده‌های یک سال را انتخاب می‌کنیم و در آن داده‌های هر فصل را جدا می‌کنیم و هر قسمت یک مجموعه داده سری زمانی مجزا می‌شود. 
</p>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
data = pd.read_csv('./data/sales_data_for_time_series.csv')
data.info()
```

```python
data['Date'] = pd.to_datetime(data['Date'])
data.info()
```

```python
data.set_index('Date', inplace=True)
data
```

```python
data['Sales'].fillna(data['Sales'].mean(), inplace=True)
data
```

<p style = "font-family:B Nazanin;">
    تابع resample تابعی است که داده‌های یک دیتافریم را بر اساس یک بازه زمانی دسته‌بندی می‌کند و می‌تواند روی هر دسته، عملیات ریاضی مانند مجموع، میانگین و ... را انجام دهد. مثلا در مثال بالا که میزان فروش در تاریخ‌های متفاوت وجود دارد می‌توان داده‌ها را به صورت ماهانه یا سالانه دسته‌بندی کرد و مجموع فروش در هر ماه یا هر سال را محاسبه کرد. تابع resample یک پارامتر ورودی می‌گیرد که نوع دسته‌بندی را مشخص می‌کند. پارامتر 'M' برای دسته‌بندی ماهانه، 'Y' برای دسته‌بندی سالانه و ...
</p>

```python
monthly_data = data.resample('M').sum() # مجموع فروش ماهانه
monthly_data
```

```python
yearly_data = data.resample('Y').count() # تعداد فروش سالانه
yearly_data
```

```python
monthly_data = monthly_data[~monthly_data.index.duplicated(keep='first')]
monthly_data
```

```python
plt.figure(figsize=(15, 7))
plt.plot(monthly_data.index, monthly_data['Sales'])
plt.axhline(y=monthly_data['Sales'].mean(), color='y', linestyle='--', linewidth=2, label='Mean')
plt.axvline(x=monthly_data['Sales'].min(), color='r', linestyle='--', linewidth=2, label='Start')
plt.axvline(x=monthly_data['Sales'].max(), color='g', linestyle='--', linewidth=2, label='End')
plt.gca().xaxis.set_major_locator(mdates.YearLocator(1)) # بازه تقسیم محور افقی (یک سال یک سال)
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime
```

```python
data = {
    'Date': ['2022-01-01', '2022-01-02', '2022-01-03', '2022-01-04', '2022-01-05'],
    'Value': [10, 15, 20, 25, 30]
}
data
```

```python
df = pd.DataFrame(data)
df
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.info()
```

<p style = "font-family:B Nazanin;">
     در یک دیتافریم اگر یک ستون به صورت datetime ذخیره شده‌باشد، یک مشخص به نام dt دارد که با استفاده از آن می‌توان به توابعی که روی datetime کار می‌کند دسترسی داشته باشیم. مثلا تابع strftime که نحوه نمایش رشته‌ای تاریخ را مشخص می‌کند. در مثال زیر یک ستون به دیتافریم اضافه می‌کنیم و همان تاریخ را با فرمتی خاص داخل آن ذخیره می‌کنیم:
</p>

```python
df['Formatted_Date'] = df['Date'].dt.strftime('%Y/%m/%d')
df
```

```python
plt.figure(figsize=(15, 8))
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Time Series Data')
plt.plot(df['Date'], df['Value'], marker='o')

plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=1))
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.xticks(rotation=45)

plt.text(df['Date'].min(), df['Value'].max(), 'Start', color='r')
plt.text(df['Date'].max(), df['Value'].max(), 'Start', color='g')

for i, txt in enumerate(df['Formatted_Date']):
    plt.annotate(txt, (df['Date'][i], df['Value'][i]), textcoords="offset points", xytext=(0, 10), ha='center')
    
plt.show()
```
