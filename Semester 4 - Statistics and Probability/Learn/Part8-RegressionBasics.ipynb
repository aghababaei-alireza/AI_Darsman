    <h1>فصل چهارم: آمار و احتمالات</h1>
    <h2>بخش هشتم: آشنایی اولیه به رگرسیون و انواع مدل‌های آن</h2>
</span>

<p style="font-family:B Nazanin">
   رگرسیون الگوریتمی است که می‌توانیم از آن برای مدل‌سازی و تحلیل‌های آماری استفاده کنیم. اگر در بین داده‌های موجود که قرار است تحلیل شوند، به مدل خاصی برسیم، می‌توانیم داده‌ها و رفتار آن را پیش‌بینی کنیم. در این مدل‌ها تعدادی ویژگی مستقل داریم که تغییر کردن آن‌ها با هم روی تغییر کردن یک سری متغیر وابسته اثر می‌گذارند. مثلا در بررسی میزان فروش یک محصول، تبلیغ کردن، راه‌اندازی یک وبسایت و ... می‌تواند میزان فروش آن محصول را کمتر یا بیشتر کند. در اینجا رگرسیون می‌تواند بر اساس آن  متغیرهای مستقل و متغیر وابسته مدلی را پیاده‌سازی می‌کند و زمانی که یک مدل به‌دست آوردیم می‌توانیم هم متغیر وابسته را پیش‌بینی کنیم، هم متغیرهای مستقل را بهینه و یا کنترل کنیم. <br>
    رگرسیون انواع و مدل‌های مختلفی دارد ولی در ساده‌ترین حالت آن اگر فرض کنیم متغیر مستقل ما x و متغیر وابسته ما y باشد، داده‌های x و y نقاطی را روی صفحه x-y مشخص می‌کنند. خط رگرسیون خطی است که یک رابطه بین x و y مشخص می‌کند که در هر نقطه مقدار به‌دست آمده برای y (با استفاده از رابطه خط رگرسیون) کمترین خطا را با مقدار واقعی y متناطر با x مورد نظر داشته باشد. 
</p>

<p style="font-family:B Nazanin">
   نکته مهم این است که متغیر وابسته حتما باید پیوسته باشد وگرنه رگرسیون امکان‌پذیر نیست. همان طور که در شکل زیر مشخص است مقدار y هر نقطه با مقدار y به‌دست آمده از خط رگرسیون خطایی دارد که محاسبه خط رگرسیون به صورتی است که مجموع این خطاها برای نقاط مختلف کمترین مقدار باشد.
</p>

<p style="font-family:B Nazanin">
   همان طور که می‌بینید در حالت رگرسیون خطی معادله خط رگرسیون به صورت $y = ax + b$ خواهد بود که در آن $b$ عرض از مبدا و $a$ شیب خط خواهد بود. ضمنا هر مقدار y واقعی با مقدار y به‌دست آمده از معادله خط رگرسیون اختلافی خواهد داشت که به آن خطا می‌گویند. a و b به گونه‌ای محاسبه می‌شوند که مجموع خطا برای نقاط مختلف کمترین مقدار ممکن باشد. 
</p>

<p style="font-family:B Nazanin">
    <b>انواع رگرسیون:</b><br>
    * رگرسیون خطی مانند رگرسیون خطی ساده، رگرسیون خطی چند متغیره<br>
    * رگرسیون غیر خطی مانند رگرسیون چندجمله‌ای، رگرسیون لجستیک، رگرسیون نمایی، رگرسیون پواسن
</p>

<p style="font-family:B Nazanin">
   به طور کلی فرمول کلی رگرسیون خطی چندمتغیره به صورت زیر است:
</p>

$$
Y = \beta_0 + \beta_1 X_1 + \beta_2 X_2 + \dots + \epsilon
$$

<p style="font-family:B Nazanin">
   که در آن $X_i$ ها همان متغیرهای مستقل و $Y$ متغیر وابسته است. $\beta_0$ عرض از مبدا و سایر $\beta_i$ ها شیب یا ضریب مربوط به هر متغیر مستقل است و در نهایت $\epsilon$ معرف خطا است. 
</p>

<p style="font-family:B Nazanin">
    <b>معیارهای کلیدی در تحلیل رگرسیون:</b><br>
    * SST یا Sum of Squares Total: مجموع مربعات اختلاف yهای واقعی و میانگینشان<br>
    * SSR یا Sum of Squares Regression: مجموع مربعات اختلاف yهای تخمین‌رده‌شده ($y'$) و میانگینشان<br>
    * SSE یا Sum of Squares Error: مجموع مربعات خطاها (اختلاف yهای واقعی و تخمین‌زده‌شده)<br>
    * ضریب تعیین یا Coefficient of Determination یا R2: معیاری برای تعیین مناسب بودن خط رگرسیون است که عددی بین صفر و یک است و هر چه به یک نزدیک‌تر باشد مناسب‌تر است. <br>
    در فرمول‌های زیر $y$ مقدار واقعی متغیر وابسته و $y'$ مقدار تخمین‌زده‌شده برای متغیر وابسته است.
</p>

$$
SST = \Sigma (y - \bar{y})^2
$$

$$
SSR = \Sigma (y' - \bar{y'})^2
$$

$$
SSE = \Sigma (y - y')^2
$$

$$
R^2 = \frac{SSR}{SST} = 1 - \frac{SSE}{SST}
$$
