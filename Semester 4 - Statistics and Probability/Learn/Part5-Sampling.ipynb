<p style="font-family:B Nazanin">
   دلایل استفاده از نمونه‌گیری و عدم انجام تحلیل روی کل جامعه: <br>
    * انجام تحلیل روی کل جامعه ممکن است هزینه بسیار زیادی برای ما داشته باشد. مثلا تحلیل حدود 10 میلیون داده ممکن است هزینه سخت‌افزاری و همچنین زمان زیادی را صرف کند. <br>
    * گاهی اوقات جامعه آنقدر یکنواخت و همگن هست که تحلیل بخشی از آن به راحتی به کل جامعه تعمیم داده می‌شود. <br>
    * گاهی اوقات لازم است یک تست را روی بخشی از جامعه انجام دهیم که هزینه‌بر یا زمان‌بر است. مثلا می‌خواهیم بررسی کنیم که یک سپر ماشین که در یک کارخانه ساخته شده‌است با چه ضربه‌ای می‌شکند. اگر بخواهیم که تمام سپرهای ساخته‌شده در کارخانه را تست کنیم که اصلا شدنی نیست و ضرر بسیار بزرگی زده می‌شود.<br>
    * ... <br>
    دو مسئله بسیار مهم در بحث نمونه‌برداری وجود دارد. یکی این که چگونه نمونه را انتخاب کنیم و دوم اینکه اندازه نمونه چقدر باشد. این دو سوالی است که در این بخش پاسخ داده خواهد شد. <br>
    <br>
    دو نوع نمونه برداری داریم: نمونه‌برداری تصادفی (Random Sampling) و نمونه‌برداری غیرتصادفی (Non-random Sampling).<br>
    زمانی از نمونه‌برداری تصادفی استفاده می‌کنیم که شانس انتخاب اعضای جامعه برای نمونه‌برداری یکسان باشد که اکثر مواقع از این روش استفاده می‌کنیم. اما اگر در جامعه ویژگی خاصی وجود داشته باشد که شانس انتخاب اعضای جامعه را متفاوت کند یا مواردی وجود دارد که نظم جامعه را به هم زده است، از روش‌های غیرتصادفی استفاده می‌کنیم. <br>
    <br>
    <b>انواع نمونه‌برداری تصادفی:</b><br>
    * <b>نمونه‌برداری ساده (Simple Sampling):</b> در این روش به صورت کاملا تصادفی و ساده تعدادی از اعضای جامعه را به عنوان نمونه انتخاب می‌کنیم. در این روش هر عضو جامعه به اندازه $\frac{1}{N}$ شانس انتخاب شدن دارد. این روش نیز به دو صورت انجام خواهد شد: <br>
    -- نمونه‌برداری ساده بدون جایگذاری: یعنی هر عضوی که برای نمونه انتخاب شد از جامعه حذف شده و اعضای بعدی از بین اعضای باقیمانده انتخاب می‌شود. <br>
    -- نمونه‌برداری ساده با جایگذاری: یعنی هر عضوی که برای نمونه انتخاب شد، از جامعه حذف نمی‌شود و اعضای بعدی نمونه مجددا از بین کل اعضای جامعه انتخاب خواهند شد. در این حالت این احتمال وجود دارد که یک عضو بیش از یک بار در نمونه انتخاب شود.
</p>

<p style="font-family:B Nazanin">
    * <b>نمونه‌برداری سیستماتیک یا منظم (Systematic Sampling):</b> در این روش عناصر نمونه بر اساس یک سری فواصل منظم انتخاب می‌شوند. برای استفاده از این روش تمام اعضای جامعه باید بر اساس یک معیار مرتب و منظم شده باشند. این روش از روش ساده، آسان‌تر است و نتیجه منظم‌تری به ما می‌دهد. در این روش نقطه شروع نیز اهمیت دارد و تغییر نقطه شروع نیز نمونه را تغییر می‌دهد. مثلا در یک اداره 100 پرونده وجود دارد که توسط دو کارمند تنظیم شده است. می‌دانیم کارمند اول تخلفی نداشته است اما کارمند دوم که پس از کارمند اول آمده‌است احتمال تخلف در تنظیم پرونده‌هایش وجود دارد. در اینجا برای بررسی پرونده‌ها اگر از روش ساده استفاده کنیم، ممکن است تعداد زیادی از پرونده‌ها مربوط به کارمند اول باشد که مطمئنیم تخلفی ندارد. پس برای نمونه‌برداری از پرونده‌ها ابتدا آن‌ها را بر اساس تاریخ مرتب می‌کنیم و سپس نقطه شروع را تاریخی انتخاب می‌کنیم که کارمند دوم آمده‌است و از آنجا با روش سیستماتیک نمونه‌برداری می‌کنیم.
</p>

<p style="font-family:B Nazanin">
    * <b>نمونه‌برداری دسته‌ای (Stratified Sampling):</b> این روش یکی از بهترین روش‌های نمونه‌برداری است. در این روش جامعه را به تعدادی دسته یا گروه تقسیم‌بندی می‌کنیم که این تقسیم‌بندی بر اساس یک معیار یا ویژگی مشترک انجام خواهد شد. سپس از هر دسته یا گروه تعدادی از اعضا را به صورت تصادفی انتخاب می‌کنیم. مزیت این روش این است که مطمئن هستیم از هر کدام از دسته‌ها حداقل یک یا چند عضو انتخاب شده‌اند و هیچ دسته‌ای بدون عضو انتخاب‌شده نیست.
</p>

<p style="font-family:B Nazanin">
    * <b>نمونه‌برداری خوشه‌ای (Cluster Sampling):</b> در این روش نیز مانند روش قبل، اعضای جامعه به چند خوشه تقسیم‌بندی می‌شوند اما در این روش، کل اعضای یک خوشه در نمونه قرار می‌گیرند. 
</p>

<p style="font-family:B Nazanin">
    * <b>نمونه‌برداری چندمرحله‌ای (Multistage Sampling):</b> این روش ترکیب روش‌های قبل است که می‌توان در چند مرحله و در هر مرحله از یکی از روش‌های قبل استفاده کرد تا نمونه‌برداری انجام شود. این روش هزینه بسیار زیادی دارد ولی در عوض دقت بسیار بالاتری ارائه می‌دهد و ضعف تمام روش‌های قبل را می‌پوشاند. به عنوان مثال در شکل زیر در ابتدا کل اعضای جامعه خوشه‌بندی شده‌اند به طوری که در هر خوشه چهار نفر آبی و 2 نفر سبز وجود داشته باشند. بعد به صورت تصادفی تعدادی از خوشه‌ها انتخاب می‌شوند و بقیه حذف می‌شوند. در نهایت در هر خوشه به صورت تصادفی 3 نفر انتخاب می‌شوند.
</p>

<p style="font-family:B Nazanin">
    <b>اندازه نمونه (Sample Size):</b> در ابتدای این بخش گفته شد که یکی از مسائل مهم در نمونه‌برداری، انتخاب اندازه نمونه است. هر چه تعداد اعضای نمونه بیشتر باشد، هزینه عملیاتی بیشتر است اما محاسبات دقیق‌تر انجام می‌شود اما هرچه اندازه نمونه کوچکتر شود، محاسبات ساده‌تر و با پردازش کمتری انجام می‌شود و احمال اینکه محاسبات و تعمیم آن‌ها به کل جامعه، با خطای بیشتری همراه باشد، بیشتر می‌شود. پس این مسئله بسیار مهم می‌شود که بتوانیم کوچکترین اندازه نمونه را پیدا کنیم که با استفاده از آن محاسباتمان دقت کافی را نیز داشته باشد. انتخاب اندازه نمونه تابع معیارهای متفاوتی است که تعدادی از آن‌ها به صورت زیر است:<br>
    * <b>اندازه جمعیت:</b> هر چه اندازه جمعیت بیشتر باشد، طبیعتا اندازه نمونه بزرگتری نیز لازم داریم. مثلا اگر اندازه جمعیت 80 میلیون داده است ممکن است اندازه نمونه را 1 میلیون انتخاب کنیم ولی اگر اندازه جمعیت 10 میلیون داده باشد، طبیعتا اندازه نمونه کوچکتر می‌شود. <br>
    * <b>تنوع داده‌ها:</b> هر چه تنوع داده‌ها کمتر باشد، پس داده‌ها همگن‌تر و یکنواخت‌تر هستند. بنابراین می‌توان نمونه کوچکتری انتخاب کرد. <br>
    * <b>مقدار حاشیه خطا و سطح اطمینان:</b> هر چه اندازه نمونه کوچکتر باشد، حاشیه خطا افزایش و سطح اطمینان کاهش می‌یابد. بنابراین اگر در مسئله‌ای مشخص کرده باشند که حاشیه خطا یا سطح اطمینان چقدر باشد (مثلا فلان قدر خطا مجاز است یا داده‌ها باید 99٪ سطح اطمینان داشته باشند و به واقعیت نزدیک باشند)، بر اساس آن می‌توان اندازه نمونه را مشخص کرد. <br>
    * <b>بودجه:</b> اندازه نمونه بزرگتر نیازمند صرف هزینه بیشتر است. پس اگر بودجه کافی وجود ندارد اندازه نمونه را باید کاهش داد. <br>
    * <b>تعیین هدف تحقیق:</b> گاهی اوقات هدف انتخاب‌شده برای یک پژوهش اندازه نمونه را می‌تواند تغییر دهد. برای برخی از اهداف دقت بالاتری لازم است. بنابراین اندازه نمونه بزرگتری نیاز است. <br>
    * <b>آزمون نمونه‌گیری پیشین:</b> با کمک گرفتن از نمونه‌گیری‌های قبلی می‌توان اندازه نمونه فعلی را محاسبه کرد. مثلا اگر از یک جامعه 50 میلیونی، قبلا یک نمونه یک میلیونی بررسی شده و خطاها زیاد بوده است، این بار باید تعداد اعضای نمونه بیشتر انتخاب شود. <br>
    * <b>فرمول‌های آماری:</b> با استفاده از فرمول‌های آماری، اگر بعضی از پارامترهای جامعه را داشته باشیم، می‌توانیم اندازه نمونه را تعیین کنیم. (اندازه نمونه بر اساس خطا، اندازه نمونه بر اساس انحراف معیار، اندازه خطا بر اساس انتخاب تصادفی) که در بخش‌های بعد در مورد این فرمول‌ها توضیح داده خواهد شد. 
</p>

<p style="font-family:B Nazanin">
    مثال‌هایی از روش‌های نمونه‌برداری: <br>
    * نمونه‌برداری ساده (Simple Sampling): در این مثال 200 عدد تصادفی با توزیع نرمال ایجاد می‌کنیم و سایز نمونه را 20 انتخاب می‌کنیم.
</p>

```python
import numpy as np
```

```python
population_data = np.random.normal(loc=50, scale=10, size=200)
population_data
```

```python
sample_size = 20
sample_data = np.random.choice(population_data, size=sample_size, replace=True) # با جایگذاری
# sample_data = np.random.choice(population_data, size=sample_size, replace=False) # بدون جایگذاری
sample_data
```

```python
print(f"Mean_Population = {np.mean(population_data)}")
print(f"Mean_Sample = {np.mean(sample_data)}")
```

<p style="font-family:B Nazanin">
     * نمونه‌برداری سیستماتیک (Systematic Sampling): در اینجا فرض می‌کنیم جامعه ما اعداد صحیح 1 تا 100 است و می‌خواهیم نمونه‌ای با اندازه 5 انتخاب کنیم. در ابتدا با توجه به اندازه جامعه و اندازه نمونه، گام حرکت را محاسبه می‌کنیم (در اینجا 100 تقسیم بر 5 مساوی 20) و سپس نقطه شروع را نیز به صورت تصادفی انتخاب می‌کنیم و در نهایت با این دو مقدار، نمونه‌های تصادفی را پیدا می‌کنیم.
</p> 

```python
import numpy as np
```

```python
population_data = np.arange(1, 101)
print(f"Population Data: {population_data}")
```

```python
sample_size = 5
step_size = len(population_data) // sample_size
print(f"Step size: {step_size}")
```

```python
start_index = np.random.randint(0, step_size)
print(f"Start Index: {start_index}")
```

```python
sample_data = population_data[start_index::step_size]
print(f"Sample Data: {sample_data}")
```

<p style="font-family:B Nazanin">
     * نمونه‌برداری دسته‌ای (Stratified Sampling): در اینجا فرض می‌کنیم که داده‌های جامعه به صورت زیر باشد:
</p>

```python
import numpy as np
```

```python
population_data = np.array([10,67,15,89,12,5,8,29,28,20,18,25,22,45])
print(f"Population Data: {population_data}")
```

<p style="font-family:B Nazanin">
     می‌خواهیم داده‌ها را دسته‌بندی کنیم. فرض کنیم که داده‌ها به این صورت دسته‌بندی می‌شوند که در بازه‌های صفر تا 9، 10 تا 19، 20 تا 29 و 30 به بالا قرار گیرند. با استفاده از تابع digitize در کتابخانه numpy می‌توان این کار را انجام داد.
</p>

```python
categories = np.digitize(population_data, bins=[0,10,20,30])
print(f"Categories = {categories}")
```

<p style="font-family:B Nazanin">
     خروجی این تابع یک لیست است که مشخص می‌کند هر کدام از اعضای جامعه در دسته چندم قرار می‌گیرند. ما در این مثال چهار دسته داشتیم که مثلا مشخص می‌شود داده اول که 10 بود، در دسته دوم و داده دوم که 67 بود در دسته چهارم قرار دارد و الی آخر. <br>
    فرض کنیم می‌خواهیم از هر دسته 2 داده انتخاب کنیم:
</p>

```python
sample_size_per_category = 2
sample_indices = []
for category in np.unique(categories):
    category_indices = np.where(categories == category)[0]
    random_indices = np.random.choice(category_indices, size=sample_size_per_category, replace=False)
    sample_indices.extend(random_indices)

sample_data = population_data[sample_indices]
print(f"Sample Data: {sample_data}")
```

<p style="font-family:B Nazanin">
     در مثالی دیگر فرض کنیم لیستی از شماره افراد (از 1 تا 9) به همراه سن آن‌ها داریم (لیست دو بعدی). می‌خواهیم افراد را بر اساس سن آن‌ها دسته‌بندی کنیم و از هر دسته 2 نفر را به صورت تصادفی انتخاب کنیم:
</p>

```python
import numpy as np
```

```python
population_data = np.array([
    [1,20], [2,22], [3,25],
    [4,18], [5,30], [6,28],
    [7,19], [8,35], [9,33]
])
print(f"Population Data: \n{population_data}")
```

```python
categories = np.digitize(population_data[:,1], bins=[0,20,30,40])
print(f"Categories: {categories}")
```

```python
sample_size_per_category = 2
sample_indices = []
for category in np.unique(categories):
    category_indices = np.where(categories == category)[0]
    random_indices = np.random.choice(category_indices, size=sample_size_per_category, replace=False)
    sample_indices.extend(random_indices)

sample_data = population_data[sample_indices]
print(f"Sample Data: \n{sample_data}")
```
