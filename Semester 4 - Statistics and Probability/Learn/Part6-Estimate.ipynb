<p style="font-family:B Nazanin">
    تا اینجا یاد گرفتیم که از بین داده‌های جامعه، نمونه‌های تصادفی را انتخاب کنیم و بر اساس داده‌های نمونه، آماره‌های مورد نیاز را پیدا کنیم. در مرحله بعد باید از روی مقادیر آماره‌ها، تخمین بزنیم که مقادیر پارامترهای جامعه چقدر است. هر چه تخمین ما دقیق‌تر باشد، پارامترهای به‌دست آمده به مقادیر واقعی نزدیک‌تر است. <br>
    <br>
    <b>استنتاج آماری:</b> فرایند استفاده از نمونه برای استنباط در مورد یک جامعه را استنتاج آماری می‌گویند. یعنی به کل فرایندی که نمونه را بررسی می‌کنیم تا از روی آن جامعه را استنباط کنیم، استنتاج آماری می‌گویند.<br>
    <b>تخمین یا برآورد (Estimate):</b> روشی است برای پیش‌بینی یا حدس مقدار یک ویژگی. <br>
    ما می‌دانیم که پارامتر مقدار یک ویژگی از روی داده‌های جامعه و آماره مقدار همان ویژگی از روی داده‌های نمونه است. رابطه آماره و پارامتر در زمان نمونه‌برداری شکل می‌گیرد. به این معنی که روش نمونه‌برداری روی ارتباط آماره و پارامتر تاثیر می‌گذارد. بنابراین به هنگام تخمین پارامتر از روی آماره باید به روش نمونه‌برداری توجه داشته باشیم. مثلا اگر از بین کل دانشجویان یک دانشگاه، فقط دانشجویان برتر را به عنوان نمونه انتخاب کردیم و میانگین معدل آن‌ها را محاسبه نمودیم، باید توجه داشته باشیم که این معدل نمی‌تواند به کل دانشجویان دانشگاه تعمیم داده شود و قطعا معدل کل دانشگاه کمتر از این مقدار است و این معدل فقط می‌تواند به دانشجویان برتر تعمیم داده شود. <br>
    در اینجا گاهی کلمه اندازه‌گیری مطرح می‌شود که باید توجه داشته باشیم که این مفهوم با تخمین متفاوت است. تخمین یعنی از روی مقدار آماره و با استفاده از حدس و پیش‌بینی مقدار یک پارامتر را محاسبه کنیم که همیشه با مقداری خطا همراه است اما اندازه‌گیری یعنی مستقیما مقدار آن پارامتر را از روی داده‌های جامعه محاسبه کنیم کع مقدار دقیقی را به ما می‌دهد. <br>
    <b>نسبت (Proportion):</b> نسبت تعداد داده‌های یک نمونه به تعداد کل داده‌های جامعه را می‌گویند. مثلا اگر یک جامعه شامل 100 نفر باشد که 20 نفر آن‌ها جوان باشند، می‌گوییم نسبت افراد جوان در این جامعه 20درصد یا یک پنجم است.<br>
    در جدول زیر نماد پارامترها و آماره‌های مربوط به آن را در جامعه و نمونه مشاهده می‌کنید.
</p>

<p style="font-family:B Nazanin">
    <b>تخمین‌گر (Estimator):</b> هر آماره‌ای ($\hat{\theta}$) که از آن برای تخمین یک پارامتر ($\theta$) استفاده کنیم، به آن تخمین‌گر می‌گویند. یک تخمین‌گر خوب باید دارای ویژگی‌های زیر باشد: <br>
    * نااریب باشد (Unbiased): یعنی میانگین یک آماره برای نمونه‌های مختلف برابر باشد با مقدار پارامتر مورد نظر برای جامعه. مثلا اگر چندین نمونه مختلف از جامعه انتخاب کردیم و برای هر کدام میانگین ($\bar{x}$) حساب کردیم و سپس میانگین این میانگین‌ها را محاسبه کردیم و این میانگین برابر با میانگین کل جامعه بود، به این آماره، یک آماره نااریب (Unbiased) می‌گویند. به عبارت دیگر امید ریاضی آماره برای نمونه‌های مختلف برابر با پارامتر شود. مثلا در مورد میانگین: $E(\bar{x})=\mu$ و یا در مورد واریانس: $E(s^2)=\sigma^2$. به طور کلی: $E(\hat{\theta})=\theta$.<br>
    * کارایی (Efficiency): در بین تخمین‌گرهایی که نااریب باشند، تخمین‌گری کارایی بیشتر و بالاتری دارد که واریانس کمتری داشته باشد. <br>
    * سازگاری (Consistency): تخمین‌گری سازگاری بالاتری دارد که با افزایش n، مقدار آماره به مقدار پارامتر نزدیک‌تر شود یعنی تخمین دقیق‌تری به ما بدهد. <br>
    * حداقل میانگین مجذور خطا (MSE): بر اساس فرمول زیر هر چه MSE کمتر باشد، تخمین‌گر ما دقیق‌تر است:<br>
    $$
    MSE = s^2+(E(\hat{\theta})-\theta)^2
    $$
</p>

<p style="font-family:B Nazanin">
    <b>توزیع نمونه‌ای (Sample Distribution):</b> اگر بیاییم توزیع مقادیر یک آماره خاص را در نمونه‌های مختلف یک جامعه به‌دست آوریم، توزیع نمونه‌ای آن آماره خاص را پیدا کرده‌ایم. مثلا فرض کنید یک جامعه یک میلیون داده دارد. اگر بیاییم و مثلا 30 نمونه مختلف هر کدام به تعداد 2000 داده را در نظر بگیریم و برای هر کدام میانگین نمونه را محاسبه کنیم و توزیع این میانگین‌ها را به‌دست آوریم، به این توزیع، توزیع نمونه‌ای میانگین می‌گویند. مثال:
</p>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
population_mean = 170
population_std = 5
population_size = 10000
population = np.random.normal(population_mean, population_std, population_size)
population
```

<p style="font-family:B Nazanin">
    می‌خواهیم از این جامعه 10000 عضوی، 500 نمونه 30 عضوی پیدا کنیم و برای هر نمونه، میانگین داده‌ها را بیابیم:
</p>

```python
num_samples = 500
sample_size = 30
sample_means = []
for _ in range(num_samples):
    sample = np.random.choice(population, size=sample_size, replace=False)
    sample_mean = np.mean(sample)
    sample_means.append(sample_mean)
    
sample_means
```

<p style="font-family:B Nazanin">
    میانگینِ میانگین‌های نمونه‌ها (میانگین توزیع میانگین‌ها) برابر است با:
</p>

```python
estimated_mean = np.mean(sample_means)
estimated_mean
```

<p style="font-family:B Nazanin">
    انحراف معیار میانگین‌ها برابر است با:
</p>

```python
estimated_std = np.std(sample_means)
estimated_std
```

<p style="font-family:B Nazanin">
    اکنون با داشتن این میانگین و انحراف معیار، می‌توانیم نمودار توزیع احتمال این میانگین‌ها را به صورت هیستوگرام رسم کنیم. ضمنا یک توزیع نرمال از داده‌های در بازه میانگین به‌اضافه و منهای 3 برابر انحراف معیار رسم می‌کنیم تا نشان دهیم توزیع داده‌ها چقدر به توزیع نرمال نزدیک است. چون می‌دانیم که در توزیع نرمال 99 درصد داده‌ها در این بازه قرار می‌گیرند.
</p>

```python
x = np.linspace(estimated_mean - 3 * estimated_std, estimated_mean + 3 * estimated_std, 1000)
x
```

```python
y = norm.pdf(x, estimated_mean, estimated_std)
y
```

```python
plt.figure(figsize=(10,6))
plt.hist(sample_means, bins=30, color='green', density=True, rwidth=0.9)
plt.plot(x, y, color='red', linewidth=2)
plt.show()
```

<p style="font-family:B Nazanin">
    <b>درجه آزادی (Degree of Freedom یا DF):</b> میزان آزادی ما برای انتخاب مقادیر در فرایند تخمین را درجه آزادی می‌گویند. به عنوان مثال فرض کنید که اگر در یک جامعه، میانگین سن افراد 35 سال است و می‌خواهیم سن 6 نفر را مثال بزنیم که میانگین آن‌ها همین 35 سال شود. به طور مثال می‌توانیم 5 عدد را به طور دلخواه کاملا آزادانه انتخاب کنیم و عدد ششم را طوری انتخاب کنیم که میانگین آن‌ها 35 شود. مثلا پنج عدد 17 و 43 و 92 و 6 و 12 را در نظر بگیریم. عدد ششم x به این صورت محاسبه می‌شود:
</p>

$
\frac{x+17+43+92+6+12}{6}=35 \quad \rightarrow \quad x + 170 = 210 \quad \rightarrow \quad \boxed{x = 40}
$

<p style="font-family:B Nazanin">
    بنابراین در این مثال می‌گوییم اگر تعداد داده‌های خواسته‌شده n باشد، درجه آزادی ما $n-1$ است ($DF = n-1$). در ادامه از این درجه آزادی برای محاسبه انحراف معیار استفاده می‌کنیم.
</p>

<p style="font-family:B Nazanin">
    <b>انواع تخمین‌گرها:</b><br>
    * <b>تخمین‌گر نقطه‌ای (Point Estimator):</b> در این نوع تخمین‌گر برای هر پارامتر یک مقدار واحد به دست می‌آید. مثلا یک نمونه از جامعه انتخاب می‌کنیم و میانگین آن را محاسبه می‌کنیم و می‌گوییم میانگین جامعه این عدد است. این روش می‌تواند خطای زیادی داشته باشد. <br>
    * <b>تخمین‌گر بازه‌ای (Interval Estimator):</b> در این نوع تخمین‌گر یک بازه برای هر پارامتر مشخص می‌کنیم. به اصطلاح دیگر برای هر پارامتر یک حد پایین (Lower Limit) و یک حد بالا (Upper Limit) مشخص می‌کنیم و می‌گوییم که این پارامتر در این فاصله اطمینان (Confidence Interval) قرار دارد. همین فاصله اطمینان و حد بالا و پایین نیز بر اساس تخمین نقطه‌ای ممکن است به‌دست آید. 
</p>

```python
import numpy as np
from scipy.stats import norm
```

```python
np.random.seed(0)
population_data = np.random.normal(loc=5, scale=2, size=100)
population_mean = np.mean(population_data)
population_std = np.std(population_data)

print(f"population_data: {population_data}")
print(f"population_mean: {population_mean}")
print(f"population_std: {population_std}")
```

```python
sample_size = 30
sample = np.random.choice(population_data, size=sample_size, replace=False)
sample_mean = np.mean(sample)
sample_std = np.std(sample)

print(f"sample: {sample}")
print(f"sample_mean: {sample_mean}")
print(f"sample_std: {sample_std}")
```

<p style="font-family:B Nazanin">
    فاصله اطمینان، بازه‌ای است که مشخص می‌کند مقدار پارامتر در آن بازه قرار می‌گیرد و سطح اطمینان مشخص می‌کند مقدار پارامتر با چه احتمالی در این بازه قرار می‌گیرد. مثلا اگر سطح اطمینان 95٪ باشد یعنی 95٪ مطمئن هستیم که مقدار ی پارامتر در بازه مشخص شده‌ی فاصله اطمینان قرار دارد. به عنوان مثال می‌گوییم به احتمال 95٪ میانگین قد افراد بین 160 تا 180 سانتیمتر است. یعنی به احتمال 5٪ خارج از این بازه (2.5٪ کمتر از 160 و 2.5٪ بیشتر از 180) است. حال اگر بخواهیم سطح اطمینان را بالاتر ببریم،‌طبیعتا باید فاصلخ اطمینان را بزرگتر کنیم. مثلا بگوییم به احتمال 99٪ میانگین قد افراد در بازه 150 تا 195 سانتیمتر است. سطح اطمینان معمولا 90٪، 95٪ یا 99٪ انتخاب می‌شود.
</p>

<p style="font-family:B Nazanin">
    برای محاسبه فاصله اطمینان (CI) از فرمول زیر استفاده می‌کنیم که در آن z همان z_score است:
</p>

$$
CI = \bar{x} \pm z\frac{s}{\sqrt{n}}
$$

<p style="font-family:B Nazanin">
    در مثال قبل سطح اطمینان را 95٪ انتخاب می‌کنیم. کلاس scipy.stats.norm تابعی به نام ppf دارد که اگر $\frac{1+confidence\_level}{2}$ را به آن بدهیم، z_score را به ما می‌دهد:
</p>

```python
confidence_level = 0.95
print((1+confidence_level)/2)
z_score = norm.ppf((1+confidence_level)/2)
margin_of_error = z_score * sample_std / np.sqrt(sample_size)
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)
print(confidence_interval)
```

<p style="font-family:B Nazanin">
    حاشیه خطای مجاز (Margin Of Error): $$Margin \quad Of \quad Error = z \times \frac{s}{\sqrt{n}}$$
    خطای استاندارد میانگین (Standard Error of Mean): $$SE_{\bar{x}} = \frac{s}{\sqrt{n}}$$
</p>

```python
import numpy as np
np.random.seed(42)
sample = np.random.normal(loc=50, scale=10, size=100)
standard_error = np.std(sample, ddof=1) / np.sqrt(len(sample))
print(f"Sample: \n{sample}")
print(f"SE: {standard_error}")
```

<p style="font-family:B Nazanin">
    تابع crosstab در ماژول pandas می‌تواند به ما کمک کند تا جداول کمکی ارزشمندی تولید کنیم. این تابع از یک دیتافریم کمک می‌گیرد و داده‌های ستون‌های مختلف آن را می‌تواند به عنوان سطر و ستون جدول جدید قرار دهد و مقادیر جدیدی را محاسبه کند. این تابع چند پارامتر ورودی دارد: پارامتر index مشخص می‌کند کدام ستون دیتافریم به عنوان سطرهای جدول جدید قرار گیرد. پارامتر columns مشخص می‌کند کدام ستون دیتافریم به عنوان ستون جدول جدید قرار گیرد. پارامتر values مشخص می‌کند برای سطر و ستون انتخاب‌شده مقادیر جدول بر اساس کدام ستون دیتافریم محاسبه شود. پارامتر aggfunc مشخص می‌کند که چه تابعی روی این مقادیر اعمال شود تا محاسبه گردند. در مثال زیر از دیتافریم data کمک می‌گیریم تا تعداد فروش هر دسته از محصولات را در کشورهای مختلف حساب کنیم. نام کشورها به عنوان سطر، دسته‌بندی‌ها به عنوان ستون، مقادیر فروش به عنوان مقادیر جدول استفاده می‌شوند و تابع count روی آن‌ها اعمال می‌شود تا تعداد آن‌ها محاسبه شود.<br>
    ضمنا پارامتر margins اگر True باشد، جمع کل سطرها و ستون‌ها را نیز اضافه می‌کند و پارامتر margins_name نامی برای این سطر و ستون جمع، انتخاب می‌کند.
</p>

```python
pd.crosstab(
    index=data['Country'], 
    columns=data['Category'], 
    values=data['Sales'], 
    aggfunc='count', 
    margins=True, 
    margins_name='Total Count'
)
```
