    <h1>فصل چهارم: آمار و احتمالات</h1>
    <h2>بخش چهارم: آمار استنباطی و مفاهیم پایه‌ای</h2>
</span>

<p style="font-family:B Nazanin">
    علم آمار به دو دسته آمار توصیفی و آمار استنباطی تقسیم‌بندی می‌شود. آمار توصیفی همان کاری است که یک داده را در نظر می‌گرفتیم و میانگین، میانه، انحراف معیار و ... را مربوط به آن داده محاسبه می‌کنیم. اما آمار استنباطی به کمک آمار توصیفی یک سری اطلاعات را از نمونه‌ای از کل جامعه به‌دست می‌آورد. در حقیقت در آمار استنباطی به جای اینکه کل جامعه (کل داده‌ها) را بررسی کنیم، یک بخشی از جامعه را به عنوان نمونه بررسی می‌کنیم و اطلاعات مورد نیاز را در مورد آن به‌دست می‌آوریم و اطلاعات به‌دست آمده را به کل جامعه تعمیم می‌دهیم. البته این نکته قابل توجه است که در این روش همیشه عدم قطعیت وجود دارد. چرا که ممکن است نمونه انتخاب‌شده با داده‌های کل جامعه تفاوت‌هایی داشته باشد. بنابراین اطلاعات به‌دست آمده از نمونه، ممکن است با کل جامعه اختلاف‌هایی داشته باشد. <br>
    <b>مفاهیم پایه آمار استنباطی:</b><br>
    * <b>جامعه (Population):</b> به مجموعه‌ای از تمام افراد، اشیا یا رویدادهایی است که دارای ویژگی‌های مشترک است. مثلا جامعه دانشجویان کامپیوتر، جامعه ماشین‌های پراید استان همدان.<br>
    * <b>نمونه (Sample):</b> گروه کوچکی از جامعه است که بر اساس قوانین و ضوابطی از کل جامعه انتخاب شده و آن را بررسی می‌کنیم تا با تجزیه و تحلیل آن به تحلیل جامعه برسیم. <br>
    N: تعداد اعضای جامعه (P) <br>
    n: تعداد اعضای نمونه (S) <br>
    * <b>پارامتر (Parameter):</b> به هر شاخص آماری که از روی جامعه گرفته شده باشد، گفته می‌شود. مثل میانگین کل داده‌های جامعه.<br>
    * <b>آماره (Statistic):</b> به هر شاخص آماری که از روی نمونه گرفته شده باشد، گفته می‌شود. مثل میانگین داده‌های نمونه. <br>
    پارامترها را معمولا با $\theta$ و آماره‌ها را با $\hat{\theta}$ نشان می‌دهند. 
</p>

<p style="font-family:B Nazanin">
    <b>مراجل انجام فرایند آمار استنباطی:</b><br>
    1. شناسایی دقیق جامعه آماری <br>
    2. انتخاب نمونه مناسب از جامعه آماری <br>
    3. تجزیه و تحلیل داده‌های نمونه <br>
    4. استنباط (تخمین) پارامترهای جامعه از روی آماره‌های نمونه <br>
    5. بررسی میزان قابل اطمینان بودن نتایج استنباط‌شده<br>
    <br>
    مثال. اعداد 2 و 4 و 6 و 8 و 10 را به عنوان جامعه در نظر می‌گیریم:
</p>

$$
P: 2, 4, 6, 8, 10\\
\mu = \frac{\Sigma{x_i}}{N} = \frac{2+4+6+8+10}{5} = \frac{30}{5} = 6 \\
\sigma^2 = \frac{\Sigma (x_i-\mu)^2}{N} = \frac{(2-6)^2+(4-6)^2+(6-6)^2+(8-6)^2+(10-6)^2}{5}=\frac{16+4+0+4+16}{5}=8\\
\sigma = \sqrt{8}
$$

<p style="font-family:B Nazanin">
    اگر بخواهیم از این جامعه نمونه‌های 3تایی در نظر بگیریم، 10 حالت انتخاب داریم (انتخاب 3 عضو از مجموعه 5 عضوی). همانطور که در ادامه می‌بینیم هر کدام از نمونه‌ها میانگین داده‌های متفاوتی دارند. پس انتخاب نمونه بسیار مهم است تا نزدیک‌ترین تخمین به پارامترهای جامعه را داشته باشیم. 
</p>
