<p style = "font-family:B Nazanin;">
    آمار، علم جمع‌آوری، تفسیر، تحلیل و نمایش داده‌هاست. اگر داده‌ها به‌درستی نشان داده‌شوند، می‌توان آن را بهتر درک کرد و تصمیم‌گیری درستی انجام داد. <br>
    اولین و ساده‌ترین هدف آمار این است که اطلاعات کاربردی از داده‌ها به ما بدهد. آمار می‌تواند داده‌ها را توصیف کند و یک سری ویژگی از داده‌ها را برای ما مشخص کند. آمار می‌تواند به ما کمک کند تا روابط بین متغیرها را به‌دست بیاوریم. <br>
    احتمالات علمی است که تلاش می‌کند تا احتمال وقوع فرایندهای تصادفی را به‌دست آورد. پس کار احتمالات این است که به کمک مدل‌های احتمالی بر اساس رویدادهای پیش‌آمده، رویدادهای آینده را پیش‌بینی کند که با این کار می‌تواند میزان ریسک را پایین بیاورد. <br>
    اهمیت آمار و احتمالات در علم داده: <br>
    &nbsp;&nbsp;&nbsp;* تحلیل داده: به کمک تکنیک‌هایی که وجود دارد، می‌توانیم داده‌ها را تحلیل کنیم و روابط بین متغیرها را مشخص کنیم. <br>
    &nbsp;&nbsp;&nbsp;* پیش‌بینی و تصمیم‌گیری: با بررسی اتفاقاتی که در گذشته رخ داده می‌توانیم پیش‌بینی کنیم که چه اتفاقاتی در آینده رخ می‌دهد و بر اساس آن تصمیم‌گیری کنیم. <br>
    &nbsp;&nbsp;&nbsp;* اعتبارسنجی: بر اساس داده‌های گذشته می‌توانیم داده‌هایی که جدید به‌دست می‌آوریم را بررسی کرده و اعتبار آن را بسنجیم. <br>
    &nbsp;&nbsp;&nbsp;* مدلسازی: بر اساس داده‌هایی که تحلیل کرده‌ایم، مدل‌های مناسب را می‌توانیم بسازیم و بر اساس آن پیش‌بینی‌های لازم را انجام دهیم. <br>
    &nbsp;&nbsp;&nbsp;* مدیریت ریسک: با بررسی داده‌ها احتمال وقوع هر رویدادی را می‌توان محاسبه کرد و احتمال وقوع خطا را کاهش داد تا ریسک تصمیمات پایین بیاید. <br>
    دو مرحله اساسی در علم آمار و احتمالات، نمونه‌برداری و تجزیه و تحلیل داده‌هاست. نمونه‌برداری یعنی انتخاب تصادفی یک زیرمجموعه کوچک‌تر از داده‌هاست. کار کردن با کل یک مجموعه داده هم زمان‌بر است و هم تفاوتی با کار کردن روی یک جامعه آماری نمونه ندارد. به طوری که این جامعه آماری نمونه مناسبی از کل داده‌ها باشد. تجزیه و تحلیل داده‌ها نیز با استفاده از ابزارهای مختلفی مثل محسابه میانگین و میانه و واریانس و رگرسیون و ... انجام می‌شود. <br>
    <b>تئوری احتمال:</b> مجموعه ابزار و تکنیک‌هایی است که به ما کمک می‌دهد تا احتمال وقوع یک رویداد را تحلیل و اندازه‌گیری کنیم. اصول اصلی تئوری احتمال عبارتند از: <br>
    &nbsp;&nbsp;&nbsp;* <b>فضای نمونه (Sample Space):</b> مجموعه‌ای از تمام نتایج ممکن یک آزمایش تصادفی است. مثلا در آزمایش انداختن یک سکه، فضای نمونه آمدن شیر یا خط است. یا در تاس ریختن، فضای نمونه اعداد 1و2و3و4و5و6 است. <br>
    &nbsp;&nbsp;&nbsp;<b>رویداد یا پیشامد (Event):</b> زیرمجموعه‌ای از فضای نمونه است که می‌توانند شامل یک یا چند نتیجه باشند. مثلا پیشامد وقوع شیر، پیشامد وقوع خط، پیشامد وقوع عدد ۵ در پرتاب تاس، پیشامد وقوع عدد زوج در پرتاب تاس. <br>
    &nbsp;&nbsp;&nbsp;<b>احتمال (Probability):</b> عددی بین صفر و یک است که احتمال وقوع یک رویداد را مشخص می‌کند. اگر احتمال یک رویداد صفر باشد، یعنی آن رویداد اصلا اتفاق نمی‌افتد. اگر احتمال یک باشد، یعنی قطعا آن اتفاق رخ خواهد داد. <br>
    <br>

    &nbsp;&nbsp;&nbsp;* Pandas: کتابخانه محبوب برای تجزیه و تحلیل داده‌ها، کار با داده‌های جدولی، csv و ...<br>
    &nbsp;&nbsp;&nbsp;* Matplotlib و Seaborn: برای رسم نمودارها<br>
    &nbsp;&nbsp;&nbsp;* Scipy: دارای توابعی برای مسائل علمی و مهندسی و آمار و احتمالات <br>
    &nbsp;&nbsp;&nbsp;* Statemodels: برای تجزیه و تحلیل‌های آماری و مدل‌سازی‌های آماری<br>
    &nbsp;&nbsp;&nbsp;* Scikit-learn: برای مدل‌سازی‌های یادگیری ماشین و الگوریتم‌های آن
</p>

```python
import numpy as np
data = np.array([10,15,20,25,30])
mean = np.mean(data)
variance = np.var(data)
print("Mean:", mean)
print("Variance:", variance)
```

```python
import numpy as np
mean = 0
std_dev = 1
sample_size = 100
data_normal = np.random.normal(mean, std_dev, sample_size)
print(data_normal)
```

```python
import pandas as pd
data = {'Name': ['mehdi', 'ali', 'reza', 'ahmad'], 'Age': [25,30,22,28]}
df = pd.DataFrame(data)
print(df)
```

```python
file_path = './data/sales.csv'
df = pd.read_csv(file_path)

mean_sales = df.groupby('Month')['Sales'].mean()
print(mean_sales)

max_sales = df.groupby('Month')['Sales'].max()
print(max_sales)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset('iris')
sns.set(style="whitegrid")
plt.figure(figsize=(8,6))
sns.scatterplot(x='sepal_length', y='sepal_width', data=data, hue='species', palette='Set1', s=100)
plt.show()
```

```python
from scipy.stats import norm

p_value = norm.cdf(1.96) - norm.cdf(-1.96)
print("p-value:", p_value)
```

```python
import statsmodels.api as sm
import numpy as np

x = np.array([1,2,3,4,5])
y = np.array([2,3.5,4.5,5.5,7])

x = sm.add_constant(x)
model = sm.OLS(y, x).fit()

print(model.summary())
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

x = np.array([1,2,3,4,5]).reshape(-1,1)
y = np.array([2,3.5,4.5,5.5,7])

model = LinearRegression()

model.fit(x,y)

print('Intercept:', model.intercept_)
print('Coefficient:', model.coef_)
```
