<p style = "font-family:B Nazanin;">
    آمار توصیفی مجموعه‌ای از تکنیک‌ها و روش‌ها است که برای توصیف، خلاصه‌سازی و تفسیر داده‌ها از آن‌ها استفاده می‌کنیم. <br>
    مراحل اصلی از تحلیل آمار توصیفی عبارتند از: <br>
    &nbsp;&nbsp;&nbsp;* جمع‌آوری داده<br>
    &nbsp;&nbsp;&nbsp;* مشخص کردن متغیرها: متغیرها و ویژگی‌هایی که قصد داریم آن‌ها را تحلیل کنیم را مشخص می‌کنیم. <br>
    &nbsp;&nbsp;&nbsp;* محاسبه معیارهای آماری: یکی از طولانی‌ترین و مهم‌ترین مراحل آماری. مانند محاسبه میانگین، مد، میانه،‌واریانس و ...<br>
    &nbsp;&nbsp;&nbsp;* تصویرسازی داده‌ها (نمودارهای توصیفی)<br>
    &nbsp;&nbsp;&nbsp;* تفسیر و نتیجه‌گیری<br>
    <br>
    معیارهای آماری (مفاهیم اصلی در آمار): <br>
    &nbsp;&nbsp;&nbsp;* میانگین (Mean): حاصل جمع داده‌ها تقسیم بر تعداد آن‌ها. 
</p>

$$9,\ 9,\ 9,\ 10,\ 10,\ 14,\ 15,\ 16,\ 19,\ 20$$

$$Mean = \frac{9+9+9+10+10+14+15+16+19+20}{10} = 13.1$$

<p style = "font-family:B Nazanin;">
    &nbsp;&nbsp;&nbsp;* مد (Mode): داده‌ای که بیشترین تکرار را در میان همه داده‌ها دارد. 
</p>

$$Mode = 9$$

<p style = "font-family:B Nazanin;">
     &nbsp;&nbsp;&nbsp;* میانه (Median): اگر داده‌ها را به صورت مرتب‌شده بنویسیم، عنصر وسط میانه است. اگر تعداد داده‌ها زوج بود، میانگین دو عدد وسط میانه است.
</p>

$$Median = \frac{10+14}{2} = 12$$

<p style = "font-family:B Nazanin;">
     &nbsp;&nbsp;&nbsp;* بازه یا محدوده (Range): تفاضل بزرگترین داده از کوچکترین داده برابر Range است.
</p>

$$Range = 20 - 9 = 11$$

<p style = "font-family:B Nazanin;">
     &nbsp;&nbsp;&nbsp;* واریانس (Variance): هر یک از داده‌ها از میانگین فاصله‌ای دارد. واریانس معرف پراکندگی داده‌ها نسبت به مقدار میانگین است که از فرمول‌های زیر محاسبه می‌شود. در نظر داشته باشید که $\mu$ میانگین کل جامعه آماری و $\bar{x}$ میانگین داده‌های فضای نمونه است.
</p>

Population Variance:
$$\sigma^2 = \frac{\sum\limits_{i = 1}^{N}(x_i-\mu)^2}{N}$$
Sample Variance:
$$S^2 = \frac{\sum\limits_{i = 1}^{n}(x_i-\bar{x})^2}{n - 1}$$

$$\sigma^2 = \frac{1}{10}\times\left[(9-13.1)^2+(9-13.1)^2+(9-13.1)^2+(10-13.1)^2+(10-13.1)^2+(14-13.1)^2+(15-13.1)^2+(16-13.1)^2+(19-13.1)^2+(20-13.1)^2\right] = 16.49$$

<p style = "font-family:B Nazanin;">
    &nbsp;&nbsp;&nbsp;* انحراف معیار (Standard Deviation): انحراف معیار برابر با جذر واریانس است. 
</p>

Population Standard Deviation:
$$\sigma = \sqrt{\frac{\sum\limits_{i = 1}^{N}(x_i-\mu)^2}{N}}$$
Sample Standard Deviation:
$$S = \sqrt{\frac{\sum\limits_{i = 1}^{n}(x_i-\bar{x})^2}{n - 1}}$$

$$\sigma = \sqrt{16.49} = 4.06$$

<p style = "font-family:B Nazanin;">
     یکی از کاربردهای انحراف معیار و واریانس مشخص‌کردن محدوده منطقی از داده‌هاست. داده‌های ما ممکن است پراکندگی زیادی داشته باشند. اما داده‌هایی که بین دو مقدار «میانگین به اضافه انحراف معیار» و «میانگین منهای انحراف معیار» هستند، داده‌های منطقی‌تری هستند و سایر داده‌ها نسبتا داده‌های پرت می‌باشند. مثلا اگر میانگین درآمد افراد یک جامعه 20 میلیون تومان و انحراف معیار آن 15 میلیون تومان باشد، محدوده منطقی درآمد افراد جامعه بین 5 تا 35 میلیون تومان است و اگر فردی درآمدش 100 میلیون تومان باشد، درآمد این فرد را در تجزیه و تحلیل دخیل نمی‌کنیم چون فاصله زیادی از محدوده منطقی داده‌ها یعنی درآمد عموم مردم جامعه دارد و تجزیه و تحلیل را دچار خطا می‌کند.
</p>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
    'Name': ['akbar', 'ashkan', 'ahmad', 'ali', 'mohamad', 'kamran', 'mehdi', 'reza'],
    'Height': [196,159,163,165,173,178,163,192]}
)

mean = df['Height'].mean()
std_dev = df['Height'].std()

plt.bar(df['Name'], df['Height'])
plt.ylim([140, 200])
plt.axhline(mean, linewidth = 3, color = 'red')
plt.axhline(mean + std_dev, linewidth = 3, color = 'green')
plt.axhline(mean - std_dev, linewidth = 3, color = 'green')

plt.show()
```

<p style = "font-family:B Nazanin;">
     در مثال بالا همانطور که در نمودار مشاهده می‌شود، اعداد قد 196، 159 و 192 که در محدوده سبزرنگ نیستند، اعداد منطقی نیستند. <br>
    &nbsp;&nbsp;&nbsp;* صدک‌ها (Percentiles): صدک‌ها نشان‌دهنده مقادیری از داده‌ها هستند که درصد مشخصی از داده‌ها کمتر از آن مقدار است. مثلا در مثال بالا (قد افراد)، صدک 20٪ داده‌ها برابر 163 است. یعنی 20 درصد از داده‌ها از عدد 163 کوچکتر است. 
</p>

```python
print(np.percentile(df['Height'], 20))
```

<p style = "font-family:B Nazanin;">
     &nbsp;&nbsp;&nbsp;* چارک‌ها (Quartiles): نوع خاصی از صدک‌ها هستند که برای درصدهای 25 و 50 و 75 درصد محاسبه می‌شوند. پس چارک اول (Q1) برابر با صدک 25٪، چارک دوم (Q2) برابر با صدک 50٪ و چارک سوم (Q3) برابر با صدک 75٪ هستند.  
</p>

```python
print(np.percentile(df['Height'], 25))
print(np.percentile(df['Height'], 50))
print(np.percentile(df['Height'], 75))
```

</p>

<p style = "font-family:B Nazanin;">
     همانطور که مشاهده می‌شود اگر تمام داده‌ها با هم برابر باشد، انحراف معیار و واریانس صفر است که نشان می‌دهد داده‌ها هیچ پراکندگی ندارند و هرچه پراکندگی داده‌ها بیشتر باشد، احراف معیار و واریانس نیز بیشتر می‌شود. حتی اگر میانگین داده‌ها تغییری نکند. <br>
    مثال دوم:
</p>

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

data = [1, 2, 5, 5, 7, 2, 3, 4, 2, 4, 1, 9]
sorted_data = sorted(data)

mean_value = np.mean(data)
median_value = np.median(data)
mode_value = stats.mode(data).mode
variance_value = np.var(data)
std_deviation_value = np.std(data)
data_range = np.ptp(data)
percentiles = np.percentile(data, [10, 40, 70, 90])
quartiles = np.percentile(data, [25, 50, 75])
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
iqr = q3 - q1

print("sorted_data: \t\t", sorted_data)
print("mean_value: \t\t", mean_value)
print("median_value: \t\t", median_value)
print("mode_value: \t\t", mode_value)
print("variance_value: \t", variance_value)
print("std_deviation_value: \t", std_deviation_value)
print("data_range: \t\t", data_range)
print("percentiles: \t\t", percentiles)
print("quartiles: \t\t", quartiles)
print("IQR: \t\t\t", iqr)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

data = [1, 2, 5, 5, 7, 2, 3, 4, 2, 4, 1, 9]
x_values = np.arange(len(data))
plt.figure(figsize = (13, 6))
plt.plot(x_values, data, marker = 'o', linestyle = '-', color = 'blue', label = 'Data')

# Mean
mean_value = np.mean(data)
plt.axhline(mean_value, color = 'red', linestyle = 'dashed', linewidth = 1, label = 'Mean')
plt.text(x_values[-1]+1, mean_value, f'Mean: {mean_value:0.2f}', color = 'red', verticalalignment = 'center')

# Median
median_value = np.median(data)
plt.axhline(median_value, color = 'green', linestyle = 'dashed', linewidth = 1, label = 'Mean')
plt.text(x_values[-1]+1, median_value, f'Median: {median_value:0.2f}', color = 'green', verticalalignment = 'center')

# Mode
mode_value = stats.mode(data).mode
plt.axhline(mode_value, color = 'purple', linestyle = 'dashed', linewidth = 1, label = 'Mean')
plt.text(x_values[-1]+1, mode_value, f'Mode: {mode_value:0.2f}', color = 'purple', verticalalignment = 'center')

# Variance
variance_value = np.var(data)
plt.axhline(variance_value, color = 'orange', linestyle = 'dashed', linewidth = 1, label = 'Mean')
plt.text(x_values[-1]+1, variance_value, f'Variance: {variance_value:0.2f}', color = 'orange', verticalalignment = 'center')

# Standard Deviation
std_deviation_value = np.std(data)
plt.axhline(std_deviation_value, color = 'brown', linestyle = 'dashed', linewidth = 1, label = 'Mean')
plt.text(x_values[-1]+1, std_deviation_value, f'Standard Deviation: {std_deviation_value:0.2f}', color = 'brown', verticalalignment = 'center')

# Range
data_range = np.ptp(data)
plt.axhline(data_range, color = 'black', linestyle = 'dashed', linewidth = 1, label = 'Mean')
plt.text(x_values[-1]+1, data_range, f'Data Range: {data_range:0.2f}', color = 'black', verticalalignment = 'center')

#Percentiles
percentiles = np.percentile(data, [10, 40, 70, 90])
plt.axhline(percentiles[0], color = 'magenta', linestyle = 'dashed', linewidth = 1, label = 'Percentiles[0]')
plt.axhline(percentiles[1], color = 'cyan', linestyle = 'dashed', linewidth = 1, label = 'Percentiles[1]')
plt.axhline(percentiles[2], color = 'blue', linestyle = 'dashed', linewidth = 1, label = 'Percentiles[2]')
plt.axhline(percentiles[3], color = 'green', linestyle = 'dashed', linewidth = 1, label = 'Percentiles[3]')
plt.text(x_values[-1]+1, percentiles[0], f'10th Percentile: {percentiles[0]:0.2f}', color = 'magenta', verticalalignment = 'center')
plt.text(x_values[-1]+1, percentiles[1], f'40th Percentile: {percentiles[1]:0.2f}', color = 'cyan', verticalalignment = 'center')
plt.text(x_values[-1]+1, percentiles[2], f'70th Percentile: {percentiles[2]:0.2f}', color = 'blue', verticalalignment = 'center')
plt.text(x_values[-1]+1, percentiles[3], f'90th Percentile: {percentiles[3]:0.2f}', color = 'green', verticalalignment = 'center')

# Quartiles
quartiles = np.percentile(data, [25, 50, 75])
plt.axhline(quartiles[0], color = 'purple', linestyle = 'dashed', linewidth = 1, label = 'Percentiles[0]')
plt.axhline(quartiles[1], color = 'blue', linestyle = 'dashed', linewidth = 1, label = 'Percentiles[1]')
plt.axhline(quartiles[2], color = 'green', linestyle = 'dashed', linewidth = 1, label = 'Percentiles[2]')
plt.text(x_values[-1]+1, quartiles[0], f'Q1 (25th Quartile): {quartiles[0]:0.2f}', color = 'purple', verticalalignment = 'center')
plt.text(x_values[-1]+1, quartiles[1], f'Q2 (50th Quartile): {quartiles[1]:0.2f}', color = 'blue', verticalalignment = 'center')
plt.text(x_values[-1]+1, quartiles[2], f'Q3 (75th Quartile): {quartiles[2]:0.2f}', color = 'green', verticalalignment = 'center')

# IQR
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
iqr = q3 - q1
plt.axhline(iqr, color = 'pink', linestyle = 'dashed', linewidth = 1, label = 'IQR')
plt.text(x_values[-1]+1, iqr, f'IQR: {iqr:0.2f}', color = 'pink', verticalalignment = 'center')

plt.xticks(x_values)
plt.xlabel('Index')
plt.ylabel('Values')
plt.title('Line Plot of Data')
plt.legend()
plt.tight_layout()
plt.show()
```

<p style = "font-family:B Nazanin;">

<p style = "font-family:B Nazanin;">
    <b>نمودار نقطه‌ای (Scatter Plot):</b> در این نمودار مقادیر متغیر به صورت نقطه‌هایی روی نمودار مشخص می‌شود. 
</p>

```python
import matplotlib.pyplot as plt
import numpy as np

population_size = 20
heights = np.random.uniform(150, 200, population_size)
weights = np.random.uniform(50, 100, population_size)

plt.figure(figsize = (10, 6))
plt.scatter(heights, weights, color = 'blue', alpha = 0.5)
plt.xlabel('Height (cm)')
plt.ylabel('Weight (Kg)')
plt.title('Height and Weight Chart')
plt.show()
```

<p style = "font-family:B Nazanin;">
    <b>نمودار دایره‌ای (Pie Chart):</b> از این نمودار معمولا زمانی استفاده می‌شود که می‌خواهیم مقدار یک متغیر را نسبت به مقدار کل نشان دهیم. در این نمودار کل مقادیر به صورت یک دایره در نظر گرفته می‌شود و مقدار هر متغیر بخشی از این دایره را شامل می‌شود. 
</p>

```python
import matplotlib.pyplot as plt
import numpy as np

activities = ['Sleeping', 'Working', 'Eating', 'Study', 'Others']
time_spent = [8, 8, 2, 3, 3]
plt.figure(figsize = (10, 6))
plt.pie(time_spent, labels = activities, autopct = '%1.1f%%', startangle = 90, colors = ['blue', 'green', 'orange', 'red', 'pink'])
plt.axis('equal')
plt.title('Time Spent Chart')
plt.legend()
plt.show()
```

<p style = "font-family:B Nazanin;">
    <b>نمودار شمعی (CandleStick Chart):</b> نمودارهای شمعی معمولا برای نشان‌دادن روند تغییرات قیمتی در یک بازه زمانی استفاده می‌شوند که اطلاعاتی مانند قیمت شروع، قیمت پایان، بالاترین قیمت و پایین‌ترین قیمت را دارند. برای رسم این گونه از نمودارها از ماژول mplfinance استفاده می‌کنیم. 
</p>

```python
import pandas as pd
import mplfinance as mpf

data = {
    'date': ['2023-08-01', '2023-08-02', '2023-08-03', '2023-08-04', '2023-08-05'],
    'open': [150, 155, 152, 154, 160],
    'high': [160, 160, 156, 158, 165],
    'low':  [145, 150, 150, 152, 158],
    'close':[155, 158, 153, 155, 163]
}

df = pd.DataFrame(data)
df['date'] = pd.to_datetime(df['date'])
df.set_index('date', inplace = True)
mpf.plot(df, type = 'candle', style = 'charles', title = 'CandleStick Chart')
```

<p style = "font-family:B Nazanin;">
    <b>نمودار باکس ویسکر (Box Plot):</b> این نمودار برای نمایش توزیع داده‌ها استفاده می‌شود. مثلا نمایش داده‌های پرت. 
</p>

```python
import matplotlib.pyplot as plt

data = [
    [12, 15, 20, 25, 30, 35, 40, 45, 50, 55],
    [15, 18, 22, 26, 30, 38, 42, 47, 52, 58],
    [10, 14, 18, 22, 27, 32, 38, 44, 50, 60]
]
plt.figure(figsize = (10, 6))
plt.boxplot(data, labels = ['A', 'B', 'C'])
plt.xlabel('Groups')
plt.ylabel('Values')
plt.title('Box and Whisker Plot')
plt.show()
```

<p style = "font-family:B Nazanin;">
    <b>نمودار هیستوگرام (Histogram Chart):</b> از این نوع نمودار برای نمایش توزیع فراوانی داده‌ها در بازه‌های مختلف استفاده می‌شود. 
</p>

```python
import matplotlib.pyplot as plt
import numpy as np

student_grades = np.random.uniform(0, 20, 30)
bins = [0, 10, 15, 18, 20]
colors = ['blue', 'green', 'orange', 'red']
N, bins, patches = plt.hist(student_grades, bins = bins, edgecolor = 'black')
for i in range(len(colors)):
    patches[i].set_facecolor(colors[i])
plt.xlabel('Grades')
plt.ylabel('Frequency')
plt.title('Histogram Student Grades')
plt.show()
```

<p style = "font-family:B Nazanin;">
    <b>نمودار حبابی (Bubble Chart):</b> مشابه با نمودار نقطه‌ای است با این تفاوت که برای مواردی کاربرد دارد که به جای دو متغیر، می‌خواهیم سه متغیر را روی نمودار نشان دهیم. در این جا هر کدام از متغیرها یک رنگ دارند و یا میزان بزرگی یا کوچکی آن نقطه مقدار آن متغیر سوم را مشخص می‌کند. 
</p>

```python
import matplotlib.pyplot as plt
import numpy as np

num_products = 20
prices = np.random.uniform(10, 100, num_products)
weights = np.random.uniform(0.5, 5, num_products)
popularity = np.random.uniform(0, 1, num_products)

plt.figure(figsize = (10, 6))
plt.scatter(prices, weights, s = popularity*100, c = popularity)
plt.xlabel('Price')
plt.ylabel('Weight')
plt.title('Bubble Chart')
plt.colorbar(label = 'Popularity')
plt.show()
```
