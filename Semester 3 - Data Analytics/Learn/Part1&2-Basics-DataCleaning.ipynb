<p style = "font-family:B Nazanin;">
   تجزیه و تحلیل داده‌ها فرایندی است که می‌تواند به کمک روش‌های کمی و کیفی به استخراج اطلاعات ارزشمند از داده‌ها بپردازد. مزایای استفاده از تجزیه و تحلیل داده‌ها عبارتند از: <br>
    &nbsp; &nbsp; * بهبود تصمیم‌گیری<br>
    &nbsp; &nbsp; * خدمات بهتر به مشتریان<br>
    &nbsp; &nbsp; * بازاریابی موثرتر<br>
    &nbsp; &nbsp; * عملیات کارآمدتر و صرفه‌جویی در هزینه‌ها <br>
    <br>
    <br>
    انواع روش‌های تجزیه و تحلیل داده‌ها عبارتند از:<br>
    &nbsp; &nbsp; * <b>تجزیه و تحلیل توصیفی:</b> در این روش داده‌های گذشته و حال بررسی می‌شود تا وضعیت فعلی توصیف شود و از روی آن بتوان روندها و الگوهای کاری را تشخیص داد. <br>
    &nbsp; &nbsp; * <b>تجزیه و تحلیل تشخیصی:</b> از روی داده‌ها تشخیص می‌دهیم که چه عوامل و دلایلی باعث ایجاد شرایط امروز شده‌است.<br>
    &nbsp; &nbsp; * <b>تجزیه و تحلیل پیش‌بینی‌کننده:</b> اگر بتوانیم تجزیه و تحلیل را به‌گونه‌ای انجام دهیم که از طریق آن بتوان پیش‌بینی انجام دهیم و یادگیری ماشین را از طریق این پیش‌بینی‌ها رقم بزنیم، به آن تجزیه و تحلیل پیش‌بینی‌کننده می‌گویند. معمولا در این روش مدلسازی آماری اتفاق می‌افتد. <br>
    &nbsp; &nbsp; * <b>تجزیه و تحلیل تجویزی:</b> اگر داده‌ها را تجزیه و تحلیل کرده و بررسی و آزمایش انجام دادیم و در نهایت توانستیم یک سری راه‌حل خاص را برای به‌دست‌آوردن نتایج مطلوب‌تر پیشنهاد بدهیم، می‌گوییم تجزیه و تحلیل تجویزی اتفاق افتاده‌است. <br>
    <br>
    در واقع این روش‌ها در ادامه یکدیگر هستند. یعنی ابتدا وضع گذشته و حال را توصیف می‌کنیم. سپس وضعیت کنونی را تشخیص می‌دهیم. از طریق این موارد آینده را پیش‌بینی می‌کنیم و بر اساس آن برای آینده راه‌حل‌های مناسب تجویز می‌کنیم. 
</p>

<p style = "font-family:B Nazanin;">
    <b>معیارهای کیفیت داده‌ها:</b><br>
    * دقت و صحت داده‌ها (Accuracy): داده‌ها باید صحیح باشند. اوپراتور یا سنسوری که داده‌ها را ثبت می‌کند، باید داده‌های صحیح را ثبت کند. <br>
    * کامل بودن داده‌ها (Completeness): تا حد امکان داده‌ها باید کامل و جامع باشد و نقصی در آن وجود نداشته باشد. <br>
    * قابل اطمینان بودن (Reliability): داده‌ها باید از مراجع و منابع قابل اطمینان و معتبر به دست‌آمده باشند. <br>
    * مرتبط بودن داده‌ها (Relevance): داده‌ها باید به هدف ما مرتبط باشد. جمع‌آوری داده‌های بی‌ربط سودی ندارد و کمکی نمی‌کند. <br>
    * قابل استفاده بودن (Usability): جمع‌بندی تمام موارد قبل قابل استفاده بودن را مشخص می‌کند. <br>
    * ثبات و سازگاری داده (Consistency): داده‌ها در تمام قسمت‌ها باید با هم سازگاری داشته باشند و تناقضی در آن‌ها دیده نشود. <br>
    * تفسیرپذیری داده‌ها (Interpretability): داده‌ها باید آنقدر شفاف و واضح باشد تا بتوان آن را تفسیر کرد. <br>
    * باورپذیری داده‌ها (Believability): داده‌ها باید منطقی و متناسب با واقعیت باشد. <br>
    * به موقع بودن داده‌ها (Timeliness): داده‌ها باید کاملا به‌روز باشد و به صورت بهینه آپدیت شده‌باشد. 
</p>

<p style = "font-family:B Nazanin;">
    <b>چهار عمل اصلی پیش‌پردازش داده‌ها:</b><br>
    پیش‌پردازش داده‌ها فرایندی است که روی داده‌های اولیه انجام می‌شود تا داده‌ها برای تجزیه و تحلیل آماده شود. این مراحل عباتند از:<br>
    * <b>پاکسازی داده‌ها (Data Cleaning):</b> مجموعه فرایندی که باعث می‌شود یک داده اصلاح شود یا اگر داده‌ای نادرست است از سیستم حذف گردد. <br>
    * <b>یکپارچه‌سازی داده‌ها (Data Integration):</b> به مجموعه فرایندهایی که باعث می‌شود داده‌ها از منابع مختلف گردآوری شده، تجمیع شوند و یکپارجه گردند، گفته می‌شود. <br>
    * <b>کاهش داده‌ها (Data Reduction):</b> فرایندی است که بدون این که به داده‌ها آسیبی وارد شود، تا حد امکان داده‌ها را کاهش دهیم تا فرایند پردازش آن‌ها در زمان کمتر با کیفیت بیشتری انجام شود. <br>
    * <b>تبدیل داده‌ها (Data Transformation) و گسسته‌سازی داده‌ها (Data Discretization):</b> به تبدیل فرمت و قالب داده‌ها گفته می‌شود به صورتی که به قالبی تبدیل شود که مناسب برای تجزیه و تحلیل باشد و همچنین تبدیل داده‌های پیوسته (که معمولا از سنسورها گرفته می‌شود) به داده‌های گسسته در این مرحله خواهد بود.  
</p>

<p style = "font-family:B Nazanin;">
    <b>فرایند پاکسازی داده‌ها:</b><br>
    داده‌هایی که نیاز به پاکسازی دارند:<br>
    * سلول خالی (Empty Cell)<br>
    * داده با فرمت اشتباه (Data in wrong format)<br>
    * داده اشتباهی (Wrong data)<br>
    * داده (رکورد) تکراری (Duplicated data)<br>
    <br>
    <b>فعالیت‌های بخش پاکسازی داده:</b><br>
    <b>1. حذف داده‌های تکراری (Remove duplicated data):</b> در این قسمت هدف حذف رکوردهایی است که تمام فیلدهایش یکسان است. ممکن است رکوردهایی وجود داشته باشد که تعدادی از فیلدهای آن با سایر رکوردها یکسان باشد ولی بعضی از فیلدهای آن متفاوت است. این رکوردها را حذف نمی‌کنیم. در اینجا راه تشخیص داده‌های تکراری با استفاده از pandas را بررسی می‌کنیم. فرض می‌کنیم دیتافریم زیر را داشته باشیم:
</p>

````python
import pandas as pd
df = pd.DataFrame({
    'Name': ['mehdi', 'ali', 'negin', 'sara', 'ahmad', 'mehdi', 'saeed', 'sara', 'negin', 'reza'],
    'Family': ['abbasi', 'rezaei', 'zandi', 'safari', 'tavakoli', 'abbasi', 'sadegi', 'safari', 'safari', 'mousavi'],
    'Age': [43,14,37,18,370,43,51,18,37,45]
})
df
````

<p style = "font-family:B Nazanin;">
     همانطور که مشخص است رکوردهای تکراری در این دیتافریم سطرهای 0 و 5 و همچنین سطرهای 3 و 7 هستند. با استفاده از تابع duplicated می‌توانیم سطرهای تکراری را تشخیص بدهیم:
</p>

````python
df = pd.DataFrame({
    'Name': ['mehdi', 'ali', 'negin', 'sara', 'ahmad', 'mehdi', 'saeed', 'sara', 'negin', 'reza'],
    'Family': ['abbasi', 'rezaei', 'zandi', 'safari', 'tavakoli', 'abbasi', 'sadegi', 'safari', 'safari', 'mousavi'],
    'Age': [43,14,37,18,370,43,51,18,37,45]
})
df.duplicated()
````

<p style = "font-family:B Nazanin;">
     این تابع یک پارامتر ورودی دلخواه به نام keep دارد. این پارامتر مشخص می‌کند کدام یک از سطرها را به عنوان تکراری محسوب کند. این پارامتر سه مقدار می‌تواند بگیرد: <br>
    - 'first': در این حالت اولین مورد از داده‌های تکراری را به عنوان داده اصلی حساب می‌کند و سایر سطرها را به عنوان سطر تکراری در نظر می‌گیرد. <br>
    - 'last': در این حالت آخرین مورد از داده‌های تکراری را به عنوان داده اصلی حساب می‌کند و سایر سطرها را به عنوان سطر تکراری در نظر می‌گیرد. <br>
    - False: در این حالت تمام سطرها را به عنوان تکراری در نظر می‌گیرد.
</p>

````python
df = pd.DataFrame({
    'Name': ['mehdi', 'ali', 'negin', 'sara', 'ahmad', 'mehdi', 'saeed', 'sara', 'negin', 'reza'],
    'Family': ['abbasi', 'rezaei', 'zandi', 'safari', 'tavakoli', 'abbasi', 'sadegi', 'safari', 'safari', 'mousavi'],
    'Age': [43,14,37,18,370,43,51,18,37,45]
})
print(df.duplicated(keep = 'first'))
print('*'*100)
print(df.duplicated(keep = 'last'))
print('*'*100)
print(df.duplicated(keep = False))
````

<p style = "font-family:B Nazanin;">
     در صورتی که بخواهیم معیار تشخیص تکراری بودن داده‌ها را ستون‌های خاصی قرار دهیم، دو راه مختلف داریم: یکی این که فقط ستون‌های مورد نظر را از دیتافریم انتخاب کنیم و تابع duplicated را روی آن اعمال کنیم: 
</p>

````python
df = pd.DataFrame({
    'Name': ['mehdi', 'ali', 'negin', 'sara', 'ahmad', 'mehdi', 'saeed', 'sara', 'negin', 'reza'],
    'Family': ['abbasi', 'rezaei', 'zandi', 'safari', 'tavakoli', 'abbasi', 'sadegi', 'safari', 'safari', 'mousavi'],
    'Age': [43,14,37,18,370,43,51,18,37,45]
})
display(df)
print(df['Name'].duplicated())
````

<p style = "font-family:B Nazanin;">
     دوم این که نام ستون‌های مورد نظر را به پارامتر subset از تابع duplicated بدهیم:
</p>

````python
df = pd.DataFrame({
    'Name': ['mehdi', 'ali', 'negin', 'sara', 'ahmad', 'mehdi', 'saeed', 'sara', 'negin', 'reza'],
    'Family': ['abbasi', 'rezaei', 'zandi', 'safari', 'tavakoli', 'abbasi', 'sadegi', 'safari', 'safari', 'mousavi'],
    'Age': [43,14,37,18,370,43,51,18,37,45]
})
display(df)
print(df.duplicated(subset = ['Name', 'Age']))
````

<p style = "font-family:B Nazanin;">
    برای حذف داده‌های تکراری از تابع drop_duplicated() در ماژول pandas می‌توانیم استفاده کنیم. پارامتر inplace این تابع اگر برابر False باشد، حذف داده‌های تکراری روی همان دیتافریم اثر نمی‌کند، بلکه یک کپی از این دیتافریم ایجاد شده و تغییرات روی آن اعمال می‌شود و در نهایت دیتافریم پاکسازی‌شده خروجی داده می‌شود: 
</p>

````python
df = pd.DataFrame({
    'Name': ['mehdi', 'ali', 'negin', 'sara', 'ahmad', 'mehdi', 'saeed', 'sara', 'negin', 'reza'],
    'Family': ['abbasi', 'rezaei', 'zandi', 'safari', 'tavakoli', 'abbasi', 'sadegi', 'safari', 'safari', 'mousavi'],
    'Age': [43,14,37,18,370,43,51,18,37,45]
})
display(df)
df2 = df.drop_duplicates(inplace = False)
display(df2)
````

<p style = "font-family:B Nazanin;">
     تابع drop_duplicates نیز پارامتر subset دارد که مشخص می‌کند معیار تشخیص داده‌های تکراری کدام ستون‌های دیتافریم باشد:
</p>

````python
df = pd.DataFrame({
    'Name': ['mehdi', 'ali', 'negin', 'sara', 'ahmad', 'mehdi', 'saeed', 'sara', 'negin', 'reza'],
    'Family': ['abbasi', 'rezaei', 'zandi', 'safari', 'tavakoli', 'abbasi', 'sadegi', 'safari', 'safari', 'mousavi'],
    'Age': [43,14,37,18,370,43,51,18,37,45]
})
display(df)
df2 = df.drop_duplicates(subset = ['Name'], inplace = False)
display(df2)
````

<p style = "font-family:B Nazanin;">
     تابع drop_duplicates نیز پارامتر ورودی keep را دارد که مشخص می‌کند از بین داده‌های تکراری کدام حفظ شده و کدام حذف شود. 'first' اولین مورد از داده‌های تکراری را نگه می‌دارد، 'last' آخرین مورد را نگه می‌دارد و False همه داده‌های تکراری را حذف می‌کند:
</p>

````python
df = pd.DataFrame({
    'Name': ['mehdi', 'ali', 'negin', 'sara', 'ahmad', 'mehdi', 'saeed', 'sara', 'negin', 'reza'],
    'Family': ['abbasi', 'rezaei', 'zandi', 'safari', 'tavakoli', 'abbasi', 'sadegi', 'safari', 'safari', 'mousavi'],
    'Age': [43,14,37,18,370,43,51,18,37,45]
})
display(df)
df2 = df.drop_duplicates(keep = 'first', inplace = False)
display(df2)
df3 = df.drop_duplicates(keep = 'last', inplace = False)
display(df3)
df4 = df.drop_duplicates(keep = False, inplace = False)
display(df4)
````

<p style = "font-family:B Nazanin;">
    <b>2. پر کردن فیلدهای بدون مقدار (Fill in missing values):</b> خالی ماندن بعضی از سلول‌ها در فرایند پردازش داده‌ها اختلال ایجاد می‌کند. برای رفع این اختلال چند راهکار داریم:<br>
    یکی از راه‌ها این است که کل آن سطر را حذف کنیم. اگر آن سلول خالی به قدری مهم باشد که نبود آن کل ارزش داده آن سطر را از بین ببرد، باید کل آن سطر را حذف کرد. <br>
    روش دوم پر کردن دستی داده‌ها است. برای دیتاست‌های کوچک این روش قابل استفاده است که یکی یکی سلول‌های خالی را بررسی کنیم و برای هر کدام مقدار مناسب را وارد کنیم. اما برای دیتاست‌های بزرگ این روش مناسب نیست.<br>
    روش سوم پر کردن با یک مقدار ثابت است. در این روش تمام سلول‌های خالی مربوط به یک ستون با یک مقدار ثابت پر می‌شود که باز هم ممکن است در بعضی موارد مناسب نباشد. <br>
    روش چهارم پر کردن با مقدار میانگین یا میانه است. در این روش میانگین یا میانه داده‌های موجود را محاسبه می‌کنیم و تمام سلول‌های خالی مربوط به آن ستون را با آن مقدار پر می‌کنیم. <br>
    روش پنجم پر کردن با محتمل‌ترین مقدار است. در این روش با استفاده از روش‌های آماری و تحلیل‌هایی حدس می‌زنیم احتمال اینکه داده‌های سلول‌های خالی چه مقداری باشد، بییشتر است و همان مقدار را برایش در نظر می‌گیریم. این روش یکی از بهترین و دقیق‌ترین روش‌هاست ولی هزینه عملیاتی بالایی دارد. 
</p>

````python
import pandas as pd
import numpy as np
df = pd.DataFrame({
    'name': ['mehdi', 'ali', np.nan, np.nan, 'reza', 'negin', np.nan, 'sara', np.nan, 'mehdi'],
    'family': ['abbasi', 'rezaei', np.nan, 'ahmadi', np.nan, 'abbasi', np.nan, 'rahmati', np.nan, 'akbary'],
    'age': [43, np.nan, np.nan, np.nan, 31, 18, np.nan, np.nan, 20, 31]
})
display(df)
````

<p style = "font-family:B Nazanin;">
     تابع isnull نشان می‌دهد که کدام سلول‌ها مقدار ندارند یا به عبارتی null یا NaN هستند:
</p>

````python
print(df.isnull())
````

<p style = "font-family:B Nazanin;">
     تابع isna نشان می‌دهد کدام سلول‌ها مقدار NaN دارند:
</p>

````python
print(df.isna())
````

<p style = "font-family:B Nazanin;">
     تابع notnull برعکس تابع isnull عمل می‌کند و هر جا داده‌ای مقدار داشته‌باشد مقدار True را برمی‌گرداند:
</p>

````python
print(df.notnull())
````

<p style = "font-family:B Nazanin;">
     توابع معرفی‌شده در بالا می‌توانند روی چند ستون خاص یک دیتافریم هم اعمال شود:
</p>

````python
print(df['age'].isnull())
````

````python
print(df['age'].notnull())
````

<p style = "font-family:B Nazanin;">
     این توابع را می‌توانیم روی ماژول pandas نیز مستقیما اعمال کنیم و نام دیتافریم را به عنوان پارامتر ورودی به آن بدهیم:
</p>

````python
print(pd.isnull(df['age']))
````

````python
print(pd.notnull(df['age']))
````

<p style = "font-family:B Nazanin;">
     اگر نخواهیم یک مقدار ثابت را جایگزین کنیم، پارامتر method پارامتری است که مشخص می‌کند روش جایگزین‌کردن چه باشد. اگر مقدار این پارامتر 'bfill' باشد، هر مقدار NaN با مقدار بعدی خودش پر می‌شود. یعنی در این روش از آخر شروع می‌کنیم و به عقب برمی‌گردیم و یکی یکی مقادیر NaN را با مقدار بعدی خودش پر می‌کنیم:

````python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'name': ['mehdi', 'ali', np.nan, np.nan, 'reza', 'negin', np.nan, 'sara', np.nan, 'mehdi'],
    'family': ['abbasi', 'rezaei', np.nan, 'ahmadi', np.nan, 'abbasi', np.nan, 'rahmati', np.nan, 'akbary'],
    'age': [43, np.nan, np.nan, np.nan, 31, 18, np.nan, np.nan, 20, 31]
})
display(df)

df2 = df.fillna(method = 'bfill')
display(df2)
````

<p style = "font-family:B Nazanin;">
     اگر مقدار پارامتر method، برابر با 'ffill' بود، هر مقدار NaN با مقدار قبلی خودش پر می‌شود. یعنی از ابتدای DataFrame شروع شده و یکی یکی مقادیر NaN با مقدار قبلی خودش پر می‌شود.

<p style = "font-family:B Nazanin;">
      تمامی موارد قبلی را نیز در این دیکشنری می‌توان به کار برد. مثلا هر ستون را با مد مربوط به خودش پر کرد:

<p style = "font-family:B Nazanin;">
      پارامتر axis مشخص می‌کند که تغییرات به صورت سطری انجام شود یا ستونی. اگر 0 باشد ستونی و اگر 1 باشد سطری اعمال می‌شود. پارامتر limit هم مشخص می‌کند حداکثر چند مقدار NaN با مقدار مشخص‌شده پر شود. مثلا در کد زیر در هر ستون حداکثر 2 مقدار با *** پر می‌شود:

<p style = "font-family:B Nazanin;">
    <b>3. برطرف کردن خطاهای نویز (Smooth Noisy Data):</b> داده‌ای که توسط یک ماشین یا سیستم به درستی قابل درک و فهم نباشد، یک نویز است. نویز حتی ممکن است در گذر زمان نیز ایجاد شود. یعنی یک سیستم داده‌هایی را خوانده، ذخیره کرده و حتی تغییر داده است اما پس از مدتی دیگر نمی‌تواند آن را بخواند. این نشان می‌دهد که این داده در گذر زمان تغییر پیدا کرده و نویز پیدا کرده است. <br>
    دلایل به وجود آمدن نویز در داده‌ها: <br>
    &nbsp;&nbsp;&nbsp; * خرابی‌های سخت‌افزاری: مثلا مشکلات سخت‌افزاری در یک سنسور که داده‌ها را جمع‌آوری می‌کند باعث می‌شود که داده‌ها با خطا ذخیره شوند.<br>
    &nbsp;&nbsp;&nbsp; * خطاهای برنامه‌نویسی: مثلا اشتباهات محاسباتی و خطاهایی که در حین برنامه‌نویسی رخ می‌دهد، داده‌ها را دچار خطا می‌کند. <br>
    &nbsp;&nbsp;&nbsp; * اشتباهات املایی: بیشتر در وارد کردن دستی داده‌ها ایجاد می‌شود که داده‌ای اشتباه در سیستم ثبت گردد. <br>
    به طور کلی هر داده‌ای که خراب باشد و اگر آن را حذف نکنیم باعث پایین آمدن کیفیت تجزیه و تحلیل ما بشود، نویز است. <br><br>
    <b>انواع نویز:</b><br>
    &nbsp;&nbsp;&nbsp; * <b>نویز کلاس (نویز برچسب):</b><br>
    &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; - <b>مقادیر متضاد:</b> مثلا دو رکورد کاملا مشابه یکدیگر هستند و فقط در یک فیلد با هم متفاوت‌اند. مثلا اطلاعات یک دانشجو شامل شماره دانشجویی و نام و نام خانوادگی و کد ملی و ... در دو رکورد مشابه است فقط در یکی فیلد فارغ‌التحصیلی True ثبت شده است و در یکی False. به این موارد نویز متضاد گفته می‌شود که فقط با تحقیق و بررسی می‌تون آن‌ها را درست کرد. یعنی دو رکورد داریم که واقعا نمی‌دانیم کدام درست و کدام غلط است. <br>
    &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; <b>طبقه‌بندی اشتباه مقادیر:</b> گاهی ممکن است به دلیل ناآگاهی یا دلایل دیگر داده‌ها در فیلدها یا ستون‌های اشتباهی ذخیره شود یا برخی از ویژگی‌ها و صفت‌های ثبت‌شده اشتباه باشند. مثلا جنس یک نوع پارچه به دلیل ناآگاهی کاربر ثبت‌کننده اشتباه ثبت شود. به این نوع نویز، نویز کلاس از جنس طبقه‌بندی اشتباه گفته می‌شود. <br>
    &nbsp;&nbsp;&nbsp;<b>نویز ویژگی:</b><br>
    &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;<b>مقادیر اشتباه:</b> مثلا قیمت یک کالا به جای این که به ریال ثبت شود، به تومان ثبت شده‌است. یا مثلا کاربر در هنگام وارد کردن سن یک شخص، به جای عدد 35، عدد 350 را وارد کرده‌است. <br>
    &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;<b>مقادیر گم‌شده:</b> مقادیری که کلا در دیتاست ثبت نشده‌است. مانند NaN یا Null<br>
    &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;<b>مقادیر ناقص:</b> مقادیری که به صورت کامل وارد نشده‌است. مثلا یک توضیح در مورد یک کالا که ناقص است و جمله‌بندی‌اش کامل نشده‌است.<br><br><br>
    <b>روش‌های برخورد با داده‌های نویزی:</b><br>
    &nbsp;&nbsp;&nbsp; - Binning Method <br>
    &nbsp;&nbsp;&nbsp; - Regression <br>
    &nbsp;&nbsp;&nbsp; - Clustering<br>
    <br>
    در این قسمت فقط روش اول بررسی می‌شود و روش‌های دیگر پس از یادگیری مباحث آماری بحث می‌شوند. <br>
    در روش Binning Method یا روش دسته‌بندی، ابتدا داده‌های اولیه را مرتب می‌کنیم. سپس داده‌ها را در چند دسته (Bin) تقسیم می‌کنیم که هر کدام معرف یک جامعه آماری کوچک‌تر است که می‌توان جداگانه نویز آن‌ها را برطرف کرد. مرحله بعد محاسبه یک مقدار جایگزین برای مقادیر آن دسته است (مثلا میانگین آن دسته) و در مرحله آخر کل داده‌های یک دسته را با مقدار محاسبه شده در مرحله قبل جایگزین می‌کنیم. مثال: 
</p>

```
1) Raw Data:
    8,16,9,15,21,21,24,30,26,27,30,34

2) Sorting:
    8,9,15,16,21,21,24,26,27,30,30,34
    
3) Bins:
    Bin1: 8,9,15,16
    Bin2: 21,21,24,26
    Bin3: 27,30,30,34
    
4) Mean:
    Bin1 Mean: 12
    Bin2 Mean: 23
    Bin3 Mean: 30
    
5) Replace:
    Bin1: 12,12,12,12
    Bin2: 23,23,23,23
    Bin3: 30,30,30,30
```

<p style = "font-family:B Nazanin;">
      در این روش به جای استفاده از میانگین روش‌های دیگری را نیز می‌توان به کار برد. از جمله این روش‌ها می‌توان به روش مرز دسته اشاره کرد. برای هر دسته مقدار مینیمم و ماکزیمم داده‌ها را پیدا کرد و برای داده‌های بین این دو مقدار، اگر به مقدار مینیمم نزدیک‌تر بود، مقدار مینیمم را جایگزین آن کرد و اگر به مقدار ماکزیمم نزدیک‌تر بود، مقدار ماکزیمم را جایگزین آن مرد. به عنوان مثال در داده‌های زیر، در دسته اول مقدار مینیمم و ماکزیمم برابر 8 و 16 است. عدد 9 به 8 نزدیک‌تر است، پس به جای آن 8 می‌گذاریم و عدد 15 به 16 نزدیک‌تر است، پس به جای آن 16 می‌گذاریم و به همین ترتیب برای همه دسته‌ها این کار را انجام می‌دهیم.
</p>

```
1) Raw Data:
    8,16,9,15,21,21,24,30,26,27,30,34

2) Sorting:
    8,9,15,16,21,21,24,26,27,30,30,34
    
3) Bins:
    Bin1: 8,9,15,16
    Bin2: 21,21,24,26
    Bin3: 27,30,30,34
    
4) Boundaries:
    Bin1 Min = 8    Max = 16
    Bin2 Min = 21   Max = 26
    Bin3 Min = 27   Max = 34
    
5) Replace:
    Bin1: 8,8,16,16
    Bin2: 21,21,26,26
    Bin3: 27,27,27,34
```

<p style = "font-family:B Nazanin;">
      پیاده‌سازی در پایتون:
</p>

````python
import numpy as np

# Load data
data = [35, 5, 11, 72, 92, 13, 15, 50, 10, 55, 204, 215]
data = np.sort(data)
print(data)
m = 3    # Number of Bins
````

````python
# Create Bins
def createBins(data, m):
    l = len(data)
    n = int(l / m)
    bins = []
    for i in range(0, m):
        bin = []
        for j in range(i*n, (i+1)*n):
            bin.append(data[j])
        bins.append(bin)
    return bins

bins = createBins(data, m)
for bin in bins:
    print(bin)
````

````python
# Bin Means
data = []
for i in range(len(bins)):
    Bin = bins[i].copy()
    mean = np.mean(Bin)
    for j in range(len(Bin)):
        Bin[j] = mean
    data.extend(Bin)
print(data)
````

````python
# Bin Boundaries
data = []
for i in range(len(bins)):
    Bin = bins[i]
    l = len(Bin)
    for j in range(l):
        if (Bin[j] - Bin[0] < Bin[l-1] - Bin[j]):
            Bin[j] = Bin[0]
        else:
             Bin[j] = Bin[l-1]
    data.extend(Bin)
print(data)
````

<p style = "font-family:B Nazanin;">
    <b>4. شناسایی و حذف داده‌های پرت (Identify and Remove Outliers):</b> داده پرت یا دورافتاده داده‌ای است که صحیح است ولی فاصله زیادی با سایر داده‌ها دارد و اختلافش با سایر داده‌ها زیاد است. مثلا اگر بخواهیم اطلاعات دانشجویان را تجزیه و تحلیل کنیم و به عنوان مثال بازه سنی دانشجویان بین 18 تا 24 سال باشد ولی دانشجویی داشته باشیم که سن 60 سال داشته باشد، اگر اطلاعات این دانشجو را نیز در تجزیه و تحلیل خود بیاوریم، کار اشتباهی نکرده‌ایم و داده‌های ما درست است ولی ممکن است این داده، مثلا در هنگام محاسبه میانگین سن دانشجویان، مقدار میانگین را به میزان زیادی تغییر دهد. <br>
    داده‌های پرت نیز می‌توانند به نوعی نویز محسوب شوند چون تجزیه و تحلیل ما را به هم می‌ریزند ولی تفاوت اینجاست که داده‌های درست نیز می‌توانند داده پرت باشند. <br><br>
    <b>دلایل به‌وجود آمدن داده پرت:</b><br>
    &nbsp;&nbsp;&nbsp; * اشتباه در طراحی مقیاس اندازه‌گیری: اگر داده‌های ما معیار درستی نداشته‌باشند و محدودیتی برای آن‌ها تعیین نکنیم، ممکن است داده پرت به وجود آید. <br>
    &nbsp;&nbsp;&nbsp; * اشتباه در ورود و جمع‌آوری داده‌ها: ممکن است در هنگام ورود اطلاعات اشتباهی رخ دهد و داده نادرستی ثبت شود که می‌توان گفت این داده همان نویز است. <br>
    &nbsp;&nbsp;&nbsp; * انتخاب نادرست نمونه‌های آماری: هنگام انجام تجزیه و تحلیل باید جامعه آماری درستی را انتخاب کنیم. مثلا جامعه آماری دانشجویان زیر 25 سال (که دیگر مشکل دانشجو با سن بالا پیش نیاید) <br> <br>
    <b>روش‌های تشخیص داده پرت:</b><br>
    &nbsp;&nbsp;&nbsp; * رسم نمودار: با رسم نمودار به صورت بصری می‌توانیم تشخیص دهیم کدام داده‌ها از سایر داده‌ها فاصله بیشتری دارند و داده پرت محسوب می‌شوند. <br>
    &nbsp;&nbsp;&nbsp; * مقایسه داده‌ها با امتیاز استاندارد (z) <br>
    &nbsp;&nbsp;&nbsp; * محدوده بین چارکی (IQR) <br>
    روش دوم و سوم در آینده توضیح داده خواهد شد.
</p>

<p style = "font-family:B Nazanin;">
    <b>5. رفع تناقضات و ناسازگاری داده‌ها (Resolve Inconsistencies):</b> اگر در یک دیتاست داده‌های تکراری ذخیره شده باشد که در بعضی از فیلدها و مقادیر با هم تفاوت (تناقض) داشته باشد، می‌گوییم تناقض در داده‌ها اتفاق افتاده است. مثلا اطلاعات یک شخص دوبار در یک دیتاست ذخیره شده است و در این دو رکورد شماره تلفن متفاوت ثبت شده است. در اینجا ناسازگاری اطلاعات را مشاهده می‌کنیم. <br>
    <b>دلایل به‌وجود آمدن ناسازگاری اطلاعات:</b><br>
    &nbsp;&nbsp;&nbsp; * یکپارچه‌نبودن سیستم ذخیره اطلاعات: مثلا اطلاعات یک شخص توسط چند سیستم یا اپلیکیشن مختلف ثبت شده‌اند و هر بار یک شماره تلفن از این فرد ذخیره شده است. یا این که شماره تلفن یک فرد تغییر کرده ولی شماره جدید در همه سیستم‌ها تغییر نکرده است. <br>
    &nbsp;&nbsp;&nbsp; * به‌روز شدن داده‌ها: در هنگام به‌روزشدن داده‌ها بخشی از داده‌ها به‌روز شده و بخشی دیگر به‌روز نشده‌اند. <br>
    &nbsp;&nbsp;&nbsp; * داده‌های چند منبع به درستی ادغام نشده‌اند. <br><br><br>
    <b>افزونگی داده‌ها (Data Redundancy):</b> به ذخیره شدن مکرر یک داده در یک سیستم یا دیتاست حتی اگر مشکل ناسازگاری یا تناقض نیز نداشته باشد، افزونگی داده گفته می‌شود. این نیز یک مشکل است که در سیستم‌های غیریکپارچه به وفور دیده می‌شود. <br>
    <b>مشکلات حاصل از فزونگی داده‌ها:</b>
    &nbsp;&nbsp;&nbsp; * فضای ذخیره‌سازی تلف‌شده <br>
    &nbsp;&nbsp;&nbsp; * به‌روزرسانی پایگاه داده دشوارتر است. <br>
    &nbsp;&nbsp;&nbsp; * منجر به ناسازگاری داده‌ها خواهد شد. <br>
    &nbsp;&nbsp;&nbsp; * بازیابی داده‌ها کند و ناکارآمد خواهد شد.<br>
    <br>
    پس نکته مهم فهم تفاوت ناسازگاری داده و افزونگی داده است. ناسازگاری داده یعنی وجود موجودیت‌های تکراری با مقادیر غیر یکسان اما افزونگی داده یعنی وجود داده‌های تکراری با مقادیر یکسان.
