    <h1>فصل سوم: تجزیه و تحلیل داده‌ها</h1>
    <h2>بخش سوم: یکپارچه‌سازی داده‌ها</h2>
</span>

<p style="font-family:B Nazanin">
   یکپارچه‌سازی داده‌ها فرايندی برای ترکیب و ادغام داده‌ها از منابع مختلف است به شکلی که در نهایت داده‌ها در یک نمای منسجم وجود داشته باشند. مثلا داده‌ها در اپلیکیشن‌های مختلف وجود دارد و یا توسط کاربرن متفاوتی ذخیره شده‌اند و می‌خواهیم آن‌ها را با هم ادغام کنیم. این داده‌ها ممکن است از نظر نوع قالب و ساختار با یکدیگر تفاوت داشته باشند و هر کدام از داده‌ها با ساختار متفاوتی ذخیره شده‌اند. <br>
    پس هدف از یکپارچه‌سازی اطلاعات این است که تمام داده‌ها منسجم و با ساختار یکسان باشند تا کار تجزیه و تحلیل داده‌ها ساده‌تر و با دقت بیشتر انجام شود. <br>
    کاربردهای یکپارچه‌سازی داده‌ها: <br>
    &nbsp;&nbsp;&nbsp; * هوش تجاری (Business Intelligence) <br>
    &nbsp;&nbsp;&nbsp; * انبار داده‌ها (Data Warehousing) <br>
    &nbsp;&nbsp;&nbsp; * مدیریت داده‌ها (Data Management) <br>
    &nbsp;&nbsp;&nbsp; * تجزیه و تحلیل داده‌ها (Data Analytics) <br>
    <br><br>
    <b>مدل‌های یکپارچه‌سازی داده‌ها:</b> <br>
    &nbsp;&nbsp;&nbsp; * <b>اتصال کامل (Tight Coupling):</b> در این روش همه داده‌ها از تمام منابع جمع‌آوری شده و در یک انبار مرکزی ذخیره می‌شوند. داده‌ها از سازمان‌ها، اپلیکیشن‌ها و منابع مختلف جمع شده، سازگار می‌شوند، داده‌های تکراری حذف می‌شود و در نهایت داده‌های یکپارچه در یک انبار متمرکز ذخیره می‌شود. در این روش هزینه انجام کار بالاتر است اما حسن این روش این است که سازگاری داده‌ها بالاتر، افزونگی داده‌ها کمتر و تحلیل داده‌ها راحت‌تر است. <br>
    &nbsp;&nbsp;&nbsp; * <b>اتصال ضعیف (Loose Coupling):</b> در این روش قرار نیست که داده‌ها از منابع مختلف در یک انبار مرکزی جمع شوند. هر منبع داده‌های خود را حفظ می‌کند ولی با استفاده از یک سیستم طراحی‌شده، این داده‌ها به صورت اشتراکی استفاده می‌شوند. در این روش هزینه تجمیع داده‌ها بسیار کمتر است اما احتمال ناسازگاری داده‌ها بیشتر خواهد بود. پیچیدگی کار با این سیستم بیشتر است و تجزیه و تحلیل داده‌ها سخت‌تر خواهد بود چون داده‌ها به صورت مجزا از هم ذخیره شده‌اند و ناسازگاری و افزونگی آن‌ها بیشتر است. <br><br>
    <b>مشکلات موجود پس از یکپارچه‌سازی داده‌ها:</b><br>
    &nbsp;&nbsp;&nbsp; * تغییر در کیفیت داده‌ها: گاهی مجبور هستیم برای ادغام دو داده با یکدیگر، برخی از مقادیر را حذف کنیم یا دقت برخی از مقادیر را کاهش دهیم. <br>
    &nbsp;&nbsp;&nbsp; * ناهمگونی داده‌ها: برخی از مقادیر ممکن است در منابع مختلف به صورت متفاوت ذخیره شده‌باشند. مثلا تاریخ‌ها در یک سیستم به صورت شمسی ذخیره شده و در سیستم دیگر میلادی باشند. <br>
    &nbsp;&nbsp;&nbsp; * برهم خوردن حفظ حریم خصوصی و امنیت داده‌ها: داده‌ها با منابع مختلف به اشتراک گذاشته می‌شوند و دیگر مختص به یک منبع خاص نیستند. این مشکل به خصوص در حالت اتصال ضعیف بیشتر رخ می‌دهد. <br>
    &nbsp;&nbsp;&nbsp; * تاثیر منفی بر عملکرد سیستم<br>
    <br>
    <b>مزایای انجام عملیات یکپارچه‌سازی:</b><br>
    &nbsp;&nbsp;&nbsp; * کاهش پیچیدگی داده‌ها<br>
    &nbsp;&nbsp;&nbsp; * یکپارچگی داده‌ها <br>
    &nbsp;&nbsp;&nbsp; * تبدیل آسان داده‌ها <br>
    &nbsp;&nbsp;&nbsp; * تصمیمات تجاری هوشمندانه‌تر <br>
    <br>
    <b>عملیاتی که در فرایند یکپارچه‌سازی داده‌ها باید انجام شود:</b><br>
    &nbsp;&nbsp;&nbsp; * یکپارچه‌سازی اسکیماها: در ابتدا باید اسکیمای دیتابیس‌ها با یکدیگر یکسان شوند. حتی اگر بخواهیم از روش اتصال ضعیف نیز استفاده کنیم، ساختار و اسکیمای دیتابیس‌ها باید با هم یکسان باشند. <br>
    &nbsp;&nbsp;&nbsp; * تشخیص و برطرف کردن افزونگی داده‌ها: در مرحله بعد (به خصوص در روش اتصال کامل) باید افزونگی داده‌ها را تشخیص دهیم و داده‌های تکراری را حذف کنیم. <br>
    &nbsp;&nbsp;&nbsp; * حل تضاد ارزش داده‌ها: باید داده‌ها بررسی شوند تا تناقضی در داده‌ها وجود نداشته باشد. مثلا اطلاعات یک شخص اگر در دو منبع ذخیره شده باشد، باید چک کنیم که مثلا شماره موبایل یکسانی داشته باشد. <br>
    <br>
    <b>روش‌های یکپارچه‌سازی داده‌ها:</b><br>
    &nbsp;&nbsp;&nbsp; * یکپارچه‌سازی دستی (Manual Integration): در این روش که برای تعداد داده کم کاربرد دارد، داده‌ها یکی یکی توسط یک کاربر بررسی شده و با یکدیگر تجمیع و ادغام می‌شوند. در این روش باید حجم داده‌ها محدود باشد وگرنه یا زمان بسیار زیادی طول خواهد کشید یا غیرممکن خواهد بود. <br>
    &nbsp;&nbsp;&nbsp; * استفاده از میان‌افزارها (Middleware Integration): گاهی سازمان‌ها سیستم‌ها و نرم‌افزارهای خود را به‌روز می‌کنند و سیستم‌ها و دیتابیس‌های جدیدی را استفاده می‌کنند، در اینجا لازم است تا یک میان‌افزار استفاده شود تا اطلاعات را از سیستم‌های قدیمی جمع‌آوری کرده و در سیستم جدید ثبت کند. <br>
    &nbsp;&nbsp;&nbsp; * یکپارچه‌سازی بر مبنای نرم‌افزار (Application-Based Integration): در این روش که به تخصص بالایی نیاز دارد، یک نرم‌افزار (اپلیکیشن) نوشته می‌شود تا داده‌های سیستم‌های قدیمی را برداشته، پاکسازی کند و پس از ادغام با یکدیگر در سیستم جدید جمع‌آوری کند. <br>
    &nbsp;&nbsp;&nbsp; * یکپارچه‌سازی دسترسی یکنواخت (Uniform Access Integration): این روش که از نوع اتصال ضعیف است، تکنیکی است که داده‌ها را بازیابی می‌کند و به طور یکنواخت از آن‌ها استفاده می‌کند اما داده‌ها همچنان در منبع اصلی خود باقی می‌ماند. <br>
    &nbsp;&nbsp;&nbsp; * یکپارچه‌سازی حافظه مشترک (Common Storage Integration): در این حالت که ترکیبی از حالت اتصال کامل و ضعیف است، در فرایند کار از اتصال ضعیف استفاده می‌کند، یعنی داده‌ها در هر لحظه که نیاز داشتیم از منابع مختلف بارکشی می‌شوند و یک کپی از آن‌ها را در منبع جدید ذخیره می‌کنیم و در گذر زمان یک پایگاه داده جدید تشکیل می‌شود. پس در گذر زمان یک انبار داده یکپارچه و جدید تشکیل می‌شود. <br>
    <br>
    <b>ابزارهای یکپارچه‌سازی داده‌ها:</b><br>
    &nbsp;&nbsp;&nbsp; * ابزارهای یکپارچه‌سازی داده On-Premise: <br>
    مانند: <u>Microsoft SQL SERVER Integration Services (SSIS)</u> و Talend و IBM InfoSphere DataStage <br>
    &nbsp;&nbsp;&nbsp; ابزارهای یکپارچه‌سازی داده Open Source: <br>
    مانند: Pentaho و <u>Apache Kafka</u> و Apache NiFi <br>
    &nbsp;&nbsp;&nbsp; * ابزارهای یکپارچه‌سازی داده مبتنی بر Cloud: <br>
    مانند: Google Cloud Data Fusion و Microsoft Azure Data Factory و Amazon Web Services (AWS) Glue
</p>
