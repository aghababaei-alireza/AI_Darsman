    <h1>فصل سوم: تجزیه و تحلیل داده‌ها</h1>
    <h2>بخش چهارم: کاهش داده‌ها</h2>
</span>

<p style="font-family:B Nazanin">
   یکی از راه‌های موثر در بهبود کیفیت تجزیه و تحلیل داده‌ها، کاهش داده‌ها و کم کردن حجم و ابعاد داده‌ها است به شرط این که داده‌های مفید از بین نروند. در کل این مبحث هرگاه از حذف یک داده یا بخشی از داده صحبت می‌کنیم، به این شرط است که اثر منفی روی نتیجه تجزیه و تحلیل داده‌ها نداشته باشد. کاهش اندازه داده‌ها می‌تواند در اثر حذف کردن یا کنار گذاشتن یک تعداد ویژگی باشد یا اینکه خود محتواها و مقادیر را حذف کنیم. مثلا فرض کنیم یک فایل اکسل داریم که اطلاعات یک میلیون شخص را در خود ذخیره کرده است که این اطلاعات در 12 ستون قرار گرفته است. به طور کلی کاهش داده‌ها در این مثال می‌تواند به صورت حذف بعضی از سطرها، حذف بعضی از ستون‌ها یا کم کردن محتوای بعضی از سلول‌ها باشد. <br>
    <br>
    <b>روش‌های کاهش داده‌ها:</b><br>
    &nbsp;&nbsp;&nbsp; * انتخاب ویژگی: در این روش، از بین ویژگی‌های موجود، یک تعدادی را انتخاب می‌کنیم و سایر ویژگی‌ها را در نظر نمی‌گیریم. (باز هم مشروط بر اینکه اثری بر نتیجه تجزیه و تحلیل داده‌ها نداشته باشد.) مثلا اگر بخواهیم روی اثر جنسیت یک مشتری بر نحوه و میزان خرید او کار و بررسی کنیم، در اینجا وجود شماره موبایل مشتری در این تجزیه و تحلیل ضروری نیست. پس این فیلد را در نظر نمی‌گیریم. <br>
    &nbsp;&nbsp;&nbsp; * استخراج ویژگی: در این روش سعی می‌کنیم با ترکیب و ادغام چند ویژگی، یک ویژگی جدید را به‌وجود آوریم به طوری که باعث شود حجم داده‌ها کاهش یابد نه این که بدتر داده‌ها را بزرگتر و زیادتر کنیم. <br>
    &nbsp;&nbsp;&nbsp; * نمونه‌برداری: در برخی از دیتاست‌ها می‌توان از بین کل داده‌ها، نمونه‌هایی را انتخاب کرد که ویژگی‌های آن‌ها معرف ویژگی‌های کل دیتاست باشد. به این روش نمونه‌برداری می‌گویند. مثلا اگر بتوانیم از بین یک میلیون داده از اطلاعات اشخاص مختلف، بتوانیم چند هزار داده را جدا کنیم که اطلاعات آن‌ها معرف اطلاعات کل افراد باشد، تجزیه و تحلیل این چندهزار داده بسیار ساده‌تر از آن یک میلیون داده است. <br>
    &nbsp;&nbsp;&nbsp; * خوشه‌بندی: در این روش سعی می‌کنیم داده‌هایی که ویژگی‌های مشابهی دارند را در یک خوشه قرار دهیم و در نهایت ما خوشه‌ها را نگهداری می‌کنیم به جای اینکه کل داده‌ها را نگه‌داریم. <br>
    &nbsp;&nbsp;&nbsp; * فشرده‌سازی: این روش باعث کم شدن حجم داده‌ها، کم شدن فضای ذخیره‌سازی و افزایش سرعت پردازش می‌شود ولی تاثیر آنچنانی در نتیجه تجزیه و تحلیل نمی‌گذارد. <br>
    <br>
    <b>تکنیک‌های کاهش داده‌ها:</b><br>
    &nbsp;&nbsp;&nbsp; * کاهش ابعاد (Dimensionality Reduction): روشی است که تلاش می‌کند ویژگی‌های غیرضروری را حذف کند (پس به نوعی ترکیبی از روش‌های انتخاب ویژگی و استخراج ویژگی است). <br>
    &nbsp;&nbsp;&nbsp; * کاهش تعداد (Numerosity Reduction): این تکنیک مانند خوشه‌بندی و نمونه‌برداری سعی می‌کند که تعداد نقاط داده‌ای را کم کند. این تکنیک به دو روش پارامتریک (مثل مدل‌های رگرسیون) و غیرپارامتریک (مثل مدل‌های هیستوگرام، خوشه‌بندی و نمونه‌برداری) انجام می‌شود. <br>
    &nbsp;&nbsp;&nbsp; * تجمیع در مکعب داده‌ای (Data Cube Aggregation): در این روش سعی می‌شود که داده‌های اصلی دیتاست در چند لایه (یا یک مکعب) قرار گیرد که باعث می‌شود حجم داده‌ها کاهش یابد. <br>
    &nbsp;&nbsp;&nbsp; * فشرده‌سازی (Data Compression): در این روش سعی می‌شود با رمزگذاری داده‌ها یا تغییر ساختار داده‌ها، حجم داده‌ها را پایین‌تر بیاوریم. <br>
    &nbsp;&nbsp;&nbsp; * گسسته‌سازی داده‌ها (Data Descritization): برخی از داده‌ها را می‌توان به دسته‌های مختلف تقسیم‌بندی کرد تا در هنگام تجزیه و تحلیل، به جای این که با حجم زیادی از داده‌های متنوع روبرو باشیم، مقادیر محدودتر و نزدیک‌تر به هم را داشته باشیم. مثلا به جای این که سن افراد را نگهداری کنیم، آن‌ها را به دسته‌های کودک، نوجوان، بزرگسال و پیر دسته‌بندی کنیم. یا به جای اینکه مقدار حقوق افراد را نگهداری کنیم، از دسته‌های حقوق کم، متوسط و زیاد استفاده کنیم. <br>
    این روش‌ها در ادامه و همچنین در فصل‌های بعدی کاملا توضیح داده خواهد شد. <br>
    <br>
    <br>
    <b>1. کاهش ابعاد (Dimensionality Reduction):</b><br>
    به ویژگی‌ها و فیلدهای مربوط به هر موجودیت یک دیتاست، ابعاد آن داده گفته می‌شود. پس به عبارتی ستون‌های موجود در یک دیتابیس می‌تواند معرف ابعاد آن دیتابیس باشد. مثلا وزن و سن و قد مربوط به یک شخص. به رکوردهایی از اطلاعات یک موجودیت، نقاط داده‌ای گفته می‌شود. پس نقاط داده‌ای، سطرهای جدول دیتابیس را مشخص می‌کند. <br>
    کاهش ابعاد داده‌ها یعنی حذف یک یا چند بعد (ویژگی) یا ادغام و ترکیب چند ویژگی و ایجاد یک ویژگی جدید. روش‌های متداول کاهش ابعاد به شرح زیر است: 
    &nbsp;&nbsp;&nbsp; <b>روش تحلیل مولفه‌های اصلی (Principal Component Analysis یا PCA):</b> در این روش ما سعی می‌کنیم از بین ویژگی‌هایی که وجود دارد، کاری کنیم که این ویژگی‌ها به تعداد کمتری ویژگی تبدیل شوند. یعنی ویژگی‌های جدیدی را استخراج کنیم. به عبارت دیگر تعداد ویژگی‌ها کمتر شده و متغیرهای جدید به‌وجود آیند. تلاش اصلی روش PCA این است که ویژگی‌هایی را با هم تلفیق و ترکیب کند که به هم وابسته‌اند. مثلا قد و وزن در یک دیتاست به هم وابسته‌اند و از ترکیب آن‌ها می‌توان یک ویژگی جدید ساخت. بنابراین PCA ویژگی‌های وابسته را با هم ترکیب می‌کند و این کار را می‌تواند در چندین مرحله انجام دهد تا در نهایت به دیتاستی برسد که تمام ویژگی‌های آن از هم مستقل هستند و هیچ وابستگی به هم ندارند. <br>
    &nbsp;&nbsp;&nbsp; * <b>روش تجزیه و تحلیل تفکیک خطی (Linear Discriminant Analysis یا LDA):</b> در این روش به کمک احتمالات تخمین می‌زنیم و بر اساس تفکیک‌پذیری داده‌ها ویژگی‌های جدیدی تولید می‌کنیم. کل تلاش روش LDA این است که در نهایت واریانس بین ویژگی‌های داده‌ای زیاد باشد و واریانس درون هر ویژگی (یعنی ویژگی‌هایی که با هم ترکیب شده‌اند) به حداقل برسد. <br>
    <br>
    <b>2. تجمیع در مکعب داده‌ای (Data Cube Aggregation):</b><br>
    در این روش سعی می‌شود که داده‌ها در لایه‌های مختلف در کنار هم قرار گیرند به صورتی که یک مکعب تشکیل شود. مثلا داده‌های مربوط به فروش چند کالا در فصل‌های مختلف سال در چند کشور را می‌توان در یک مکعب قرار داد. به طوریکه یک بعد مشخص‌کننده نوع کالا، یک بعد مشخص‌کننده فصل سال و یک بعد مشخص‌کننده نام کشور باشد، در این صورت با جمع زدن مقادیر فروش در هر سطر می‌توان اطلاعات مفیدی با حجم بسیار کم را استخراج کرد. 
</p>
